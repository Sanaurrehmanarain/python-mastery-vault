### Nested *if*

````python
age = 34

# nesting
if age >= 18:
    if age >=80:
        print("cannot drive")
    else:
        print("can drive")
else:
    print("cannot drive")
````

````python
# Nested if example
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
    
    # Nested if inside the first if
    if age >= 60:
        print("You are also a senior citizen.")
    else:
        print("You are not a senior citizen.")
else:
    print("You are a minor.")
````

**How it works:**
- First if checks if the person is 18 or older.
- If true, it goes inside and checks another condition: is the person 60 or older?
- If the first condition is false, it skips the nested check and prints "You are a minor."

````python
# Nested if example: Grading system
marks = int(input("Enter your marks: "))

if marks >= 50:
    print("You passed the exam.")
    
    if marks >= 75:
        print("You got a distinction!")
        
        if marks == 100:
            print("Perfect score! Excellent job!")
        else:
            print("Great work!")
    else:
        print("Good job, but you can aim higher.")
else:
    print("You failed. Better luck next time.")
````

**How it works:**
- First if checks if marks are 50 or more (pass).
- Inside that, another if checks if marks are 75 or more (distinction).
- Inside that, another if checks if marks are exactly 100 (perfect score).

### Equivalent Version Using elif (Flattened Logic)
This version avoids deep nesting while keeping the same outputs for each case.

````python
# Grading system using elif (equivalent outputs)
marks = int(input("Enter your marks: "))

if marks < 50:
    print("You failed. Better luck next time.")
elif marks < 75:
    print("You passed the exam.")
    print("Good job, but you can aim higher.")
elif marks == 100:
    print("You passed the exam.")
    print("You got a distinction!")
    print("Perfect score! Excellent job!")
else:
    # Here marks is >= 75 but not 100
    print("You passed the exam.")
    print("You got a distinction!")
    print("Great work!")
````

**Why use elif here?**
- It’s flatter and often easier to read.
- Each condition is mutually exclusive, so elif naturally fits.
- You still get the same messages as the nested if version.

WAP to check if a number entered by the user is odd or even.

````python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
````

````python
# 2nd method, Pthonic way :)
num = int(input("Enter a number: "))

print("Even" if num % 2 == 0 else "Odd")
````
