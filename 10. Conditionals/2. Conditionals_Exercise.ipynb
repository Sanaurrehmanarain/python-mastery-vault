## Exercise 1 - Nesting `if`/`else` blocks can be cumbersome

Galactic Exploration Agency astronauts are preparing to explore a new planet. The planet's environment can be classified as "ideal", "tolerable", "harsh", or "toxic". The type of environment determines how many hours the astronauts can be outside of their spaceship exploring the planet.

If the conditions are "ideal", astronauts can explore for 12 hours. If the conditions are "tolerable", they can explore for 8 hours. If the conditions are "harsh", they can explore for 3 hours. And, if the conditions are "toxic", astronauts can only explore for an hour.

### Part A

Using a nested `if`/`else` statement, write a program that updates the `hours_to_explore` variable with the correct number of hours based on the value of the `environment` variable.

**Solution**

```python
environment = "tolerable"  # or "ideal", "harsh", "toxic"
hours_to_explore = -1  # placeholder

if environment == "ideal":
    hours_to_explore = 12
else:
    if environment == "tolerable":
        hours_to_explore = 8
    else:
        if environment == "harsh":
            hours_to_explore = 3
        else:
            if environment == "toxic":
                hours_to_explore = 1

print(f"Because the environment is {environment}, astronauts can explore the planet for {hours_to_explore} hours.")
```

## Exercise 2 - Sometimes nesting is the best option

### 🚨 New Concept: When to nest conditionals

There are two main situations where nesting conditionals can make your code easier to read.

The first scenario occurs when you want to **execute intermediate actions** within a conditional block. For example...

```python
grade = 75  # Example value

if grade >= 60:
    print("Passed")
    if grade >= 90:
        print("With Distinction")
    elif grade >= 75:
        print("With Merit")
else:
    print("Failed")
```

Here we want to output that the grade is passing anytime it is at least 60. We can do this by including the intermediate print statement on line 4. Then we can go on to check whether the other conditions are met.

The second scenario where you may opt to nest conditionals is when you want to **ensure that some condition is met before continuing on to check other conditions**. For exmaple...

```python
age = 21  # Example value

if age >= 0:
    if age < 13:
        print("Child")
    elif age < 20:
        print("Teenager")
    elif age < 65:
        print("Adult")
    else:
        print("Senior")
else:
    print("Invalid age")
```

Here we want to check that `age` is a valid value (i.e. at least 0) before checking the rest of the conditions, since we would not want to continue our checks if the age was invalid.

Similarly, in the grades example both the "passed with distinction" and "passed with merit" require the grade to be at least 60. Thus, we should check that condition first.

### Part A

The mechanics need to check whether the fleet's spaceships are ready to launch. The launch sequence involves multiple steps that need to be executed sequentially:

1. Fuel Check: Ensure there's enough fuel.
2. Crew Check: Ensure the crew is present and ready.
3. Navigation Check: Ensure navigation systems are functioning.

If the fuel check fails, a warning should be issued immediately without checking further. Similarly, if the crew check fails, no further checks are needed.

Write a program using nested conditionals to perform this launch readiness check.

**Solution**

```python
fuel_ready = True  # Example value
crew_ready = False  # Example value
navigation_ready = True  # Example value

if fuel_ready:
    if crew_ready:
        if navigation_ready:
            print("Launch ready")
        else:
            print("Navigation check failed")
    else:
        print("Crew check failed")
else:
    print("Fuel check failed")
```

Note that the above and below code blocks are logically equivalent. However similar to Exercise 1 Part A, it is best practice to explicitly check for the success state. Putting the success state in the catch-all `else` block as it is below can lead to unintended behavior and bugs in programs.

```python
fuel_ready = True  # Example value
crew_ready = False  # Example value
navigation_ready = True  # Example value

if not fuel_ready:
    print("Fuel check failed")
elif not crew_ready:
    print("Crew check failed")
elif not navigation_ready:
    print("Navigation check failed")
else:
    print("Launch ready")
```

## Exercise 3 - The order of conditional checks matters

### 🚨 New Concept: The order of conditional checks matters

When creating conditional blocks, it is important to note that the order in which conditions are checked influences the output of a conditional block.

For example, the following code calculates a discount percentage based on the amount of money spent. If the buyer spends more than \$1000 they get a 20% discount, if they spend more than \$500 they get a 10% discount, otherwise they get a 5% discount.

```python
amount_spent = 1200

if amount_spent > 1000:
    discount = 20
elif amount_spent > 500:
    discount = 10
else:
    discount = 5

print(f"Discount applied: {discount}%")
```

This code applies the correct discount because it checks first to see if the `amount_spent` is greater than 1000. If it is not, then it checks to see if `amount_spent` is greater than 500. The `elif` statement functionally checks if `amount_spent` is greater than 500 but less than 1000.

However, if we swap the order of conditional checks like this...

```python
amount_spent = 1200

if amount_spent > 500:
    discount = 10
elif amount_spent > 1000:
    discount = 20
else:
    discount = 5

print(f"Discount applied: {discount}%")
```

We get the wrong result.

- Temperature <= 50°C: Safe, engine is operating within normal parameters.

Write a program that checks the engine temperature and prints out the appropriate action based on the priority order.

Change the order of checks and observe the difference in outputs.

```python
engine_temperature = 85  # Example value

if engine_temperature > 90:
    action = "Engine is overheating, mission abort required"
elif engine_temperature > 70:
    action = "Warning, cooling systems need to be activated"
elif engine_temperature > 50:
    action = "Caution, recommend reducing power output"
else:
    action = "Safe, engine is operating within normal parameters"

print(action)
```

Incorrect order of checks:

```python
# Reversed Order
if engine_temperature > 50:
    action = "Caution, recommend reducing power output"
elif engine_temperature > 70:
    action = "Warning, cooling systems need to be activated"
elif engine_temperature > 90:
    action = "Engine is overheating, mission abort required"
else:
    action = "Safe, engine is operating within normal parameters"

print(action)
```

## Exercise 4 - The ternary operator

### 🚨 New Concept: The ternary operator

The ternary operator provides a concise way to write conditional statements.

The ternary operator takes the form: **value_if_true `if` condition `else` value_if_false**

Instead of using a full if/else block, you can evaluate conditions inline like this...

```python
age = 17
status = "adult" if age >= 18 else "minor"
print(status)
```

For simple conditional statements, the ternary operator can make code more readable. However for more complex conditional blocks, the traditional `if`/`elif`/`else` structure is best.

### Question

The Galactic Exploration Agency has a policy that a spaceship's food supply is evaluated before every mission. If the supply level is above 50%, it's considered "Sufficient"; otherwise, it's "Insufficient."

Write a program using the ternary operator to classify the food supply level.

**Solution**

```python
food_supply = 75  # Example value

status = "Sufficient" if food_supply > 50 else "Insufficient"
print(status)
```

### Question
User will enter a floating point number let say 238.915. Your task is to find out the integer portion before the point (in this case 238) and then check if that integer portion is an even number or not?

Solution 1 (Simple)

```python
# Ask user for a floating-point number
x = float(input("Enter a real number: "))

# Get the integer part before the decimal
integer_part = int(x)  # int() truncates toward zero

# Check if it's even or odd
if integer_part % 2 == 0:
    print(f"Integer part: {integer_part} -> Even")
else:
    print(f"Integer part: {integer_part} -> Odd")
```

Solution 2 (With validation)

```python
while True:
    raw = input("Enter a real number: ").strip()
    try:
        x = float(raw)           # try to convert to float
        break                    # success -> exit the loop
    except ValueError:
        print("Invalid number. Please try something like 238.915, -3.7, or 51.")

# Get the integer portion before the decimal (truncates toward zero)
integer_part = int(x)

# Check even/odd
if integer_part % 2 == 0:
    print(f"Integer part: {integer_part} -> Even")
else:
    print(f"Integer part: {integer_part} -> Odd")
```
