There are a few best practices in Python to make control flow efficient and readable so your code doesn’t run unnecessary if statements and exits early when it can.

Below are the most useful patterns—with short, focused examples you can copy and run.

## 1) Use if / elif / else for mutually exclusive conditions
**Problem:** Using multiple independent if statements means more than one block may run (and every condition is checked).

**Solution:** Use elif so only the first matching branch runs—Python stops checking after a match.

❌ Not ideal:

```python
marks = 85

if marks >= 50:
    print("Pass")
if marks >= 75:
    print("Distinction")
if marks == 100:
    print("Perfect")
# Output: Pass, Distinction   (multiple ran)
```

✅ Best practice (elif short-circuits):

```python
marks = 85

if marks < 50:
    print("Fail")
elif marks < 75:
    print("Pass")
elif marks == 100:
    print("Pass + Distinction + Perfect")
else:
    print("Pass + Distinction")
# Only one branch runs
```

## 2) Use early returns (guard clauses) inside functions
Early returns let you exit as soon as you know the answer, avoiding deep nesting and unnecessary checks.

```python
def classify_marks(marks: int) -> list[str]:
    # Guard clauses
    if marks < 0 or marks > 100:
        return ["Invalid marks"]

    if marks < 50:
        return ["You failed. Better luck next time."]

    messages = ["You passed the exam."]

    if marks < 75:
        messages.append("Good job, but you can aim higher.")
        return messages

    messages.append("You got a distinction!")
    if marks == 100:
        messages.append("Perfect score! Excellent job!")
    else:
        messages.append("Great work!")
    return messages
```

## 3) Use short-circuiting with and / or
Python evaluates boolean expressions left to right and stops early when the result is known.

- A and B: if A is False, Python won’t evaluate B.
- A or B: if A is True, Python won’t evaluate B.

```python
def is_valid_user(user):
    return user is not None and user.is_active and user.has_paid  # stops as soon as a check fails
```

## 4) In loops, use break, continue, and for-else
- `break: `stop the loop early when you’ve found what you need.
- `continue:` skip to next iteration.
- `for ... else:` the else runs **only if the loop didn’t break**.

```python
nums = [3, 5, 12, 7, 8]

# Find the first even number > 10
for n in nums:
    if n % 2 == 0 and n > 10:
        print("Found:", n)
        break
else:
    print("No even number > 10 found")
```

## 5) Consider match / case for clear branching
Structural pattern matching can replace messy chains.

```python
def http_status(code: int) -> str:
    match code:
        case 200 | 201:
            return "Success"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500 | 502 | 503:
            return "Server Error"
        case _:
            return "Unknown"

http_status(201)
```

## 6) Use any() / all() for clean condition groups
These short-circuit too, and they’re expressive.

```python
# All checks must pass
if all([user.is_active, user.email_verified, user.has_paid]):
    grant_access()

# Any one condition triggers action
if any([x > 10 for x in nums]):
    print("At least one number > 10")
```

#### Quick side-by-side example (before vs after)
Before (multiple checks + redundant work):

```python
marks = int(input("Enter marks: "))

if marks >= 50:
    print("You passed the exam.")
if marks >= 75:
    print("You got a distinction!")
if marks == 100:
    print("Perfect score! Excellent job!")
elif marks < 50:
    print("You failed. Better luck next time.")
```

After (mutually exclusive + early return):

```python
def feedback(marks: int) -> list[str]:
    if marks < 0 or marks > 100:
        return ["Invalid marks"]

    if marks < 50:
        return ["You failed. Better luck next time."]

    # Passed
    if marks < 75:
        return ["You passed the exam.", "Good job, but you can aim higher."]

    # Distinction
    if marks == 100:
        return ["You passed the exam.", "You got a distinction!", "Perfect score! Excellent job!"]

    return ["You passed the exam.", "You got a distinction!", "Great work!"]

print("\n".join(feedback(int(input("Enter marks: ")))))
```
