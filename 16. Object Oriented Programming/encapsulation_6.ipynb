```python
class car:
    def __init__(self, speed, color):
         self.speed = speed
         self.color = color
         
ford = car(150, "red")
honda = car(170, "black")
audi = car(220, "blue")
bmw = car(250, "green")

ford.speed = 190 # anybody can change any value in the code (without encapsulation)

print(ford.speed)
print(ford.color)
```

```python
# ENCAPSULATION
class car:
    def __init__(self, speed, color):
         self.__speed = speed # putting double underscore before the varaible name will make it private and no one can alter or change it
         self.__color = color
    
    # set mathod
    def set_speed (self, value):
         self.__speed = value

    # get method
    def get_speed(self):
         return self.__speed
    
      # set mathod
    def set_color (self, value):
         self.__color = value

    # get method
    def get_color(self):
         return self.__color
             
ford = car(150, "red")
honda = car(170, "black")
audi = car(220, "blue")
bmw = car(250, "green")

ford.set_speed(300) # the speed is set to 300 and will not change now
# ford.__speed = 400 # now you can't change the speed as it is a private variable now


print(ford.get_speed())
print(ford.get_color()) 
```

```python
class Rectangle:
    def __init__(self, height, width):
        self.__height = height # make the height private
        self.__width = width # make the width private
    
    def set_height(self, height): # setter
        self.__height = height

    def get_height(self): # getter
        return self.__height
    
    def set_width(self, width): # setter
        self.__width = width

    def get_width(self): # getter
        return self.__width
    
    def area(self):
        return self.__height * self.__width


rect1 = Rectangle(20, 60)
rect2 = Rectangle(50, 40)

print(rect1.area())
print(rect2.area())      
```
