***PYTHON INHERITANCE***
- Classes in Python can be extended, creating new classes which retain characteristics of the base class.
- This process, known as inheritance.
- Involves a superclass and a subclass.
- The subclass inherits the members of the superclass, on top of which it can add its own members.

#### Example of multiple inheritance

````python
class Polygon: # super class
    __width = None # private member variable
    __height = None # private member variable

    def set_values(self, width, height): # setter method
        self.__width = width
        self.__height = height

    def get_width(self): # getter method
        return self.__width
    
    def get_height(self): # getter method
        return self.__height
    
class Shape: # super class
    __color = None

    def set_color(self, color):
        self.__color = color

    def get_color(self):
        return self.__color

class Rectangle(Polygon, Shape): # subclass | Rectangle is inheritting from Polygon & Shape classes
    def area(self):
        return self.get_width() * self.get_height()
    
class Triangle(Polygon, Shape):  # subclass | Triangle is inheritting from Polygon & Shape classes
    def area(self):
        return self.get_width() * self.get_height() / 2
    
rect = Rectangle() # rect is an instance of Rectangle() class
tri = Triangle() # tri is an instance of Triangle() class

rect.set_values(30, 10)
tri.set_values(50, 40)

rect.set_color("red")
tri.set_color("blue")

print(rect.area())
print(tri.area())
print(rect.get_color())
print(tri.get_color())
````

We didn't initialize a constructor `(__init__ method)` in the Polygon class because the set_values method is explicitly used to assign values to the private attributes __width and __height.

If we had used `__init__`, we would need to pass values when creating an instance, like:

```python
class Polygon:
    def __init__(self, width, height):
        self.__width = width
        self.__height = height

Then, we would create objects like:

rect = Rectangle(30, 10)
tri = Triangle(50, 40)
```
Since the current implementation uses a separate set_values method instead of `__init__`, objects are created first and initialized later. This provides flexibility, allowing objects to be instantiated without immediately defining their dimensions.
