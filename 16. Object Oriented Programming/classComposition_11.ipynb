## Definition of Class Composition
Class composition is a design principle in object-oriented programming where one class contains an instance of another class as a member, instead of inheriting from it. It follows the "has-a" relationship, meaning an object of one class "has" an object of another class.

**Composition** represents `part-of` relationship.

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition: Car "has-a" Engine

    def drive(self):
        self.engine.start()
        print("Car is moving")

my_car = Car()
my_car.drive()
```

#### Key Points:
- "Has-a" Relationship – A class contains another class as an attribute.
- Better Code Reusability – More flexible than inheritance since components can be replaced independently.
- Encapsulation – The contained class (Engine) is hidden within the containing class (Car).
- Reduces Code Complexity – Avoids deep inheritance trees.

```python
class Salary:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus

    def annual_salary(self):
        return (self.pay * 12) + self.bonus
    
class Employee:
    def __init__(self, name, age, pay, bonus):
        self.name = name
        self.age = age
        self.obj_salary = Salary(pay, bonus)

    def total_salary(self):
        return self.obj_salary.annual_salary()

emp = Employee("max", 25, 15000, 6000)
print(emp.total_salary())
        
```
