```python
class Hello:
    def __init__(self, name):
        self.a = 10
        self._b = 20
        self.__c = 30

hello = Hello("name")
print(hello.a)
print(hello._b)
print(hello.__c)
```

You got an error, why?.. because you can print a and _b but you can not call/ access __c (private member variable) outside the class.

```python
class Hello:
    def __init__(self, name):
        self.a = 10 # public member variable, you can use this inside or outside the class
        self._b = 20
        self.__c = 30
    
    def public_method(self):
        print(self.a)
        print(self.__c)
        print("public")

hello = Hello("name")
print(hello.a)
print(hello._b)
hello.public_method()
```

Now you didn't get any error, becuase you can use your private member variables inside the class or any method of the class.

```python
class Hello:
    def __init__(self, name):
        self.a = 10 
        self._b = 20
        self.__c = 30
    
    def public_method(self):
        print(self.a)
        print(self.__c)
        print("public")

    def __private_method(self): # private method
        print("private")

hello = Hello("name")
print(hello.a)
print(hello._b)
hello.public_method()
hello.__private_method
```

why you get an error? Because same as private variables, you can not use private method outside the class. But you can use it inside the class. See the example below.

```python
class Hello:
    def __init__(self, name):
        self.a = 10 
        self._b = 20
        self.__c = 30
    
    def public_method(self):
        print(self.a)
        print(self.__c)
        print("public")
        self.__private_method() # use of private method inside the class

    def __private_method(self): # private method
        print("private")

hello = Hello("name")
print(hello.a)
print(hello._b)
hello.public_method()
```
