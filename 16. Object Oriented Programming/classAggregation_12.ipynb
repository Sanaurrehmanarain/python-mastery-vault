**Aggregation represents `Has-A` relationship.

```python
class Salary:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus

    def annual_salary(self):
        return (self.pay * 12) + self.bonus
    
class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.obj_salary = salary

    def total_salary(self):
        return self.obj_salary.annual_salary()
    
salary = Salary(15000, 8000)
emp = Employee("Max", 25, salary)
print(emp.total_salary())
```

#### Explanation of the above code
self.obj_salary.annual_salary()

In the Employee class, self.obj_salary is an instance of the Salary class that is passed as an argument during object creation. This demonstrates aggregation, where one class (Employee) contains an instance of another class (Salary).

When we call:

emp.total_salary()

This executes:

return self.obj_salary.annual_salary()

Here's what happens step by step:

1. self.obj_salary refers to the Salary object (salary = Salary(15000, 8000)) that was passed to Employee.
2. self.obj_salary.annual_salary() calls the annual_salary() method of the Salary class.
3. annual_salary() calculates the salary:

(15000×12)+8000 = 188000

4. The computed value 188000 is returned.

Thus, emp.total_salary() returns 188000, which is printed.
