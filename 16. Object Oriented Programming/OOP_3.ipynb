# Explanation of Object-Oriented Programming Concepts

## 1. **Class (`class student`)**
- A **class** is a blueprint for creating objects.
- In this case, `student` is a class representing students.

## 2. **Class Attribute (`college_name = "ABC College"`)**
- A **class attribute** is shared across all instances of the class.
- Any object of `student` can access this attribute.

## 3. **Constructor (`__init__`)**
- A special method that gets called automatically when an object is created.
- Initializes **instance attributes** (`self.name` and `self.marks`).

## 4. **Instance Attributes (`self.name`, `self.marks`)**
- Unique to each object (student).
- `self.name = name`: Assigns a value to the `name` attribute for each object.
- `self.marks = marks`: Assigns marks to the student.

## 5. **Methods (`welcome()`, `get_marks()`)**
- Functions defined inside a class that operate on objects.
- `welcome(self)`: Prints a welcome message with the student's name.
- `get_marks(self)`: Returns the marks of the student.

## 6. **Object (`s1 = student("karan", 97)`)**
- `s1` is an **instance** of the `student` class.
- `"karan"` and `97` are passed to the constructor.

## 7. **Method Calls (`s1.welcome()`, `s1.get_marks()`)**
- `s1.welcome()`: Calls the `welcome` method, printing `"Welcome student, karan"`.
- `print(s1.get_marks())`: Calls `get_marks()` and prints `97`.


Q: Create student class that takes name & marks of 3 subjects as arguments in constructor. Then create a method to print the average.

````python
class student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_avg(self):
        sum = 0
        for val in self.marks:
            sum += val
        print(f"Hi {self.name} your avg score is: {sum / 3}")


s1 = student("Tony Stark", [98, 86, 89])

s1.get_avg()
````

### Static Method
Methods that don't use the self parameter (work at class level)

````python
class MathOperations:
    @staticmethod # Decorator
    def add(a, b):
        return a + b

# Calling the static method without creating an object
result = MathOperations.add(5, 3)
print(result)  
````

- Decorators allow us to wrap another function in order to extend the behaviour of the wrapped function, without permanently modifying it.
- It behaves like a normal function but is inside the class.
- No need to create an object; it is called using the class name (MathOperations.add(5, 3)).

## 1. Abstraction
Hiding the implementation details of a class and only showing the essential features to the user.

````python
class car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False

    def start(self):
        self.clutch = True
        self.acc = True
        print("Car Started..")

car1 = car()
car1.start()

# The next example in "Encapsulation" is a better representative 
# of "Abstraction" as that is using private attributes.
````

## 2. Encapsulation
Wrapping data and functions into a single unit (object).

### Private (like) attributes & methods
#### Conceptual Implementations in python
Private attributes & methods are meant to be used only wihtin the class and are not accessible from outside the class.

````python
class Car:
    def __init__(self):
        self.__acc = False  # Private attribute
        self.__brk = False  # Private attribute
        self.__clutch = False  # Private attribute

    def start(self):
        self.__clutch = True
        self.__acc = True
        print("Car Started..")

    def is_running(self):
        return self.__acc  # Controlled access

car1 = Car()
car1.start()

# Trying to access private attribute directly (not recommended)
# print(car1.__acc)  # This will raise an AttributeError
print(car1.is_running())  # Correct way to check if the car is running
````

✅ Encapsulation: Uses private attributes (__acc, __brk, __clutch) to prevent direct access.

✅ Abstraction: The user only sees the start() method and doesn't worry about internal mechanics.

✅ Controlled Access: The is_running() method provides a safe way to check if the car is running.

## 3. Inheritance
When one class (child/derived) derives the properties & methods of another class (parent/base).
```python
class Car:
    ....

class ToyotaCar(Car):
    ....

````python
# Single Inheritance
class Car:
    @staticmethod
    def start():
        print("car started...")

    @staticmethod
    def stop():
        print("car stopped...")

class ToyotaCar(Car):
    def __init__(self, name):
        self.name = name

car1 = ToyotaCar("Fortuner")
car2 = ToyotaCar("Prius")

print(car1.name)
car1.start()
# Why we didnot use print(car1.start())?. 
# The start() method is a staticmethod, meaning it does not 
# return anything but directly prints a message. 
# When you call print(car1.start()), it prints "car started..." 
# and then prints None because start() does not return a value.
````

### Inheritance types:
- Single Inheritance
- Multi-level Inheritance
- Multiple Inheritance

````python
# Multilevel Inheritance
# In multilevel inheritance, a class inherits from a derived class.
class Car:
    @staticmethod
    def start():
        print("Car started...")

    @staticmethod
    def stop():
        print("Car stopped...")

class ToyotaCar(Car):  # Inherits from Car
    def __init__(self, name):
        self.name = name

class HybridToyotaCar(ToyotaCar):  # Inherits from ToyotaCar
    def __init__(self, name, battery_life):
        super().__init__(name)
        self.battery_life = battery_life

car1 = HybridToyotaCar("Prius Hybrid", "10 hours")
print(car1.name)           
print(car1.battery_life)   
car1.start()              
````

````python
# Multiple Inheritance
# In multiple inheritance, a class inherits from two or more parent classes.
class Engine: # parent 1
    @staticmethod
    def engine_type():
        print("Hybrid Engine")

class Car: # parent 2
    @staticmethod
    def start():
        print("Car started...")

    @staticmethod
    def stop():
        print("Car stopped...")

class ToyotaCar(Car, Engine):  # Inherits from both Car and Engine, **child class**
    def __init__(self, name):
        self.name = name

car1 = ToyotaCar("Prius")
print(car1.name)  
car1.start()        
car1.engine_type()  
````

#### Super method
super() method is used to access methods of the parent class. Means The super() method is used to call a method from the parent class inside the child class.

````python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} car started...")

class ToyotaCar(Car):
    def __init__(self, brand, name):
        super().__init__(brand)  # Calling the parent class constructor
        self.name = name

    def show_details(self):
        print(f"Brand: {self.brand}, Model: {self.name}")

car1 = ToyotaCar("Toyota", "Fortuner")
car1.start()         
car1.show_details() 
````

#### class method
A class method is bound to the class & receives the class as an implicit first argument. In other words A class method is defined using @classmethod and takes cls as the first parameter. It can be used to modify class attributes or create instances in an alternative way.

Note - static method can't access or modify class state & generally for utility.
```python
class Student:
    @classmethod # decorator
    def college(cls):
        pass

````python
class Car:
    brand = "Toyota"  # Class attribute

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_brand(cls, new_brand):
        cls.brand = new_brand  # Modifies the class attribute

    @classmethod
    def from_string(cls, car_string):
        name = car_string.split("-")[1] # Extract the car model name from a string formatted as "Brand-Model".
        return cls(name)

# Using class method to change the class attribute
Car.change_brand("Honda")
print(Car.brand)  

# Using class method as an alternative constructor
car1 = Car.from_string("Honda-Civic")
print(car1.name)  
print(car1.brand) 
````

#### Property
We use @property decorator on any method in the class to use the method as a property. The @property decorator is used to define getter, setter, and deleter methods for class attributes, allowing controlled access.

````python
class student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math

    @property
    def percentage(self):
        return str((self.phy + self.chem + self.math) / 3) + "%"

std1 = student(98, 97, 99)
print(std1.percentage)

std1.phy = 86
print(std1.percentage)
````

````python
class Car:
    def __init__(self, brand, model):
        self._brand = brand  # Private attribute (convention)
        self._model = model  

    @property
    def brand(self):
        """Getter method for brand"""
        return self._brand

    @brand.setter
    def brand(self, new_brand):
        """Setter method for brand"""
        if new_brand:
            self._brand = new_brand
        else:
            raise ValueError("Brand name cannot be empty")

    @brand.deleter
    def brand(self):
        """Deleter method for brand"""
        print("Deleting brand...")
        del self._brand

# Creating an instance
car1 = Car("Toyota", "Fortuner")

# Accessing property (calls getter)
print(car1.brand)  

# Modifying property (calls setter)
car1.brand = "Honda"
print(car1.brand)  

# Deleting property (calls deleter)
del car1.brand  
````

## 4. Polymorphism
Polymorphism allows different classes to have methods with the same name but different implementations.
#### Operator Overloading
When the same operator is allowed to have different meaning according to the context.

### Operators & Dunder functions

| Operation    | Operator | Dunder Method      |
|--------------|----------|--------------------|
| # addition   | a + b    | a.``__add__``(b)       |
| # subtraction| a - b    | a.``__sub__``(b)       |
| # multiplication| a * b | a.`__mul__`(b)       |
| # division   | a / b    | a.`__truediv__`(b)   |
| # modulo     | a % b    | a.`__mod__`(b)       |

````python
# Here "+" has a different meaning according to the context.
print(1 + 3) # here + is addtion 
print("Hello" + " World") # here + mean concatenate
print([1, 2, 3] + [4, 5, 6]) # here + mean merge
````

````python
# 1. Example using complex numbers
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real, "i +", self.img, "j")

num1 = Complex(1, 3)
num1.showNumber()

num2 = Complex(4, 6)
num2.showNumber()
````

````python
# 2. Now we want to add complex nums using normal function
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real, "i +", self.img, "j")

    def add(self, num2):
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        return Complex(newReal, newImg)

num1 = Complex(1, 3)
num1.showNumber()

num2 = Complex(4, 6)
num2.showNumber()

num3 = num1.add(num2)
num3.showNumber()
````

````python
# 3. If i want to add simply by num3 = num1 + num2, we can do it by using 
# dunder __add__ function.
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real, "i +", self.img, "j")

    def __add__(self, num2):
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        return Complex(newReal, newImg)

num1 = Complex(1, 3)
num1.showNumber()

num2 = Complex(4, 6)
num2.showNumber()

num3 = num1 + num2 # now we can do this 
num3.showNumber()
````

Example 1: Method Overriding (Runtime Polymorphism)

````python
class Car:
    def start(self):
        print("Car is starting...")

class Toyota(Car):
    def start(self):
        print("Toyota is starting...")

class BMW(Car):
    def start(self):
        print("BMW is starting...")

# Using polymorphism
cars = [Toyota(), BMW()]

for car in cars:
    car.start()  
````

- The start() method is overridden in both Toyota and BMW classes.
- The method call depends on the object type at runtime.

Example 2: Method Overloading (Compile-Time Polymorphism Simulation)

Python does not support true method overloading, but it can be simulated using default arguments.

````python
class Car:
    def speed(self, max_speed=None):
        if max_speed:
            print(f"Car is running at {max_speed} km/h")
        else:
            print("Car is running...")

car = Car()
car.speed()         
car.speed(120)      
````

The same method speed() works with and without an argument.

Example 3: Polymorphism with Functions

````python
class Toyota:
    def fuel_type(self):
        return "Petrol"

class Tesla:
    def fuel_type(self):
        return "Electric"

def car_fuel(car):
    print(car.fuel_type())

car_fuel(Toyota())  
car_fuel(Tesla())   
````

The function car_fuel() works with different classes having the same method name (fuel_type()).

Polymorphism ensures that the same function or method name can be used for different types of objects.

Q: Define a Circle class to create a circle with radius r using the constructor. 

Define an Area() method of the class which calculates the area of the circle.

Define a Perimeter() method of the class which allows you to calculate the perimeter of the circle.

````python
class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return (22 / 7) * self.radius ** 2
    
    def perimeter(self):
        return 2 * (22 / 7) * self.radius

c1 = Circle(21)
print(c1.area())
print(c1.perimeter())
````
