# Difference Between Composition and Aggregation

Both **composition** and **aggregation** are types of object relationships in OOP, where one class contains another. The key difference lies in **ownership and dependency**.

| Feature        | Composition | Aggregation |
|--------------|-------------|-------------|
| **Definition**  | Strong relationship where the contained object **cannot** exist independently of the container. | Weak relationship where the contained object **can** exist independently. |
| **Lifespan**    | The contained object is **destroyed** when the container object is destroyed. | The contained object **remains** even if the container is destroyed. |
| **Example**     | A `Car` class contains an `Engine`—if the car is destroyed, the engine is too. | A `Team` class contains `Players`—players can exist outside the team. |
| **Implementation** | The container class **creates** the contained object inside its constructor. | The contained object is **passed as an argument** to the container class. |

Example of Composition

```python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Car:
    def __init__(self, model, horsepower):
        self.model = model
        self.engine = Engine(horsepower)  # Creating Engine inside Car

car = Car("Tesla", 500)
```

Here, the Engine is part of Car and cannot exist independently.

Example of Aggregation

```python
class Player:
    def __init__(self, name):
        self.name = name

class Team:
    def __init__(self, team_name, players):
        self.team_name = team_name
        self.players = players  # Players exist independently

player1 = Player("Alice")
player2 = Player("Bob")
team = Team("Warriors", [player1, player2])
```

Here, Player objects exist before being added to Team, so they can exist independently.

Key Takeaway
- Use composition when the contained object cannot function independently (e.g., Engine inside a Car).
- Use aggregation when the contained object can exist on its own (e.g., Players in a Team).
