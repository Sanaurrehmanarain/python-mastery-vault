An **abstract** method in Python OOP is a method declared in a base class but without implementation. It forces subclasses to provide their own implementation. Abstract methods are defined using the @abstractmethod decorator from the abc module.

```python
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract base class
    @abstractmethod
    def make_sound(self):
        pass  # No implementation

class Dog(Animal):
    def make_sound(self):
        return "Bark"

# animal = Animal()  # Error: Cannot instantiate abstract class
dog = Dog()
print(dog.make_sound())  # Output: Bark
```

Explanation:

1. Animal is an abstract class.
2. make_sound is an abstract method, so any subclass must override it.
3. Cannot instantiate an abstract class directly.
4. Dog class implements make_sound(), so it can be instantiated.

```python
from abc import ABC, abstractmethod
class Shape(ABC): # super class
    @abstractmethod
    def area(self): pass

    @abstractmethod
    def perimeter(self): pass

class Square(Shape): # subclass
    def __init__(self, side):
        self.__side = side

    def area(self):
        return self.__side * self.__side
    
    def perimeter(self):
        return 4 * self.__side
    
square = Square(5)
print(square.area())
print(square.perimeter())
```
