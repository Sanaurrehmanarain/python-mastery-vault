#### super()
super() is a built-in function in Python that allows a subclass to call methods from its parent (or superclass). It is commonly used in inheritance to invoke a method from the base class without explicitly naming it.

```python
# Calling Parent Class Methods:
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        super().show()  # Calls the 'show' method of Parent
        print("Child method")

obj = Child()
obj.show()
```

```python
# Calling Parent’s __init__ Constructor:
class Parent:
    def __init__(self, name):
        self.name = name
        print("Parent constructor called")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Calls Parent's constructor
        self.age = age
        print("Child constructor called")

obj = Child("Alice", 25)
```
