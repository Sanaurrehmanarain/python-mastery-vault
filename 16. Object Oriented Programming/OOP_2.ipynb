```python
class car:
    pass # we use pass to make an empty class

ford = car() # ford, honda, etc str objects or instances of the class car.
honda = car()
audi = car()
bmw = car()

ford.speed = 150 # speed is a attribute or variables that holds data
honda.speed = 170
audi.speed = 220
bmw.speed = 250

ford.color = "red" # color is a attribute
honda.color = "black"
audi.color = "blue"
bmw.color = "green"

print(ford.speed)
print(ford.color)


# class car with __init__ method you will see in init_method_4 file.
```
