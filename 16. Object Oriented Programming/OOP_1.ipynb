# Object-Oriented Programming
### Key Characteristics
- **Languages:** C++, Java, Python
- **Core Unit:** Class
- **Focus**: Creating and interacting with objects.
- **Object Definition:** A single software unit that combines data and methods.
- **Attributes:** Represent the object's data. In other words, they are variables that hold data.
- **Methods:** Represent the functions or behaviors associated with the object. In other words, functions inside the class are called methods.
### Example: Travel Application (Object-Oriented Approach)

| **Aspect**              | **Procedural Programming**            | **Object-Oriented Programming**     |

By organizing logic around objects and their interactions, object-oriented programming enables better scalability, maintainability, and reusability in software development.

In Python, a `class` is a blueprint for creating objects. It defines the structure and behavior of objects by grouping data (attributes) and functionality (methods) together. A class acts as a template for objects, encapsulating data and methods to create reusable code.

#### Key Concepts of Classes in Python
* Attributes: Variables that hold data associated with an object. They are defined inside the class.
* Methods: Functions that define the behavior of an object. They operate on the attributes of the object.
* Objects: Instances of a class created using the class blueprint. Each object has its own set of attributes and can use the methods defined in the class

````python
# Syntax for Defining a Class
class ClassName:
    # Class attributes (shared by all instances)
    class_attribute = "shared_value"

    def __init__(self, attribute1, attribute2):
        # Instance attributes (unique to each object)
        self.attribute1 = attribute1
        self.attribute2 = attribute2

    def method_name(self):
        # Method to define behavior
        return f"Attributes: {self.attribute1}, {self.attribute2}"
````

````python
# Example: A Simple Class
class Person:
    # Class attribute
    species = "Human"

    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

    def greet(self):
        # Method
        return f"Hello, my name is {self.name} and I am {self.age} years old."

# Creating objects
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Accessing attributes and methods
print(person1.greet())  # Output: Hello, my name is Alice and I am 25 years old.
print(person2.greet())  # Output: Hello, my name is Bob and I am 30 years old.
````

#### Key Features of Classes in Python
1. Encapsulation: Groups data and behavior into a single unit.
2. Inheritance: Allows one class to inherit attributes and methods from another class.
3. Polymorphism: Allows methods to be overridden in child classes.
4. Abstraction: Hides implementation details and exposes only the necessary functionality.
#### Special Method: `__init__`
The `__init__` method is a constructor in Python. It initializes the attributes of an object when the object is created.

````python
class Example:
    def __init__(self, value):
        self.value = value  # Initialize the 'value' attribute

obj = Example(10)
print(obj.value)  # Output: 10
````

In Python, classes are a cornerstone of object-oriented programming, enabling efficient organization and reuse of code.
