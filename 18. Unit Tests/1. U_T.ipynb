# Unit Test
## 1. What is a Unit Test?

A **unit test** is a way to test small, individual parts of your code (called units), usually functions or classes, to make sure they work as expected.

Think of it as asking: *"Does this function always give the right output for different inputs?"*

Unit testing helps you:

- Catch bugs early

- Prevent breaking old code when adding new features

- Build confidence that your program works as intended

## Python’s Built-in Testing Module

Python has a built-in library called `unittest`.

A test is written inside a class that inherits from `unittest.TestCase`.


## 2. A Simple Function (to be tested)

```python
def add(a, b):
    return a + b

def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero is not allowed")
    return a / b
```

## 3. Writing a Unit Test

```python
import unittest

class TestMathFunctions(unittest.TestCase):

    def test_add(self):
        self.assertEqual(add(2, 3), 5)   # 2 + 3 = 5
        self.assertEqual(add(-1, 1), 0)  # -1 + 1 = 0

    def test_divide(self):
        self.assertEqual(divide(10, 2), 5)   # 10 / 2 = 5
        self.assertRaises(ValueError, divide, 10, 0)  # Division by zero
```

## 4. Running Tests in Jupyter Notebook

In a script, you’d normally run unittest.main(), but in Jupyter this can cause issues.
Instead, use this:

```python
unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestMathFunctions))
```

## 5. Common Assertions

| Method                             | Description                    | Example                                   |
| ---------------------------------- | ------------------------------ | ----------------------------------------- |
| `assertEqual(a, b)`                | Check if `a == b`              | `assertEqual(2+2, 4)`                     |
| `assertNotEqual(a, b)`             | Check if `a != b`              | `assertNotEqual(2+2, 5)`                  |
| `assertTrue(x)`                    | Check if `x` is True           | `assertTrue(5 > 3)`                       |
| `assertFalse(x)`                   | Check if `x` is False          | `assertFalse(3 > 5)`                      |
| `assertRaises(error, func, *args)` | Check if function raises error | `assertRaises(ValueError, divide, 10, 0)` |

## 6. Another Example: Testing a Multiply Function

```python
def multiply(a, b):
    return a * b

class TestMultiply(unittest.TestCase):
    def test_positive(self):
        self.assertEqual(multiply(3, 4), 12)

    def test_negative(self):
        self.assertEqual(multiply(-2, 5), -10)

unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestMultiply))
```
