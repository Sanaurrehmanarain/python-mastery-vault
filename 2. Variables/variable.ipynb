# VARIABLES

```python
miles_hiked = 1.5
oz_consumed = 12

total_miles = 4.73
total_oz_of_water = 40

current_cr = oz_consumed / miles_hiked # current consumption rate
print(f"Current water consumption rate (oz per mile): {current_cr}") # method 1
print("Current water consumption rate (oz per mile):", current_cr) # method 2
print("Current water consumption rate (oz per mile): " + str(current_cr)) # method 3
```

```python
max_cr = total_oz_of_water / total_miles # maximum consumption rate
print(f"Maximum water consumption rate (oz per mile): {max_cr}")
```

```python
blaze_wins = 13
titans_wins = 12

blaze_text = "Blaze wins: "
titan_text = "Titan wins: "

output = blaze_text + str(blaze_wins) +", " + titan_text + str(titans_wins) # method 1

print(output)

print(f"Blaze wins: {blaze_wins}, Titan wins: {titans_wins}") # method 2. Using f strings. f means formatted strings literals, this is a better version.
```

### To assign multiple values to multiple variables on the same line of code.

```python
# Assigns "Red" to x, "Green" to y, and "Blue" to z
x, y, z = "Red", "Green", "Blue"
print(f"x = {x}, y = {y}, z = {z}")
```

```python
# Assigns purple to a, b, and c
a = b = c = "Purple"
print(f"a = {a}, b = {b}, c = {c}")
```
