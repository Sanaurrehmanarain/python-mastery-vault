This one is a bit tricky! Remember, Python does not allow us to concatenate strings and numeric data directly. We must cast our numeric variables to strings with the `str()` function.

### 🚨 New Concept: F-strings

While writing code like `blaze_score_text + str(blaze_wins) + ", " + titans_score_text + str(titans_wins)` does work, it is not the most streamlined way to embed variable values in strings. To improve upon this, Python provides f-strings (also called formatted string literals). We can create an f-string by simply writing the character `f` before the first quotation marks of the string. Then, we can include variables anywhere within the string by placing them within curly braces `{}`. For example...

```python
month = "January"
day = 1

"Today is " + month + " " + str(day) + "!"
```

Can be written with a single f-string like this...

```python
f"Today is {month} {day}!"
```

Using f-strings makes code more readable than using the concatenation operator `+` for joining strings. Additionally, when using f-strings, we are not required to cast numeric values to strings.

### Question

Rewrite the string `Blaze wins: 13, Titan wins: 12` using a single f-string.

**Solution**

With the score text variables...

```python
f"{blaze_score_text}{blaze_wins}, {titans_score_text}{titans_wins}"
```

Or without using the score text variables...

```python
f"Blaze wins: {blaze_wins}, Titan wins: {titans_wins}"
```

## Exercise 2 - Multiple assignments

### 🚨 New Concept: Multiple assignments

Python allows us to assign multiple values to multiple variables on the same line of code. This is a convenient way to make your code more concise. For example...

```python
# Assigns "Red" to x, "Green" to y, and "Blue" to z
x, y, z = "Red", "Green", "Blue"
f"x = {x}, y = {y}, z = {z}"
```

You can also assign the same value to each variable like this...

```python
# Assigns purple to a, b, and c
a = b = c = "Purple"
f"a = {a}, b = {b}, c = {c}"
```

## Exercise 3 - Variable mutability and reassignment

### 🚨 New Concept: Variable mutability

Python allows us to update the values stored in our variables during our program. This feature is called "mutability". It looks like this...

```python
my_var = "my string"
my_var
```

```python
my_var = 3
my_var
```

Here Python allowed us to reassign an integer value to the string variable `my_var`.

### 🚨 New Concept: Other assignment operators

In addition to the `=` assignment operator, Python provides several other assignment operators which serve as shorthand for lines of code like this `x = x + 2`. These assignment operators include `+=`, `-=`, `*=`, `/=`, and many others (that can be found here: https://www.w3schools.com/python/gloss_python_assignment_operators.asp). For example, line 2 in the following cell can be rewritten using the `+=` assignment operator...

```python
x = 1
x = x + 2
x
```

like this...

```python
x = 1
x += 2
x
```

### Question

The championship game is underway! It is your job to keep track of the scores.

1. Points scored in the first quarter: Blaze - 23, Titans - 20
2. Points scored in the second quarter: Blaze - 30, Titans - 28
3. Points scored in the third quarter: Blaze - 29, Titans - 30
4. Points scored in the fourth quarter: Blaze - 26, Titans - 34

Using the variables `blaze_score` and `titans_score` and the `+=` assignment operator, update the scores for each quarter of the game. Then output the final scores of each team.

**Solution**

```python
# 1. Points scored in the first quarter: Blaze - 23, Titans - 20
blaze_score += 23
titans_score += 20

# 2. Points scored in the second quarter: Blaze - 30, Titans - 28
blaze_score += 30
titans_score += 28

# 3. Points scored in the third quarter: Blaze - 29, Titans - 30
blaze_score += 29
titans_score += 30

# 4. Points scored in the fourth quarter: Blaze - 26, Titans - 34
blaze_score += 26
titans_score += 34

f"Titans score: {titans_score}, Blaze score: {blaze_score}"
```
