 A **module** in Python is a file containing Python code (functions, classes, or variables) that can be imported and used in other Python programs. However, modules can also be built-in (like math or sys), not just user-defined Python files.

```python
import myfunctions  # import the file myfunctions.py, which we recently created.
# If myfunctions is in another directory, use the following:
# from "directory path" import myfunctions
print(myfunctions.add(2, 3))
print(myfunctions.multiply(5, 5))
```

```python
# 2nd method using alias
import myfunctions as fn

print(fn.add(2, 3))
print(fn.multiply(5, 5))
```
