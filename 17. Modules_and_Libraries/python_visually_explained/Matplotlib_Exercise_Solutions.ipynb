### Question

## Exercise 2 - Bar charts

### 🚨 New Concept: Labeling bar chart bars

The `pyplot` module provides the `text()` function to affix labels to a bar in a bar chart.

**Syntax**
```python
text(x, y, s, fontdict=None, **kwargs)
```

**Parameters:**
1. **x** (`float`) - The x-coordinate of the text position in data coordinates.
2. **y** (`float`) - The y-coordinate of the text position in data coordinates.
3. **s** (`str`) - The string of text to be displayed at the specified coordinates.
4. **fontdict** (`dict`, optional) - A dictionary to override default text properties (e.g., font size, family, style).
5. **kwargs** (`**kwargs`, optional) - Additional keyword arguments to control the appearance of the text. This includes properties such as `color`, `fontsize`, `rotation`, `verticalalignment`, `horizontalalignment`, etc.

**Commonly used keyword arguments:**
- **fontsize**: Controls the size of the text. Can be a number or a string (e.g., `'small'`, `'large'`).
- **color**: Sets the color of the text. Can be a string (e.g., `'red'`, `'blue'`) or an RGB tuple.
- **rotation**: Rotates the text in degrees (e.g., `45`, `90`).
- **horizontalalignment** (or **ha**): Aligns text horizontally relative to the x-coordinate. Possible values: `'center'`, `'left'`, `'right'`.
- **verticalalignment** (or **va**): Aligns text vertically relative to the y-coordinate. Possible values: `'center'`, `'top'`, `'bottom'`.
- **bbox**: Adds a bounding box around the text. Accepts a dictionary of properties to define box style (e.g., `boxstyle`, `facecolor`, `alpha`).

In the below chart, we use a `for` loop to affix the number of sunny days as a label above each bar. We pass in the x location of the bar as the `x` argument, they y location of the bar plus 0.5 buffer as the `y` argument, specify the number of days (casted to a string) as the `s` argument, and specify horizontal alignment with the `ha` keyword argument.

````python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sunny_days = [10, 12, 15, 20, 25, 28, 30, 29, 25, 20, 15, 12]

# Increases the bar chart's width so that the x-labels do not overlap
plt.figure(figsize=(13, 6))

# Creates the bar chart using the given months and sunny_days data
plt.bar(months, sunny_days)

plt.title('Sunny Days per Month')
plt.xlabel('Month')
plt.ylabel('Sunny Days')

# Uses a loop to place a text label above each bar in the chart
for i in range(len(months)):
    plt.text(i, sunny_days[i] + 0.5, str(sunny_days[i]), ha='center')

plt.show()
````

### Question

The online store sells products in three different categories: Electronics, Clothing, and Home Goods. You are given the total sales for each category over the past week. Create a bar chart to compare the sales across these categories.

To create the chart:

1. Use a `matplotlib` function to create a bar chart with the categories on the x-axis and the sales figures on the y-axis.
2. Add a title to the chart, and label the x-axis as "Category" and the y-axis as "Sales ($)".
3. Customize the chart by...
   1. Changing the color of the bars to blue, green, and orange
   2. Adding numerical labels on top of each bar to show the sales figures
4. Display the chart.

**Solution**

````python
import matplotlib.pyplot as plt

categories = ['Electronics', 'Clothing', 'Home Goods']
sales = [1200, 850, 950]

plt.bar(categories, sales, color=['blue', 'green', 'orange'])

plt.title('Sales by Category Over the Past Week')
plt.xlabel('Category')
plt.ylabel('Sales ($)')

for i in range(len(categories)):
    plt.text(i, sales[i] + 20, str(sales[i]), ha='center')

plt.show()
````

To create a bar chart to show the sales data by category, we use `matplotlib`'s `bar()` function. We pass in as arguments the category data the x-variable and the sales data as the y-variable as well as the colors we wish the bars to be. Like Exercise 1, we add a title and labels to the chart. We use a `for` loop and the `text()` function to place a text label with the sales number above each bar. The `text()` function takes the x-position and y-position of the label as the first two arguments, the string to display as the third argument, and the horizontal alignment as the keyword argument `ha`. Finally, we use `show()` to display our chart.

## Exercise 3 - Pie charts

### 🚨 New Concept: Creating a pie chart with slice labels

The `plyplot` module provides the `pie()` function to create pie charts. The function takes two arguments that allow you to label the slices of the pie chart: `labels` and `autopct`.

The `labels` argument allows us to specify labels for each slice. In this example, we want to label each slice with the values in the `majors` list.

The `autopct` argument controls the display of the percentage values on the wedges. It allows you to specify a format string. For example, `'%1.1f%%'` displays the percentage to one decimal place.

````python
import matplotlib.pyplot as plt

majors = ['Engineering', 'Business', 'Arts', 'Sciences', 'Law', 'Others']
enrollment = [120, 80, 50, 100, 40, 30]

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(enrollment, labels=majors, autopct='%1.1f%%')
plt.title('Student Enrollment by Major')
plt.show()
````

### Question

The store has collected data on the age distribution of its customers. The data shows the percentage of customers in different age groups. Create a pie chart to visualize this distribution.

To create the chart:

1. Use a `matplotlib` function to create a pie chart that represents the age distribution.
2. Add a title to the chart.
3. Customize the pie chart by...
   1. Adding labels to each slice
   2. Displaying the percentage value on each pie wedge
   3. Using the given `wedge_colors` as the colors for the pie wedges
4. Add a legend and specify its title and location so that it doesn't overlap the pie chart.
5. Display the chart.

**Solution**

````python
import matplotlib.pyplot as plt

age_groups = ['18-25', '26-35', '36-45', '46-55', '56+']
percentages = [25, 40, 20, 10, 5]
wedge_colors = ['blue', 'green', 'orange', 'red', 'purple']

plt.pie(percentages, labels=age_groups, autopct='%1.1f%%', colors=wedge_colors)

plt.title('Customer Age Distribution')

plt.legend(age_groups, title="Age Groups", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
````

To show the consumer age distribution data as a pie chart, we use the `pie()` function. The `pie()` function takes the data as it's first argument. Then, it takes the labels, slice percentage label format, and slice colors as keyword arguments. We add a title to the graph and use the `legend()` function with the age groups as the first argument. We also specify the legend title, location, and `bbox_to_anchor` (an argument used in conjunction with the `loc` argument to control placement of the legend). Finally, we `show()` the chart.
