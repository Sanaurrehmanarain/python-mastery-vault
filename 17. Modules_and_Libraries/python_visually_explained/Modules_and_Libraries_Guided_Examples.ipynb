```python
import matplotlib.pyplot as plt 

days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
miles_hiked = [9.7, 8.4, 10.2, 10.8, 7.4]

average_miles = sum(miles_hiked) / len(miles_hiked)

plt.plot(days, miles_hiked, 'o-', label='Miles Hiked')
plt.axhline(y=average_miles, color='orange', linestyle='--', label='Average Miles Hiked')

plt.xlabel('Days')
plt.ylabel('Miles Hiked')
plt.title('Miles Hiked over 5 Days')
plt.legend()

plt.show()
```
