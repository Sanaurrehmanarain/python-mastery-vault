## Exercise 1 - The `math` module

### Part A

You must determine the minimum size in square kilometers of protected areas for a certain species of deer. Biologists have determined that the minimum area of a conservation zone for a herd of these deer is given by the following formula:

$ \text{minimum area (km}^2) = 10 + \sqrt{n} $, where $n$ is the number of individual deer in the heard.

The park contains five distinct herds, with 16 deer, 25 deer, 36 deer, 49 deer, and 45 deer, respectively.

Calculate the minimum conservation area for each of the herds.

**Solution**

To calculate the square root of the values in the `herd_sizes` list, we use the given equation, the `math` module's `sqrt()` function, and either a `for` loop:

````python
import math

herd_sizes = [16, 25, 36, 49, 45]

protected_area_sizes = []
for herd in herd_sizes:
    protected_area_sizes.append(10 + math.sqrt(herd))

print("Minimum protected area sizes (in square kilometers):", protected_area_sizes)
````

Or a list comprehension:

````python
import math

herd_sizes = [16, 25, 36, 49, 45]

protected_area_sizes = [10 + math.sqrt(herd) for herd in herd_sizes]
print("Minimum protected area sizes (in square kilometers):", protected_area_sizes)
````

Both solutions apply the `sqrt()` function to each item in the `herd_sizes` list and create a list called `protected_area_sizes` which contains the square kilometers that each herd requires.

### Part B

You have found a new species of frog in the forest. To predict how the frog population will grow, you track how many frogs there are in a pond on two dates 4 weeks apart. You would like to know the weekly growth rate of the population over this period.

According to your bio-statistician colleague, we can calculate the weekly growth rate of a population $r$ by applying the formula

$r = \left(e^{\left(\frac{\ln(P_t) - \ln(P_0)}{t}\right)} - 1\right) * 100$,

where $t$ is the length of the observation period in weeks, $P_0$ is the initial population, and $P_t$ is the population at the end of week $t$.

Use the above equation and functions provided within the `math` module to calculate the weekly growth rate over the observation period.

**Solution**

````python
P0 = 50
Pt = 1000
t = 4

r = (math.exp((math.log(Pt) - math.log(P0)) / t) - 1) * 100
print(f"Growth rate: {r :.2f}% per week")
````

To calculate the weekly growth rate, we need to use the `math` module's `exp()` and `log()` functions. The `exp()` function calculates the constant *e* to the power of its argument. In this case, we pass in the difference in natural logs of `Pt` and `P0` divided by the observation period `t`. By default, the `log()` function calculates the natural log of its argument.

### Part C

An unexpected early winter has reduced food availability for herbivores in the forest. You need to estimate the emergency food supplies required to sustain these animals until spring. You know the average daily food intake for different species and must ensure there is enough food for each one.

The emergency food supplies only come in whole units. So, a species with nutritional needs 1.7 units of food per day would require 2 units of emergency food.

Create a new list where each element is a species **total food unit requirements** over the winter season. To do this, use the `math` module to **round up** each element in the `daily_food_intakes` list and multiply that by the number of days in the winter season (120 days).

**Solution**

````python
daily_food_intakes = [3.8, 5.4, 6.4, 4.7, 7.8]
total_days = 120

food_supplies_ceil = [math.ceil(food_intake) * total_days for food_intake in daily_food_intakes]
print("Emergency food supplies (rounded up):", food_supplies_ceil)
````

We use a list comprehension to create a new list where each item is 120 days times the food intake, rounded up. In the list comprehension's expression, we use the `math` module's `ceil()` function to round up each value before multiplying it by 120. The loop variable `food_intake` iterates through the list `daily_food_intake`.

If we did not round the `daily_food_intake` values up with the `ceil()` function, we would get significantly smaller food values.

````python
food_supplies = [food_intake * total_days for food_intake in daily_food_intakes]

print("Emergency food supplies (without rounding):", food_supplies)
````

Alternatively, you could use a `for` loop to solve this problem:

````python
food_supplies_ceil = []
for food_intake in daily_food_intakes:
    food_supplies_ceil.append(math.ceil(food_intake) * total_days)

print("Emergency food supplies (rounded up):", food_supplies_ceil)
````

### 🚨 New Concept: The `random` module's `uniform()` function

The `random.uniform(a, b)` function generates a random floating-point number within a specified range. The parameters `a` and `b` define the lower and upper bounds of the range, respectively. The `uniform(a, b)` function returns a value $x$ where $a <= x <= b$.

It provides continuous random numbers, meaning it can generate any floating-point value within the given range.

````python
import random

# Generate a random floating-point number between 1.5 and 5.5
random_number = random.uniform(1.5, 5.5)
print(random_number)
````

In this code, `random.uniform(1.5, 5.5)` will produce a random float like `3.847`, `2.651`, or any other value between `1.5` and `5.5`. Each time the code is run, the value will vary within that range.

This function is useful when you need to simulate or model random decimal data, such as generating random coordinates, probabilities, or measurements.

### Part A

You and your team have tagged several rabbits from birth and have been tracking their ages, which are stored in the list below. The maximum lifespan of a rabbit in your forest is 12 years. Given a its age, a rabbit is equally likely to die at any point in time between now and its 12th "birthday".

For example, a rabbit that is 10 years old is equally likely to die at anytime in the next two years, and a 4-year-old rabbit is equally likely to die anytime in the next 8 years.

For that reason, use the `random` module's `uniform()` function to model the potential remaining lifespan of each rabbit. Then, calculate the total projected lifespan for each rabbit and print the results.

Note that, because we are using a function which generates random numbers, your output will differ each time you run your code.

**Solution**

````python
import random

rabbit_ages = [1.5, 3.2, 5.0, 7.8, 2.4]
max_lifespan = 12

# Use the uniform() function to model the remaining lifespans for each rabbit
remaining_lifespans = [random.uniform(0, max_lifespan - age) for age in rabbit_ages]

# Calculate total lifespans for each rabbit
total_lifespan = [age + remaining for age, remaining in zip(rabbit_ages, remaining_lifespans)]

# Print results
for i, age in enumerate(rabbit_ages):
    print(f"Rabbit {i+1}: Current Age = {age:.2f} years, Remaining Lifespan = {remaining_lifespans[i]:.2f} years, Projected Lifespan = {total_lifespan[i]:.2f} years")
````

To simulate the remaining lifespans for each of the rabbits, we use a list comprehension along with the `random` module's `uniform()` function. This function generates a float between the two arguments that we pass into it. For example, the first rabbit is 1.5 years old, so they have a maximum of 10.5 years remaining. With the arguments 0 and `max_lifespan - age` the `uniform()` function generates a float between 0 and 10.5 for Rabbit 1. The list comprehension applies the `uniform()` function to each of the rabbit ages, creating a new list that stores the remaining lifespans of each of the rabbits.

Then, to calculate the total lifespan of each rabbit, we use another list comprehension and the `zip()` function. The `zip()` function pairs each rabbit's age (from the `rabbit_ages` list) with its simulated remaining lifespan (from the `remaining_lifespan` list). We sum each rabbits' current age with its simulated remaining lifespan to calculate its total estimated lifespan.

Finally, we use a `for` loop and the `enumerate()` function to print the results.

### 🚨 New Concept: The `random` module's `choices()` function

The `random.choices()` function from Python's `random` module returns a randomly selected element (or multiple elements) from a given sequence. The `random.choices()` function allows you to specify how many items to select, and it can return the same item more than once (i.e., with replacement).

**Syntax**
```python
random.choices(population, weights=None, k=1)
```

- **`population`**: The sequence (list, tuple, etc.) from which to choose.
- **`weights`**: An optional parameter that assigns a relative weight to each element, influencing the probability of selection. If not provided, each element is equally likely.
- **`k`**: The number of items to choose (default is 1).

The function returns a list of `k` randomly selected elements. If `k > 1`, it can return duplicate elements because selection is done with replacement. For example, this code will randomly select 3 fruits from the list. The result might be `['apple', 'banana', 'apple']`, allowing for duplicates.

````python
import random

# Define a list of options
options = ['apple', 'banana', 'cherry', 'date']

# Select 3 random choices from the list, with equal probability
random_selection = random.choices(options, k=3)
print(random_selection)
````

You can also specify weights for the `choices()` function to use when picking from the given list. In this case, `banana` has a higher probability (weight of 0.7) of being chosen compared to the other fruits, which have a weight of 0.1:

````python
# Select 3 random choices, giving 'banana' a higher chance of being selected
weighted_selection = random.choices(options, weights=[0.1, 0.7, 0.1, 0.1], k=3)
print(weighted_selection)
````

The `random.choices()` function is useful when you need to randomly sample elements from a collection, especially when repetitions are allowed or when weighted probabilities are involved.

### Part B

Since you are concerned about the declining rabbit population, you want to simulate the impacts of different weather conditions (sunny, rainy, stormy, drought) on the number of rabbits in your forest for each of the next 12 months.

Based on historical data: 40% of the time the weather is sunny, 30% of the time the weather is rainy, 10% of the time the weather is stormy, and 20% of the time the forest is in a drought.

Additionally, when it is sunny the rabbit population increases by 10% per month, when it is rainy the population increases by 3% per month, when it is stormy the population decreases by 4% per month, and when the forest is in a drought the population decreases by 10% per month.

The current rabbit population is approximately 2,400.

Use the `random` module's `choices()` function to simulate weather events and their associated population growth rates over the next 12 months. The weathers event should be drawn according to their probabilities. The result of this will be a list of 12 population growth rates for the next 12 months. Then, apply the 12 population growth rates to the rabbit population to arrive at a final simulated population value for 12 months in the future. Round this value down since we cannot have a fractional rabbit!

**Solution**

````python
import math
import random

current_rabbit_pop = 2400
weather_probabilities = [0.4, 0.3, 0.1, 0.2]
pop_growth_rates = [1.1, 1.03, 0.96, 0.9]

# Simulate population growth rates over the next 12 months
simulated_pop_growth_rates = random.choices(pop_growth_rates, weather_probabilities, k=12)

# Apply the population growth rates to the rabbit population
simulated_pop = current_rabbit_pop
for growth_rate in simulated_pop_growth_rates:
    simulated_pop *= growth_rate

print(f"Rabbit population in 12 months: {math.floor(simulated_pop)} rabbits")
````

To simulate the population growth rates over the next 12 months, we use the `random.choices()` function. We specify the `pop_growth_rates` list as the values to be randomly chosen from by the `choices()` function, the `weather_probabilities` list as the weights, and 12 as the number of values to be chosen from the `pop_growth_rates` list. This call generates a list of 12 `simulated_pop_growth_rates`.

Next, we apply the `simulated_pop_growth_rates` to the `current_rabbit_pop` using a `for` loop. Using the `*=` operator allows us to cumulatively grow of the population over the 12 months.

Finally, we output the simulated population after 12 months.

### Part C

In Part B, we simulated how weather conditions affect the rabbit population over a 12 month period. Now, you'd like to expand your code to simulate **multiple** 12 month population forecasts.

Build on your code from Part B by embedding your simulation logic in `for` loop that runs 10 times. After each iteration, append the simulated outcome to the `simulated_pops` list. Finally, output the list of 10 populations forecasts.

**Solution**

````python
import math
import random

current_rabbit_pop = 2400
weather_probabilities = [0.4, 0.3, 0.1, 0.2]
pop_growth_rates = [1.1, 1.03, 0.96, 0.9]

simulations = 10
simulated_pops = []

for _ in range(simulations):
    # Simulate population change factors over the next 12 months
    simulated_pop_growth_rates = random.choices(pop_growth_rates, weather_probabilities, k=12)

    # Apply the population change factors to the rabbit population
    simulated_pop = current_rabbit_pop
    for growth_rate in simulated_pop_growth_rates:
        simulated_pop *= growth_rate
    simulated_pops.append(math.floor(simulated_pop))

print(f"Simulated populations after 12 months: {simulated_pops}")
````

To run our simulation 10 times, we embed our code from Part B in a `for` loop that runs 10 times. At the end of the loop, we append the simulated population value to the new `simulated_pops` list. This list resides outside the `for` loop so that it may be updated with the results of each iteration's simulation. Lastly, we print the results of each of the 10 simulations.

### Part D

Now that you can simulate many populations at once, you'd like to analyze these values. Specifically, you'd like to calculate the mean, standard deviation, and quartiles of the simulated population forecasts.

Using the simulated populations list you generated in Part C and the `statistics` module, compute the mean, standard deviation, and quartiles of the simulations.

**Solution**

````python
import statistics

mean = statistics.mean(simulated_pops)
print(f"Mean of the simulated rabbit populations: {mean}")

stdev = statistics.stdev(simulated_pops)
print(f"Standard deviation of the simulated rabbit populations: {stdev}")

quartiles = statistics.quantiles(simulated_pops)
print("Quartiles")
print(f"  25th percentile: {quartiles[0]}")
print(f"  50th percentile: {quartiles[1]}")
print(f"  75th percentile: {quartiles[2]}")
````

As in Exercise 2, the `statistics` module does the heavy lifting for us here. To calculate the mean and standard deviation, we simply call the `mean()` and `stdev()` functions respectively. To calculate the quartiles of the simulated populations, we pass our `simulated_pops` into the `quantiles()` function. By default, this function calculates quartiles since it has the default parameter `n=4`. The `quantiles()` function returns a list containing each of the quartiles.

### Part E

Sometimes it can be useful to reproduce the random values generated by a `random` module function. To do this, the `random` module allows you to set a seed using the `seed()` function.

Read about the `seed()` function here (https://docs.python.org/3/library/random.html#random.seed) and modify your code from Part C to use the seed value `87`. Because we are using a common seed value, your output should exactly match the expected output below.

**Solution**

````python
import math
import random

current_rabbit_pop = 2400
weather_probabilities = [0.4, 0.3, 0.1, 0.2]
pop_growth_rates = [1.1, 1.03, 0.96, 0.9]

simulations = 10
simulated_pops = []

random.seed(87)

for _ in range(simulations):
    # Simulate population change factors over the next 12 months
    simulated_pop_growth_rates = random.choices(pop_growth_rates, weather_probabilities, k=12)

    # Apply the population change factors to the rabbit population
    simulated_pop = current_rabbit_pop
    for growth_rate in simulated_pop_growth_rates:
        simulated_pop *= growth_rate
    simulated_pops.append(math.floor(simulated_pop))

print(f"Simulated populations after 12 months: {simulated_pops}")
````

Since we specified `87` as the seed, the results of the `random.choices()` function are reproducible.
