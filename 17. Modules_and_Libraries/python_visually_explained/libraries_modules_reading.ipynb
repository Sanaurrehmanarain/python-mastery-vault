LIBRARY
Imagine Python's built in function as the books on your personal bookshelf at home. These books, just like built in functions, are instantly accessible but limited in number. When the book you need can't be found at home, you may be able to find it by visiting your municipal library.

![alt text](<Screenshot 2025-01-26 112231.png>)

In our analogy, Python libraries act as the programming equivalent of municipal libraries. The expand the set of available functions when Python's built in functions fall short.

![alt text](<Screenshot 2025-01-26 112640.png>)

Python offers a vast ecosystem of libraries, each serving different programming needs. 

![alt text](<Screenshot 2025-01-26 112933.png>)

Among them, the Python Standard library stands out as one of the most important.

![alt text](<Screenshot 2025-01-26 113126.png>)

Just as a municipal library is organized into sections, each grouping books of the same genre, the Python Standard library is organized into modules, each grouping related functions. For instance, the Python Standard library has a Math mondule that contains functions performing mathematical operations, such as taking the square root or the logarithm of a number. The library also includes a random module which provides functions for generating random numbers, which is useful for simulations, and a statistics module, which provides functions for calculating common statistics like the mean and the median.   

![alt text](<Screenshot 2025-01-26 113720.png>)

Functions in the PSL are not as readily avaulable as Pythons's built in functions.

![alt text](<Screenshot 2025-01-26 114653.png>)

To use a function from a module of the standard library, we must first check out the function from the library by importing it into our program.

![alt text](<Screenshot 2025-01-26 114844.png>)

To import the square root function sqrt from the math module, we write **from math import sqr**.

![alt text](<Screenshot 2025-01-26 115116.png>)

In that statement, the from math part instructs Python that we want to retrieve a function from the math module. The import sqrt part specifies which function we want to import from that module.

Besides the sqr func the math module contains many other helpful functions.

Just as your municipal library doesn't contain all the books ever written, the PSL doesn't include func for every possible programming needs. Fortunately, there are other libraries called external libraries that can help in those situations.

![alt text](<Screenshot 2025-01-26 115912.png>)

External libraries have been created by the Python community to address specific programming needs, ranging from data analysis and machine learning to web development, data visualization, and more. To gain access to these external libraries, you must first get a library card by installing them in an interactive notebook environment like Google Colab or VS Code. Getting this library card can be done using Python's preferred installer program, also known as Pip.
