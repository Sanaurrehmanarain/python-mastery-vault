## Exercise 1 - Loading data into a DataFrame

### 🚨 New Concept: Creating a DataFrame from data

A DataFrame is a key data structure in Pandas. Its like a table in a spreadsheet or database, consisting of rows and columns, where:

- **Rows** represent individual records or data points (e.g., a person's information).
- **Columns** represent different variables or features (e.g., age, name, salary).

Each column can have a different data type (like numbers, text, or dates), and the DataFrame allows for easy data manipulation and analysis, such as filtering, grouping, and calculating statistics.

Think of it as a flexible table with labeled rows and columns that lets you perform powerful data operations quickly.

Data presented as a collection (like a list or dictionary) of a collection can be converted into a DataFrame. For example, the following list of lists can be converted into a DataFrame:

```python
import pandas as pd

data = [
    [1, 'Alice', 24, 'New York', 55000, 'HR'],
    [2, 'Bob', 27, 'San Francisco', 62000, 'Engineering'],
    [3, 'Charlie', 22, 'Los Angeles', 48000, 'Marketing'],
    [4, 'David', 32, 'Chicago', 73000, 'Finance'],
    [5, 'Eva', 29, 'Boston', 68000, 'Sales']
]

pd.DataFrame(data)
```

## Exercise 2 - Filtering a DataFrame

### 🚨 New Concept: Filtering data in a DataFrame

Filtering data in a DataFrame is essential to focus on specific subsets of data, making analysis more efficient and targeted. You might filter data to:

1. **Exclude irrelevant rows**: Remove rows that don't meet certain criteria (e.g., values below a threshold).
2. **Analyze specific conditions**: Narrow the data to analyze certain conditions (e.g., sales from a particular region).
3. **Improve performance**: Work with smaller, more manageable datasets by removing unneeded data.

**How to filter data in a DataFrame by condition**:

`df[df['column'] > value]` filters rows where a column's values are greater than a specified value.

The condition, `df['column'] > value`, checks if each value in the specified column is greater than `value` and returns a `pandas` Series. A Series is basically just a DataFrame with one column.

If the the column value is greater than `value`, the corresponding entry in the Series is `True`. If not, the entry is `False`.

For example, we can filter the below data frame to only display students born after 01-01-2001:

```python
import pandas as pd

data = {
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'Diana Prince', 'Evan Davis'],
    'Student_ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Major': ['Computer Science', 'Mathematics', 'Mathematics', 'Biology', 'Mathematics'],
    'Year_in_School': ['Sophomore', 'Junior', 'Senior', 'Freshman', 'Senior'],
    'Birthday': ['2002-05-14', '2001-08-22', '2000-12-01', '2003-03-17', '1999-07-09']
}

df = pd.DataFrame(data)

df[df["Birthday"] > "2001-01-01"]
```

The filter `df["Birthday"] > "2001-01-01"` tests each value in the `"Birthday"` column and creates a Series:

```python
df["Birthday"] > "2001-01-01"
```

We then use that Series to filter the DataFrame:

```python
df[df["Birthday"] > "2001-01-01"]
```

## Exercise 3 - Calculations with DataFrames

### 🚨 New Concept: Creating new DataFrame columns

After a DataFrame is created, you can add a new column to it like this:

```python
import pandas as pd

data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Region_A_Sales': [2300, 3400, 4100, 3200, 2800, 3600],
    'Region_B_Sales': [1800, 2200, 2700, 2100, 1900, 2500],
}

df = pd.DataFrame(data)

# Add new columns
df['Region_A_Tax'] = df['Region_A_Sales'] * 0.2
df['Region_B_Tax'] = df['Region_B_Sales'] * 0.2

df
```

### Part A

Add a new column to the DataFrame that calculates the total delay for each flight.

**Solution**

```python
flight_info['Total_Delay'] = flight_info['Departure_Delay'] + flight_info['Arrival_Delay']
flight_info
```

We add a new column to a DataFrame similar to how we add a new dictionary entry; we simply specify the name of the column in between square brackets. Then, we calculate the total delay (arrival + departure) by adding the DataFrame columns together.

## Exercise 4 - Sorting DataFrames

### Part A

Identifying the flights with the longest total delays is crucial for prioritizing efforts to reduce delays. Sorting the data by total delay will help you pinpoint the worst offenders.

Sort the DataFrame by `Total_Delay` in descending order to identify flights with the most significant delays.

**Solution**

```python
sorted_delays = flight_info.sort_values(by='Total_Delay', ascending=False)
sorted_delays
```

To sort the DataFrame by `Total_Delay` in descending order, we use the `.sort_values()` method. We pass in the column we wish to sort, and specify `ascending=False` to sort in descending order.

### Part B

Departure delays can have a cascading effect on the entire schedule. By sorting flights by departure delay in ascending order, you can see which flights are typically on time or early, helping you identify patterns or successful strategies.

Sort the DataFrame by `Departure_Delay` in ascending order to identify flights with minimal or no departure delays.

**Solution**

```python
sorted_delays = flight_info.sort_values(by='Departure_Delay', ascending=True)
sorted_delays
```

Similar to Part A, we use the `.sort_values()` method. This time, to sort in ascending order, we specify the keyword argument `ascending=True`.
