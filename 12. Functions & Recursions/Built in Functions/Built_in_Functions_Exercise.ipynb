## Exercise 1 - Outputting data

### 🚨 New Concept: Escape sequences

When writing out a longer string, it is often desirable to split the string up into multiple lines or insert a tab for formatting. Python provides us with special character sequences to accomplish tasks like this. These sequences always begin with the escape character backslash `\`. To tell Python that we would like a new line to be inserted in our string when outputted, we can use the new line character sequence `\n` like this...

````python
print("Hello\nworld")
````

Similarly, we can insert a tab by using the sequence `\t` like this...

````python
print("Hello\tworld")
````

Note that if you try to type the `\` character as part of a string, you will get an error (or at least a syntax warning).

````python
# print("The backslash character: \")
````

If you wish to type the literal `\` character, you need to escape it by typing `\\`. For example...

````python
print("The backslash character: \\")
````

Additionally, if you want to include `'` or `"` characters within a string and do not want to worry about which way you need to declare the string, you can escape single and double quotes with the backslash like this...

````python
print("She said \"My name is Mary O\'Connor\".")
````

A full list of escape sequences can be found here: https://www.w3schools.com/python/gloss_python_escape_characters.asp

## Exercise 2 - Finding the absolute difference

Next, Alex is interested in knowing the absolute difference in his day-to-day step counts. Create two variables, `todays_steps = 11240` and `yesterdays_steps = 14290`, and compute the absolute difference.

**Solution**

To calculate the absolute difference, we use the `abs()` function and pass in the difference between the two step variables as an argument.

````python
todays_steps = 11240
yesterdays_steps = 14290

print(f"The absolute difference is {abs(todays_steps - yesterdays_steps)} steps.")
````

Swapping the order of `todays_steps` and `yesterdays_steps` does not change the final result because we are calculating the *absolute* difference.

````python
todays_steps = 11240
yesterdays_steps = 14290

print(f"The absolute difference is {abs(yesterdays_steps - todays_steps)} steps.")
````

Note that we can use an f-string just like a normal string inside a print statement. F-strings are very useful within print statements for creating output code. If we chose not to use an f-string, our output could look like this...

````python
print("The absolute difference is", abs(todays_steps - yesterdays_steps), "steps.")
````

## Exercise 3 - Identifying the greatest value

Beyond being interested in the absolute difference from yesterday to today, Alex is also interested in identifying the most steps he took in the last seven days.

Given that in the last seven days Alex took 9820, 15400, 12980, 10090, 13820, 11240, and 14290 steps, use a built-in function to identify the most steps Alex took.

**Solution**

````python
print(f"The most steps that Alex took was {max(9820, 15400, 12980, 10090, 13820, 11240, 14290)}.")
````

## Exercise 4 - Identifying the least value

In the spirit of analyzing extremes, Alex would like you to identify the fastest time in his last five 100 meter dash attempts: 14.89, 14.02, 15.67, 13.91, 13.87. Use a built-in function to do this.

**Solution**

````python
print(f"Alex's fastest 100 meter time is {min(14.89, 14.02, 15.67, 13.91, 13.87)} seconds.")
````

## Exercise 5 - Rounding results

Alex's goal is to exercise for an average of 45 minutes for per day. Given Alex's workout plan from Exercise 1, find the average number of minutes Alex's will exercise per day. Then use a built-in function to display that result to 1 decimal place.

Alex's workout plan
- Monday: run for 45 minutes
- Tuesday: yoga for 60 minutes
- Wednesday: bike for 90 minutes
- Thursday: strength train for 60 minutes
- Friday: swim for 30 minutes
- Saturday: basketball for 60 minutes
- Sunday: rest

**Solution**

````python
avg_minutes = (45 + 60 + 90 + 60 + 30 + 60 + 0) / 7

print(f"Alex will exercise for an average of {round(avg_minutes, 1)} minutes per day. This does meet his goal!")
````
