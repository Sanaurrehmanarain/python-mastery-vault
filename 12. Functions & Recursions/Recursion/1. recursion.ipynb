# What is **Recursion**?

Recursion is when a function calls itself in order to solve a smaller version of the same problem.

Every recursive function has two main parts:

1. Base case → When to stop (prevents infinite loop).

2. Recursive case → Function calls itself with a smaller/simpler input.

👉 Think of it like breaking a big problem into smaller problems until it’s so small that the answer is obvious.

### Sum of numbers (1 to n)

````python
def sum_numbers(n):
    if n == 0:             # Base case
        return 0
    else:                  # Recursive case
        return n + sum_numbers(n-1)
````

````python
sum_numbers(5)
````

Step-by-step for `sum_numbers(3)`:

- `sum_numbers(3) = 3 + sum_numbers(2)`

- `sum_numbers(2) = 2 + sum_numbers(1)`

- `sum_numbers(1) = 1 + sum_numbers(0)`

- `sum_numbers(0) = 0` (base case)

Now add back: `1 + 0 = 1 → 2 + 1 = 3 → 3 + 3 = 6`

✅ Result = 6

If n is 0 or negative, countdown outputs the word, “Blastoff!”. Otherwise, it outputs n
and then calls itself, passing n-1 as an argument.

````python
def countdown(n):
    if n <= 0:              # Base case
        print("Blastoff!")
    else:
        print(n)
        countdown(n - 1)    # Recursive case

countdown(3)
````

### Reverse a String

````python
def reverse_string(s):
    if len(s) == 0:             # Base case
        return ""
    else:
        return s[-1] + reverse_string(s[:-1])
````

````python
reverse_string('cat')
````

Step-by-step for `"cat"`:

- `reverse_string("cat") = "t" + reverse_string("ca")`

- `reverse_string("ca") = "a" + reverse_string("c")`

- `reverse_string("c") = "c" + reverse_string("")`

- `reverse_string("") = ""`

Put together → `"t" + "a" + "c" = "tac"`

A function that calls itself is recursive. As another example, we can write a function
that prints a string n times:

````python
def print_n_times(string, n):
    if n > 0:
        print(string)
        print_n_times(string, n - 1)
````

If n is positive, print_n_times displays the value of string and then calls itself, pass‐
ing along string and n-1 as arguments.
If n is 0 or negative, the condition is false and print_n_times does nothing.

````python
print_n_times("hello there", 3)
````

For simple examples like this, it is probably easier to use a for loop. But we will see
examples later that are hard to write with a for loop and easy to write with recursion,
so it is good to start early.

````python
for i in range (3):
    print('hello there')
````

## 🧮 Recursive Function: Factorial
### ✅ What Is a Factorial?
The factorial of a number `n` (written as `n!`) is the product of all positive integers less than or equal to `n`.

For example:
- 5! = 5×4×3×2×1=120
- 0! = 1 (by definition)

### 🧠 Recursive Definition
A factorial can be defined recursively as:

- **Base case**: 

factorial(0)=1

0! = 1

- **Recursive case**: 

factorial(𝑛)=𝑛×factorial(𝑛−1)

n! = n(n − 1)!

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Example usage
print(factorial(5)) 
````

### 🔍 Explanation
- `factorial(5)` calls `factorial(4)`

- `factorial(4)` calls `factorial(3)`

- ...

- Until it reaches `factorial(0)`, which returns `1`
 
- Then all the calls resolve in reverse order:

    - `factorial(1)` returns `1 × 1 = 1`

    - `factorial(2)` returns `2 × 1 = 2`

    - ...

    - `factorial(5)` returns `5 × 24 = 120`

````python
print(factorial(3))  
print(factorial(0)) 
````

After factorial, the most common example of a recursive function is fibonacci,
which has the following definition:
- fibonacci (0) = 0
- fibonacci (1) = 1
- fibonacci (n) = fibonacci (n − 1) + fibonacci (n − 2)

````python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
````

````python
fibonacci(3)
````

## Checking Types
In programming, type checking means verifying that the input (argument) to a function is of the correct type (e.g., integer, string, float, list).

In Python, you can do this using `isinstance(variable, type)`

```python
x = 5
print(isinstance(x, int))   # True
print(isinstance(x, str))   # False
```
### Why do we use type checking?

1. Error prevention → Stops the program from running into unexpected errors.
Example: factorial of "hello" makes no sense.

2. Code safety → Makes sure the function only works with valid inputs.

3. Better debugging → Clear error messages instead of strange runtime crashes.

### How is it beneficial?

Without type checking, your recursive factorial function could break:

```python
factorial("hello")  
# This would crash because "hello" - 1 is invalid.
```
With type checking:

```python
factorial("hello")  
# factorial is only defined for integers.
```
So the program doesn’t crash—it fails safely.

### Example: Your factorial function explained

````python
def factorial(n):
    if not isinstance(n, int):      # Check if n is not an integer
        print('factorial is only defined for integers.')
        return None
    
    elif n < 0:                     # Check if n is negative
        print('factorial is not defined for negative numbers.')
        return None
    elif n == 0:                    # Base case: factorial(0) = 1
        return 1
    else:                           # Recursive case
        return n * factorial(n-1)
````

How it works step by step:

Suppose `factorial(3)` is called:

1. `factorial(3)`

    - not negative, not 0 → returns `3 * factorial(2)`

2. `factorial(2)`

    - returns `2 * factorial(1)`

3. `factorial(1)`

    - returns `1 * factorial(0)`

4. `factorial(0)`

    - base case → returns `1`

Now we multiply backwards:
`1 * 1 = 1`
`2 * 1 = 2`
`3 * 2 = 6`

✅ `factorial(3) = 6`
