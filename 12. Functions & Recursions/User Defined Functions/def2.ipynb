- If nothing enters as a arguments for name and age, the default values will be "unknown name" and 0. 
- single * before marks mean we can put as many arguments as we want.
- dounble ** means we will use keys with the associate values for marks,
- for example * => 95, 80, 90, 88 as many arguments as we want for marks,
- ** => english = 95, maths = 80, biology = 90, science = 88 (keys with its values)'''

```python
def  student (name = "Unknown name", age = 0, **marks): 
    print("name: ", name)
    print("age", age)
    # print("marks", marks) # it will give the output in a tuple if it is single *, if it is **marks it will give you in dictionary.
    for key, value in marks.items(): # thats why we use for loop to change to proper output
        print(key, ":", value)

# student() it will give you the default values.
student()
```

```python
student("Sana", 37, english = 95, maths = 80, biology = 90, science = 88)
```

#### You can write a same function in many ways:

```python
# method 1
x = int(input("Enter a number: "))

if x % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# method 2
def main():
    x = int(input("Enter a number: "))
    if is_even(x):
        print("Even")
    else:
        print("Odd")
    

def is_even(x):
    if x % 2 == 0:
        return True
    else:
        return False

main()
```

```python
# method 3
def main():
    x = int(input("Enter a number: "))
    if is_even(x):
        print("Even")
    else:
        print("Odd")
    

def is_even(x):
    return True if x % 2 == 0 else False

main()
```

```python
# method 4
def main():
    x = int(input("Enter a number: "))
    if is_even(x):
        print("Even")
    else:
        print("Odd")

def is_even(n):
    return n % 2 == 0
main()
```
