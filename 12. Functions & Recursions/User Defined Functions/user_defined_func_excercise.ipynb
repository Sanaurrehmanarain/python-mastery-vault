##### Exercise 1 - Loan payment calculator
You expect that clients will often want to calculate the monthly payment of a loan. To be prepared, you decide to create a Python function that calculates monthly loan payments quickly and accurately. This tool will help you give clear and reliable answers to your clients.

Write the function calculate_loan_payment():

- Parameters: the loan amount, the annual interest rate (ex: the value 12 = 12%), the number of years to repay the loan
- Function Body
    1. Calculate the monthly interest rate
    2. Determine the number of payments per year
    3. Use the monthly interest rate, number of payments per year, and loan amount to calculate the monthly loan payment
- Return value: the monthly loan payment

Use the following equations for calculating the monthly payment:

Calculate the monthly interest rate: monthly interest rate=annual interest rate12⋅100

- Divides the annual interest rate by 12 (for the number of months)
- Divides the monthly interest rate by 100 to convert it into a decimal number

Calculate the total number of payments: number of payments=years×12

- Converts the number of years into the corresponding number of months, since loan payments occur monthly

Calculate the monthly payment: monthly payment=loan amount×monthly interest rate1−(1+monthly interest rate)−number of payments

- Uses the the loan amount and previously determined monthly interest rate and number of payments as inputs
- Determines the total monthly payment, spreading the loan's full cost (principal plus interest) evenly over the loan term in the numerator
- Adjusts the numerator by ensuring each payment reflects the correct balance between interest and principal, with the interest portion decreasing as the loan balance is paid down

```python
# Your function
def calculate_loan_payment(loan_amount, annual_interest_rate, years):
    monthly_interest_rate = annual_interest_rate / 12 / 100
    number_of_payments = years * 12
    monthly_payment = (loan_amount * monthly_interest_rate) / (1 -(1 + monthly_interest_rate) ** -number_of_payments)
    return monthly_payment

# Sample function call
monthly_payment = calculate_loan_payment(100000, 5, 30)
print(f"Monthly Payment: ${monthly_payment:.2f}")
```

In this function, it's important to calculate the monthly interest rate and the number of payments before calculating the monthly payment, since these values are needed to accurately calculate the monthly payment.

##### Exercise 2 - Retirement savings calculator
##### 🚨 New Concept: Keyword arguments
Up to this point, we have been passing arguments to functions in the exact order that the function parameters specify. This is known as using "positional arguments."

However, when passing arguments to a function that expects many parameters, it can be easy to forget the order in which the arguments should be passed. As a result, you might accidentally pass arguments to a function in the wrong sequence, leading to errors and bugs.

For example, the function below uses several parameters to create a user profile:

```python
def configure_profile(username, display_name, email, bio, notifications_enabled):
    profile = {
        "username": username,
        "display_name": display_name if display_name else username,
        "email": email,
        "bio": bio,
        "notifications_enabled": notifications_enabled
    }

    for key, value in profile.items():
        print(f"{key}: {value}")
```

When passing in arguments to the function, we need to be careful to pass the arguments in the proper order:

```python
user = configure_profile("alexsmith", "Alex Smith", "alex@example.com", "Avid traveler and photographer.", False)
```

It is hard to tell from looking only at the function call what each argument represents inside the function.

To address this, Python allows us to use keyword arguments when calling functions. By using keyword arguments, we can clearly assign values to specific function parameters, making the function call more readable. For example, the following function call will produce the same output as the previous function call:

```python
user = configure_profile(
    username="alexsmith",
    display_name="Alex Smith",
    email="alex@example.com",
    bio="Avid traveler and photographer.",
    notifications_enabled=False
)
```

In this call we can clearly see what each argument represents within the function.

When using keyword arguments, as opposed to positional arguments, the position of the arguments in the function call doesn't matter. Thus, we can change their order and obtain the same result:

```python
user = configure_profile(
    bio="Avid traveler and photographer.",
    email="alex@example.com",
    username="alexsmith",
    display_name="Alex Smith",
    notifications_enabled=False
)
```

In summary, keyword arguments offer several benefits:
1. Improved Readability: They make function calls clearer and easier to understand, especially when dealing with multiple parameters.
2. Flexibility in Argument Order: They allow you to rearrange the order of arguments.
3. Error Prevention: They help programmers avoid errors that can occur from passing arguments in the wrong order.
##### Question
One of your key objectives is to help clients achieve a secure retirement. A smart way to prepare is by committing to regular annual contributions. Clients often wonder how much they will have accumulated in their retirement account by the time they retire if they follow this approach.

For instance, a client starting with $100,000 in savings and contributing $7,000 annually might ask how much they could expect to save over 30 years, assuming an annual return of 5%.

The function you'll create in this exercise will allow you to demonstrate the long-term impact of these contributions on their retirement savings.

Write the function calculate_retirement_savings():

- Parameters: the initial savings amount, annual contribution, annual return rate, number of years until retirement
- Function Body
    1. Initialize a total savings variable
    2. Use a for loop to grow the initial savings amount by the annual return rate for the specified number of years, inside the loop:
        1. First, add the annual contribution to the total savings
        2. Then, grow the total savings using the growth equation
- Return value: the total savings at the time of retirement
Use the following equation to calculate the retirement savings:

total savingsyear 1=total savingsyear 0∗1+annual return rate100

```python
# Your function
def calculate_retirement_savings(initial_savings, annual_contribution, annual_return_rate, years):
    total_savings = initial_savings
    for _ in range(years):
        total_savings += annual_contribution
        total_savings *= (1 + annual_return_rate / 100)
    return total_savings

retirement_savings =  calculate_retirement_savings(
    initial_savings = 10000,
    years = 30,
    annual_contribution = 5000,
    annual_return_rate = 5
)   # Call the function with keyword arguments

print(f"Retirement Savings: ${retirement_savings:.2f}")
```

The calculate_retirement_savings() function takes four arguments: initial_savings, annual_contribution, annual_return_rate, and years. Within the function, the for loop runs once for each year. During each loop iteration, we increase the total savings by the annual contribution amount. Then we account for the additional year of total_savings growth using the given equation.

By calling the function with keyword arguments, the code is more readable and less error prone.

##### Question
Many of your clients will be curious about how their investments might grow over time. To provide a clear picture of potential growth, you decide to create a function that projects the future value of a client's investments. This tool will help you demonstrate the power of compound interest, boosting your clients' confidence in their long-term financial strategies.

Write the function project_investment_growth():

- **Parameters**: the initial investment amount, the number of years, an optional annual growth rate (default to 7%)
- **Function Body**
    1. Calculate the future value of the initial investment amount
- **Return value**: the future value of the investment

Use the following equation to project the growth of an investment:

future value=initial amount×(1+annual rate100)years

```python
# Your function
def project_investment_growth(initial_amount, years, annual_rate = 7):
    future_value = initial_amount * ((1 + annual_rate / 100) ** years)
    return future_value

# Sample function calls
future_value = project_investment_growth(5000, 20)
print(f"Future Value: ${future_value:.2f}")

future_value = project_investment_growth(5000, 20, 3)
print(f"Future Value (rate specified): ${round(future_value, 2)}")
```

The project_investment_growth() function has two required parameters (initial_amount and years) and one optional parameter (annual_rate), which defaults to 7%. Inside the function, we use the investment growth formula to calculate and return the future value of the investment.

Calling the function without a annual rate, as in the first sample function call, means that the function will use the default 7% rate. However, if we specify an annual rate, then the function will use that rate instead.

##### Exercise 4 - Budget analysis
##### 🚨 New Concept: Multiple return values
Python allows us to return multiple values from a single function. For example, in the function below, we calculate several statistics from a list of grades and return them:

```python
# The scores parameter is a list of numerical grades
def grade_statistics(scores):
    total = sum(scores)
    count = len(scores)
    average = total / count
    highest = max(scores)
    lowest = min(scores)

    return average, highest, lowest
```

These values are returned as a tuple, which can be unpacked and used later in the program.

```python
scores = [85, 90, 78, 92, 88]
average, highest, lowest = grade_statistics(scores)

print(f"Average grade: {average}")
print(f"Highest grade: {highest}")
print(f"Lowest grade: {lowest}")
```

Advantages of returning multiple values:

1. **Improves Readability and Maintenance**: Functions that return multiple values make the code easier to read and maintain by clearly indicating that related information is being returned together, which can then be quickly unpacked and used.

2. **Handles Complex Operations**: Some tasks require multiple outputs, such as obtaining the average, highest, and lowest scores from a list of grades. Returning multiple values allows you to manage these tasks efficiently with a single function call.

##### Question
When working with a new client, one of the first steps is to understand their budget—specifically, how much they earn compared to how much they spend. To help with this, you decide to create a function that analyzes a client’s budget by calculating their total monthly expenses and determining whether they have a surplus or a deficit based on their monthly income. This tool will enable you to quickly assess their financial situation and provide tailored advice.

Use the following equation to calculate the savings or deficit:

savings or deficit=monthly income−monthly expenses

- Calculates the monthly savings or deficit amount using monthly income and total monthly expenses

Write the function analyze_budget():

- Parameters: the monthly income as an integer, a list of monthly expenses
- Function Body
    1. Calculate the total monthly expenses
    2. Calculate the amount of surplus or deficit
- Return values: the amount of surplus or deficit and the total expenses

```python
# Your function
def analyze_budget(monthly_income, expenses):
    total_expenses = sum(expenses)
    savings_or_deficit = monthly_income - total_expenses
    return savings_or_deficit, total_expenses

# Sample function call
monthly_income = 4000
monthly_expenses = [1200, 300, 150, 600, 200]

savings_or_deficit, total_expenses = analyze_budget(monthly_income, monthly_expenses)
print(f"Surplus: ${savings_or_deficit:.2f}" if savings_or_deficit > 0 else f"Deficit: ${abs(savings_or_deficit):.2f}")
print(f"Monthly income: ${monthly_income} | Monthly expenses: ${total_expenses}")
```

The analyze_budget() function has two parameters: monthly_income and expenses. In the function, we first sum the expenses to get the total expenses. Then, we subtract the total expenses from the monthly income to determine the amount of surplus or deficit. The function returns both the surplus or deficit amount and the total monthly expenses.

##### Exercise 5 - Track expenses by category
##### 🚨 New Concept: Arbitrary arguments
In Python, arbitrary arguments enable a function to accept any number of inputs, providing flexibility in handling different input sizes. There are two types of arbitrary arguments:

1. Arbitrary positional arguments (using *args)
2. Arbitrary keyword arguments (using **kwargs)

Arbitrary Positional Arguments (*args)

When you prefix a parameter in a function header with an asterisk (*), it allows the function to accept any number of positional arguments. These arguments are gathered into a tuple:

```python
def my_function(*args):
    print(type(args))

my_function("Hello", "World", "!")
```

You can then iterate over or manipulate the tuple as needed. For example, this function accepts any number of names and prints each one using a for loop:

```python
def greet(*names):
    print("Greeting...")
    for name in names:
        print(f"Hello, {name}!")
    print("Done greeting.\n")

greet("Alice", "Bob", "Charlie")
greet("Dana", "Edgar")
greet("Fred", "Greta", "Haley", "Iris")
```

```python
def add(*args):
    sum = 0
    for i in range(len(args)):
        sum += args[i]
    return sum

add(2, 4, 5, 6, 7)
```

```python
# You can simplify the loop using Python's built-in sum() function
def add(*args):
    return sum(args)

add(2, 4, 5, 6, 7)
```

Arbitrary Keyword Arguments (**kwargs)

When you prefix a parameter with two asterisks (**), it allows the function to accept any number of keyword arguments. These arguments are gathered into a dictionary:

```python
def my_function(**kwargs):
    print(type(kwargs))

my_function(key0="Hello", key1="World", key2="!")
```

You can then access the items by key. For example, this function accepts any number of pet details and prints them out using a for loop:

```python
def describe_pet(**pet_details):
    for key, value in pet_details.items():
        print(f"{key}: {value}")

describe_pet(name="Buddy", species="Dog", age=5)
```

The arbitrary keyword arguments name="Buddy", species="Dog", age=5 are passed into the describe_pet function and collected into a dictionary called pet_details. In this dictionary, name, species, and age are the dictionary keys and "Buddy", "Dog", and 5 are the dictionary values. Inside the function, the for loop iterates through each item in the pet_details dictionary, printing both the key and its corresponding value.

```python
def key_values(**kwargs):
    for i in kwargs:
        print(f"Key is: {i}, and Value is: {kwargs[i]}")

key_values(a = 3, b = "dino", c = True, d = 4.5)
```

##### Question
Clients often need to see how their spending is distributed across different categories to make informed decisions about budgeting. To assist with this, you decide to create a function that analyzes a client's expenses by calculating the total expenses and breaking down each category's impact on their overall budget. This tool will be invaluable in helping clients visualize their spending habits and make adjustments as needed.

Write the function analyze_expenses():

- Parameters: any number of expenses (passed as arbitrary keyword arguments)
- Function Body
    1. Calculate the sum of the expenses passed in
    2. Use a for loop to iterate through each of the expenses and print:
        1. the category name
        2. the amount of the expense
        3. the percentage of the total budget that expense takes up
- Return value: the sum of all expenses

```python
def analyze_expenses(**expenses):
    total_expenses = sum(expenses.values())
    print("Expense categories:")
    for key, value in expenses.items():
        print(f"  {key}: ${value} | {round((value / total_expenses), 2) * 100}%")
    print(f"Total expenses: ${total_expenses}")

    return total_expenses

# Sample function call
total_expenses = analyze_expenses(rent=1200, groceries=400, utilities=150, entertainment=200)
```

The analyze_expenses() function takes arbitrary keyword arguments for the parameter expenses, which is a dictionary. Inside the function, we sum the values of expenses to calculate the total expenses. Then, we use a for loop to iterate through the expense categories, printing each amount along with its percentage of the total budget. Finally, the function returns the total expenses.

Let say you are a teacher and you have different student records containing id of a student and the markslist in each subject where different students have taken different number of subjects. All these records are in hard copy. You want to enter all the data in computer and want to compute the average marks of each student and display.

```python
def get_data_from_user():
    student_data = {}
    while True:
        studentID = input("Enter student ID: ")
        if studentID in student_data:
            print(studentID, "is already inserted")
            continue

        markslist = input("Enter the marks (comma-separated): ")
        try:
            marks = [float(mark.strip()) for mark in markslist.split(",")]
        except ValueError:
            print("Please enter valid numeric marks.")
            continue

        student_data[studentID] = marks

        more_student = input('Enter "no" to quit insertion or press Enter to continue: ')
        if more_student.lower() == "no":
            break
    return student_data
```
