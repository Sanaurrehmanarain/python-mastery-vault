Method 1: use of def function without parameter

```python
def hello():
    print("hello,")

name = input("What is your name? ")
hello()
print(name)
```

Method 2: def function with parameter

```python
def hello(to):
    print("hello,", to)

name = input("What is your name? ")
hello(name)
```

Method 3: def function with default parameter

```python
def hello(to = "world"):
    print("hello,", to)

hello()
name = input("What is your name? ") # if you give any argument it will take, and will not take the default one.
hello(name)
```

Its important to assign return values to variables if you'd like to use them later <br> in your program.

```python
def greet(name):
    return "Hello " + name

greeting = greet("Alice")
greet("Bob")
print(greeting)
```

The return value of one function call can be used as an argument to another <br> function call.

```python
def add(a, b):
    return a + b

result = add(5, 10)
print(result)
print(add(result, 20))
```

Python allows us to use multiple return statements inside a function.

```python
def compute(a, b):
    result = a * b
    if result > 10:
        return result - 3
    else:
        return result + 4

print(compute(3, 2))
print(compute(5, 3))
```

This function uses a global variable. Can you predict the output?

```python
counter = 10

def increment(value):
    counter += value
    return counter

print(increment(5))
print(increment(3))
print(counter)
```

**Parameter**
Variable name in a function's header

**Argument**
Value passed into a function

While both arguments and parameters refer to function inputs, arguments are the actual values used in function calls, whereas parameters are the variable names used in function definition.

```python
def rectangle_area(length, width): # The variable names in the head lenght & width are called the function's parameters.
    area = length * width
    return area # Essential for sending data outside the function

rectangle_area(100, 140) # values passed into a a function are called arguments.
```

```python
print(f"Zone 2 area: {rectangle_area(260, 70)}")
print(f"zone 3 area: {rectangle_area(140, 220)}")
```

```python
distance = 10 # In miles
temp = 75 # In Farenheit
terrain_difficulty = 2 # On a scale from 1 (easiest) to 3 (hard)


# calc_hours
baseline_hours = distance / 3
terrain_factor = 1 + (terrain_difficulty / 2)
temp_factor = temp / 70
adjusted_hours = baseline_hours * temp_factor * terrain_factor

# calc_calories
baseline_calories_per_hour = 160
terrain_factor = 1 + (terrain_difficulty / 3)
total_calories = baseline_calories_per_hour * terrain_factor * adjusted_hours

# print_result
print(f"Estimated time hiking: {round(adjusted_hours, 2)} hours.")
print(f"Estimated calories burned: {round(total_calories, 2)} calories")
```

Code could be more clear. We can make it more easier to understand and readabale by encapsulating these three blocks into three different functions.

```python
def calc_hours(distance, temp, terrain_difficulty):
    baseline_hours = distance / 3
    terrain_factor = 1 + (terrain_difficulty / 2)
    temp_factor = temp / 70
    return baseline_hours * temp_factor * terrain_factor

def calc_calories(adjusted_hours, terrain_difficulty):
    baseline_calories_per_hour = 160
    terrain_factor = 1 + (terrain_difficulty / 3)
    return baseline_calories_per_hour * terrain_factor * adjusted_hours

def print_results(adjusted_hours, total_calories):
    print(f"Estimated time hiking: {round(adjusted_hours, 2)} hours.")
    print(f"Estimated calories burned: {round(total_calories, 2)} calories")

    # Since the soul purpose of print_results function is to print the output, 
    # we do not need the function to return any value. Thus we omit the RETURN statement.

distance = 10 # In miles
temp = 75 # In Farenheit
terrain_difficulty = 2 # On a scale from 1 (easiest) to 3 (hard)

adjusted_hours = calc_hours(distance, temp, terrain_difficulty)
total_calories = calc_calories(adjusted_hours, terrain_difficulty)
print_results(adjusted_hours, total_calories)
```
