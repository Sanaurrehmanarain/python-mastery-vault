Q: WAF to find the factorial of n. (n is the parameter)

```python
def cal_fact(n):
    fact = 1
    for i in range (1, n + 1):
        fact *= i
    return fact

cal_fact(5)
```

# RECURSION
When a function calls itself repeatedly. Recursion is when a function calls itself to solve a problem step by step until it reaches a stopping point (base case).

```python
# Recursive function
# Example:
# Prints n to 1 backwards
def show(n):
    if n == 0: # Base case (stopping point)
        return
    print(n)
    show(n - 1) # Recursive call

show(5)
```

##### How It Works:
1. Base Case: When n == 0, the function stops (return).
2. Recursive Case: Prints n, then calls show(n - 1).
3. Recursive Execution Flow:

show(5) → prints 5 → calls show(4)

show(4) → prints 4 → calls show(3)

show(3) → prints 3 → calls show(2)

show(2) → prints 2 → calls show(1)

show(1) → prints 1 → calls show(0)

show(0) → stops recursion

### Difference between Normal Functions and Recursions

Recursion repeatedly calls the same function, which adds new frames to the call stack each time, consuming more memory. In contrast, normal (iterative) function calls do not create multiple stack frames for each iteration, as variables are updated in place. The key difference is that recursion uses additional stack memory for each function call, while an iterative function maintains a single frame in memory.

Recursive Function (Uses Stack Memory)

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(factorial_recursive(5))  # Output: 120
```

Explanation: This function calls itself, creating new stack frames for each call.

Iterative Function (Uses Constant Memory)

```python
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial_iterative(5))  # Output: 120
```

Explanation: This function uses a loop, modifying variables in place without extra stack usage.

Q: Write a recursive function to calculate the sum of first n natural numbers.

```python
# Recursive Function
def calc_sum(n):
    if n == 0:
        return 0
    return calc_sum(n - 1) + n

calc_sum(5)
```

```python
# Iterative (Normal) Function
def calc_sum_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(calc_sum_iterative(5)) 
```

```python
# Formula-Based Approach (Most Efficient)
def calc_sum_formula(n):
    return (n * (n + 1)) // 2  # Sum formula: n(n+1)/2

print(calc_sum_formula(5))  # Output: 15
```

Q: Write a recursive function to print all elements in a list. Hint: use list & index as parameters.

```python
# Recursive function
def print_list(list, idx = 0):
    if idx == len(list): # Base case: stop when index reaches list length
        return
    print(list[idx]) # Print current element
    print_list(list, idx + 1) # Recursive call with next index

fruits = ["mango", "apple", "banana", "kiwi", "pineapple", "grapes"]

print_list(fruits, 0)
```

```python
# Iterative function
def print_list_iterative(list):
    for item in list:
        print(item)

print_list_iterative(fruits)
```
