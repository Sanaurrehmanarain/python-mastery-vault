# TUPLES ARE IMMUTABLE

```python
tup = (2, 56, 234, 345, 435, 1, 3, 4, 3)

print(type(tup))
print(tup[3]) # you can access to any element in tuple by using indexing.
```

```python
tup[1] = 5 # tuples are immutable. you can not change or assign new items.
```

```python
tup[1]
```

```python
tup[1 : 3]
```

```python
# returns index of first occurance
tup.index(4)
```

```python
# counts total occurances
tup.count(3)
```

## WAP to count the number of students with the "A" grade in the following tuple.

```python
grade = ("C", "D", "A", "A", "B", "B", "A")

grade.count("A")
```

## Sort them from "A" to "D".

 Method 1 
 
 tuples are immutable, meaning they do not have a sort() method. However, you can sort a tuple by converting it to a list first or using sorted().

```python
sorted_grade = sorted(grade) # Returns a sorted list
print(sorted_grade)
```

Method 2

Convert to List, Sort, and Convert Back

```python
grade_list = list(grade) # Convert to list
grade_list.sort() # Sort the list
sorted_grade_tuple = tuple(grade_list) # Convert back to tuple

print(sorted_grade_tuple)
```

✅ Use sorted() if you just need a sorted result as a list.

✅ Use list().sort() + tuple() if you need the sorted tuple.
