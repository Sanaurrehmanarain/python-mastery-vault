Explanation:
1. x = "global" is a global variable.
2. Inside func(), x = "local" creates a new local variable x that exists only inside func().
3. The global x is NOT changed.
4. print(x) inside func() refers to the local x, so it prints "local".

```python
def func():
    x = "local"
    print(x)

x = "global"
func()
print(x)
```

Explanation:
1. x = "global" is defined globally.
2. Inside func(), x = "local" is a local variable, so it prints "local".
3. The global x is not affected by the local x inside func().
4. print(x) outside func() prints "global".

```python
def func():
    global x
    print("1-------", x)
    x = "local"
    print("2-------", x)

x = "global"
func()
print("3-------", x)
```

Explanation:
1. x = "global" is defined globally.
2. Inside func(), we use global x. This means:
    - Python does not create a new local variable x.
    - Instead, it refers to the global x directly.
3. print("1-------", x) prints "global" because x was "global" before modification.
4. x = "local" modifies the global x (instead of creating a local variable).
5. print("2-------", x) now prints "local".
6. After calling func(), the global x has been changed to "local", so print("3-------", x) prints "local".

The `nonlocal` keyword is used inside nested functions (a function inside another function). It allows a variable to be modified in an enclosing (but non-global) scope.

```python
def outer_func():
    x = 50  # x is in the enclosing (non-global) scope

    def inner_func():
        nonlocal x  # Refers to 'x' from outer_func
        x = 100  # Modifies 'x' from outer_func

    print("1-------", x)  # Prints 50
    inner_func()  # Calls inner_func, which modifies 'x' to 100
    print("2-------", x)  # Prints 100

x = 20  # Global variable (not related to the 'x' inside outer_func)
outer_func()
print("3-------", x)  # Prints 20 (global 'x' remains unchanged)
```

Explanation
1. x = 50 is defined inside outer_func(), making it a local variable for outer_func.
2. inner_func() is a nested function inside outer_func(), and it uses nonlocal x.
    - This means x inside inner_func() refers to the x from outer_func() (enclosing scope).
    - The assignment x = 100 modifies the x from outer_func(), not a new local variable.
3. Flow of execution:
    - print("1-------", x) prints 50 before calling inner_func().
    - inner_func() is called, modifying x from 50 to 100.
    - print("2-------", x) prints 100 after modification.
4. The global x = 20 is completely separate from the x inside outer_func(), so:
    - print("3-------", x) prints 20 (unchanged global x).

```python
def outer():
    x = "outer variable"  # This is in the enclosing scope

    def inner():
        nonlocal x  # Refers to 'x' from the enclosing function (outer)
        print("1------", x)  # Prints "outer variable"
        x = "modified by inner"  # Modifies 'x' from outer function
        print("2------", x)  # Prints "modified by inner"

    inner()
    print("3------", x)  # 'x' was changed by inner()

outer()
```

Explanation:
1. x = "outer variable" is created inside outer(), so it belongs to the enclosing function's scope.
2. Inside inner(), we use nonlocal x, which means:
    - Python does not create a new local variable x inside inner().
    - Instead, x refers to the variable x from outer().
3. print("1------", x) prints "outer variable".
4. x = "modified by inner" modifies the x from outer().
5. print("2------", x) prints "modified by inner".
6. After calling inner(), print("3------", x) inside outer() prints "modified by inner", showing that the value was changed.
