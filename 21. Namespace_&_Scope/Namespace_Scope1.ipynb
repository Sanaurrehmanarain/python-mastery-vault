##### **GLOBAL VARIABLE**
- Available anywhere
- Stored in the global namespace
- Has a global scope

##### **LOCAL VARIABLE**
- Available in a function
- Stored in a function's local namespace
- Has a local scope

```python
# Example 1
unit = "feet"
length = 100
width = 140
area = length * width

print(f"Area: {area} {unit} squared")
```

```python
# Example 2
unit = "feet"

def rectangle_area(length, width):
    area = length * width
    print(f"Area: {area} {unit} squared")

rectangle_area(100, 140)
rectangle_area(240, 70)
```

##### **GLOBAL NAMESPACE**
In example 1

**UNIT, LENGTH, WIDTH and AREA are in global namespace**
- it is a primary container
- it is accessible anywhere

In example 2

**UNIT** is in global namespace and also upon execution, the function name **rectangle_area** also stored in **global namespace** and they can be accessed anywhere in the program for the remainder of the execution.

When variables are created in created within the function. Python does not store them inside the global namespace. Instead, Python creates a new namespace associated with the function called the function's **local namespace**.

The parameters **length & width** are in **local namespace** as well as the variable **area** which in inside the body of the function. Variables in the glocal namespace (i.e. unit) are accessible from everywhere in the program, including during the execution of a function call. After all the lines of code in the function have been executed, the program wraps up the execution of the function call by **discarding the local namespace** along with its content, and returns to the main program.

```python
def rectangle_area(length, width):
    area = length * width
    return area

def circle_area(radius):
    area = 3.14159 * radius ** 2
    return area

clearing_area = rectangle_area(10, 12)
tent_area = rectangle_area(8, 8)
campfire_area = circle_area(1.5)

free_space = clearing_area - tent_area - campfire_area
print(f"Free space in the clearing: {free_space} feet squared")
```
