##### Exercise 2 - Functions with the same parameter name
Rule: Function parameters are treated as local variables during the execution of the function they belong to. If multiple functions defined in the global scope use the same parameter name, it does not affect the execution of any of the functions. This is because each function has its own local scope, which is independent of the scopes of other functions.

What will this program output?

```python
def calc_triangle_area(base, height):
    area = 0.5 * base * height
    return area

def calc_rect_area(base, height):
    area = base * height
    return area

print(calc_triangle_area(5, 3))
print(calc_rect_area(5, 3))
```

The call to calc_triangle_area() returns 7.5 and the call to calc_rect_area() returns 15. The fact that both functions use base and height as parameter names has no effect on the execution of either of the functions.

##### Exercise 3 - Parameter and global variable with the same name
Rule: As we saw in Exercise 2, function parameters are considered part of a function's local namespace. Therefore when parameter names and global variable names conflict, the more local name is used.

What will this program output?

```python
unit = "feet"

def calc_square_area(length, unit):
    return f"{length**2} {unit} squared"

print(calc_square_area(4, "meters"))
```

This program outputs 16 meters and not 16 feet because Python chooses the most local unit variable to return on line 4. Since function parameters are more local than global variables, Python choses to use "meters" which was pass in as a parameter and not the global unit = "feet".

##### Exercise 4 
- Global variable, parameter, local variable with the same name
Building on Exercise 3, there may be a naming conflict between a local variable, parameter name, and global variable. Again, the most local name is used.

What will this program output?

```python
unit = "feet"

def calc_square_area(length, unit):
    unit = "yards"
    return f"{length**2} {unit} squared"

print(calc_square_area(4, "meters"))
```

The program will output 16 yards. As we saw in Exercise 3, the global unit variable will not be used since there are more local options available. When we create a variable inside a function with the same name as a parameter, we override the parameter's value. This means that on line 5, the unit variable references the function's local unit variable which contains the value "yards".

The namespace at the time the function returns looks the same as it did in Exercise 3; however, unit is equal to "yards", not "meters":
