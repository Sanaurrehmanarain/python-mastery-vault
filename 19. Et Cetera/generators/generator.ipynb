## 🔹 List vs Generator (Memory Usage)

```python
import sys

my_list = [i for i in range(10000)]
print(sum(my_list))
print(sys.getsizeof(my_list), 'bytes')

my_gen = (i for i in range(10000))
print(sum(my_gen))
print(sys.getsizeof(my_gen), 'bytes')
```

####  Explanation

- List comprehension `[ ]`

    - Creates and stores all 10,000 numbers in memory.

    - Memory usage: ~85 KB.

    - Good when you need to reuse the data multiple times.

- Generator comprehension `( )`

    - Produces numbers on demand (lazy evaluation).

    - Memory usage: ~100 bytes (tiny, regardless of size).

    - Good for large datasets or when you only need to iterate once.

**List →** faster for repeated access, but uses more memory.

**Generator →** memory efficient, but can be iterated only once.

# GENERATORS
Generators are simple way of creating iterators. A generator is a function that returns the iterator object which we can iterate upon. A generator is a special type of function in Python that automatically creates an iterator using the yield keyword.

Unlike a normal function, which returns a value and exits, a generator pauses at yield, remembers its state, and resumes from there when called again. This makes it memory efficient and easy to use for large data processing.

Example: Generator for Even Numbers

```python
def even_numbers(max_num):
    num = 0
    while num <= max_num:
        yield num  # Yield pauses execution and remembers the state
        num += 2

# Using the generator in a for loop
for num in even_numbers(10):
    print(num)  
```

Example: Custom Iterator for Even Numbers

```python
class EvenNumbers:
    def __init__(self, max_num):
        self.max_num = max_num
        self.current = 0

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.current > self.max_num:
            raise StopIteration  # End iteration
        val = self.current
        self.current += 2
        return val

# Using the custom iterator
evens = EvenNumbers(10)
for num in evens:
    print(num) 
```

How generator is Different from a Normal Function?

A normal function would return all values at once in a list:

Example: Normal Function for Even Numbers

```python
def even_numbers_list(max_num):
    result = []
    num = 0
    while num <= max_num:
        result.append(num)
        num += 2
    return result

print(even_numbers_list(10))  
```

Disadvantages of this approach:

- Uses more memory (stores all numbers in a list).
- Slower if the range is large.

With a generator, values are generated one at a time and don’t consume unnecessary memory.

Using next() to Manually Control the Generator

You can also manually iterate over a generator using next():

```python
gen = even_numbers(6)
print(next(gen))  
print(next(gen))  
print(next(gen))  
print(next(gen))  
print(next(gen)) # # Raises StopIteration
```

This is useful when you want to process elements one at a time rather than all at once.

Key Advantages of Generators

- ✅ Memory-efficient (does not store all values at once).
- ✅ Lazy Evaluation (generates values when needed).
- ✅ Easy to implement (simpler than custom iterator classes).

```python
def my_func():
    yield "a" # if your fuction contains at least one yield keyword, then you func is called a generator func.
    yield "b"
    yield "c"

x = my_func()
print(next(x))
print(next(x))
print(next(x))
```

When to Use Which?

- Use an iterator when you need more control over iteration behavior (e.g., implementing a custom iterable object).
- Use a generator when working with large data sets, infinite sequences, or streaming data, since it’s more memory efficient.

## Key Differences Between Iterators and Generators

| Feature         | Iterator (Custom Class)                       | Generator (Function with `yield`)        |
|---------------|--------------------------------|---------------------------------|
| **Implementation** | Uses a class with `__iter__()` and `__next__()` | Uses a function with `yield` |
| **Memory Usage** | Stores state explicitly in instance variables | Generates values on-the-fly, saving memory |
| **Ease of Use** | More complex, requires a class and multiple methods | Simpler, just define a function with `yield` |
| **State Management** | Must manually track iteration state | Automatically remembers state between `yield` calls |
| **Performance** | Can be slower for large datasets | More efficient for large or infinite sequences |
| **Reusability** | Can be restarted by creating a new instance | Cannot be restarted once exhausted, but can be re-called |

```python
def list_iterator(list):
    for i in list:
        yield i

a = [1, 2, 3, 4, 5, 6]
my_list = list_iterator(a)

print(next(my_list))
print(next(my_list))
print(next(my_list))
print(next(my_list))
print(next(my_list))
print(next(my_list))
print(next(my_list))
```

We can also use for loops for this.

```python
def list_iterator(list):
    for i in list:
        yield i

a = [1, 2, 3, 4, 5, 6]
my_list = list_iterator(a)

for x in my_list:
    print(x)
```
