# Command Line Arguments & `argparse`
### 1. What are Command Line Arguments?
Command-line arguments are values passed to a Python script when it's executed from the command line. They allow users to provide input without modifying the script.

Example:
Running a script with arguments:
```sh
python script.py arg1 arg2
```
Here, arg1 and arg2 are command-line arguments.

### 2. Accessing Command Line Arguments Using sys.argv
Python provides sys.argv to retrieve command-line arguments.

Example:
```python
import sys

print("Script Name:", sys.argv[0])  # First argument is always the script name
print("Arguments:", sys.argv[1:])  # Remaining arguments
```
Run in terminal:
```sh
python script.py hello world
```
Output:
```less
Script Name: script.py
Arguments: ['hello', 'world']
```
🚨 Limitations of sys.argv:

- No automatic parsing (all arguments are strings).
- No help messages or validation.

### 3. Using argparse for Better Argument Handling
The argparse module provides a powerful way to handle command-line arguments with automatic parsing, validation, and help messages.

Example: Using argparse to Accept Arguments

```python
import argparse

# Create parser
parser = argparse.ArgumentParser(description="A simple script to demonstrate argparse")

# Add arguments
parser.add_argument("name", type=str, help="Your name")
parser.add_argument("age", type=int, help="Your age")
parser.add_argument("--city", type=str, default="Unknown", help="Your city (optional)")

# Parse arguments
args = parser.parse_args()

# Print values
print(f"Hello {args.name}, you are {args.age} years old and live in {args.city}.")
```
Running the script:

python script.py Alice 25 --city "New York"

Output:

Hello Alice, you are 25 years old and live in New York.

### 4. Key Features of argparse
- ✅ Positional arguments (mandatory inputs).
- ✅ Optional arguments (--option or -o flags).
- ✅ Type checking (e.g., int, float).
- ✅ Default values (e.g., default="Unknown").
- ✅ Help messages (help="description").
