# PYTHON DEBUGGER
The `Python Debugger (pdb)` is a built-in tool that helps you pause and inspect your code while it runs. It allows you to find bugs by stepping through the code line by line, examining variables, and understanding where things go wrong.

### How to Use pdb
There are two main ways to use Python’s debugger:

1. Using pdb.set_trace() in the code
2. Running the script with python -m pdb script.py


### 1️⃣ Debugging with pdb.set_trace()
You can insert pdb.set_trace() anywhere in your script to pause execution and enter debug mode.

````python
import pdb

def divide(a, b):
    pdb.set_trace()  # Debugger starts here
    result = a / b
    return result

print(divide(10, 2))
````

🔍 When the debugger starts, you can use commands like:

- n → Move to the next line
- s → Step into a function
- c → Continue execution
- p variable → Print the value of a variable

### 2️⃣ Running a Script in Debug Mode
You can run a Python script in debug mode from the command line:
```sh
python -m pdb my_script.py
```
This allows you to step through the script before it runs.

### 3️⃣ Example Debugging Session
Code with a Bug:

````python
def add_numbers(a, b):
    total = a + b
    return total

x = 5
y = "10"  # This will cause an error (str + int)
print(add_numbers(x, y))
````

Debugging with `pdb`

````python
import pdb

def add_numbers(a, b):
    pdb.set_trace()  # Pause execution here
    total = a + b  # Error occurs here
    return total

x = 5
y = "10"  # This will cause an error
print(add_numbers(x, y))
````

🔍 Debugging Output
```bash
> test.py(5)add_numbers()
-> total = a + b
(pdb) p a
5
(pdb) p b
'10'
(pdb) n
TypeError: unsupported operand type(s) for +: 'int' and 'str'
```
Here, we see that `b` is a string `('10')`, causing a TypeError when adding it to `a` (an integer).
