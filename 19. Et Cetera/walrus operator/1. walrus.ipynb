# Walrus Operator
The walrus operator (`:=`) in Python is called the assignment expression.

It was introduced in Python 3.8 and lets you assign a value to a variable as part of an expression.

## Normal Assignment vs Walrus Operator

Normally, assignment (`=`) is a separate step:

```python
n = len([1, 2, 3])
if n > 2:
    print(f"List is long ({n} elements)")
```

With the walrus operator, you can do it inline:

```python
if (n := len([1, 2, 3])) > 2:
    print(f"List is long ({n} elements)")
```

Here, `n := len([1, 2, 3])` both calculates the length and assigns it to `n` inside the condition.

## Practical Examples

### 1. While loop with input

```python
# Without walrus
line = input("Enter text: ")
while line != "quit":
    print("You entered:", line)
    line = input("Enter text: ")

# With walrus
while (line := input("Enter text: ")) != "quit":
    print("You entered:", line)
```

### 2. Filtering data

```python
numbers = [1, 2, 3, 4, 5, 6]

# Keep even numbers and print them
evens = [n for n in numbers if (rem := n % 2) == 0]
print(evens)  # [2, 4, 6]
```

### 3. Avoiding repeated function calls

```python
import math

# Without walrus
x = math.sqrt(25)
if x > 4:
    print(x)

# With walrus
if (x := math.sqrt(25)) > 4:
    print(x)  # prints 5.0
```

The walrus operator `:=` lets you assign and use a value in the same expression, which often makes code shorter and avoids repeating calculations.
