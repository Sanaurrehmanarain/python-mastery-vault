# Type Hints
**type hints** are an important part of writing modern, clean, and maintainable Python. Let’s go step by step with explanations and examples you can drop into your Jupyter Notebook.

#### What are Type Hints?

- ype hints (or type annotations) let you tell Python what type of data you expect for variables, function parameters, and return values.

- They don’t change how the code runs — Python still allows dynamic typing.

- But they improve:

    - Readability (code is easier to understand at a glance).

    - Tooling (IDEs give better autocomplete & error checking).

    - Static analysis (with tools like `mypy` to catch type errors before runtime).

##### ✅ Basic Syntax
```python
# Variable type hints
age: int = 25
name: str = "Alice"
is_active: bool = True
scores: list[int] = [90, 80, 95]
```

## ✅ Functions with Type Hints

````python
def greet(name: str, age: int) -> str:
    return f"Hello, {name}. You are {age} years old."

print(greet("Alice", 25))
````

- `name: str` → parameter must be a string.

- `age: int` → parameter must be an integer.

- `-> str` → function returns a string.

## ✅ Complex Type Hints

Python’s `typing` module (or built-in types in Python 3.9+) gives richer annotations:

````python
from typing import Dict, List, Tuple, Optional

# List of integers
numbers: List[int] = [1, 2, 3]

# Dictionary with string keys and int values
ages: Dict[str, int] = {"Alice": 25, "Bob": 30}

# Tuple of specific types
point: Tuple[int, int] = (10, 20)

# Optional type (int or None)
maybe_age: Optional[int] = None
````

## ✅ With Generators

````python
from typing import Generator

def count_up_to(n: int) -> Generator[int, None, None]:
    for i in range(n):
        yield i
````

#### 🔹 Why Use Type Hints?

1. Clarity: makes code easier to understand for humans.

2. Early error detection: static checkers like `mypy` can catch mistakes:
```python
def add(x: int, y: int) -> int:
    return x + y

add("hello", 5)  # mypy will flag this as an error
```
4. IDE Support: better autocomplete and inline documentation.

## ✅ Example with Comprehensions and Type Hints

````python
names: list[str] = ["Alice", "Bob", "Charlie"]

# List comprehension with type hint
name_lengths: list[int] = [len(name) for name in names]

# Dict comprehension with type hint
name_dict: dict[str, int] = {i: name for i, name in enumerate(names)}
````

## 🔹 Python Type Hint Cheat Sheet
### ✅ Basic Types

| Type Hint | Example                  | Meaning               |
| --------- | ------------------------ | --------------------- |
| `int`     | `age: int = 25`          | Integer               |
| `float`   | `pi: float = 3.14`       | Floating-point number |
| `str`     | `name: str = "Alice"`    | String                |
| `bool`    | `is_active: bool = True` | Boolean               |
| `None`    | `data: None = None`      | No value              |

### ✅ Collections

| Type Hint         | Example                                | Meaning                           |
| ----------------- | -------------------------------------- | --------------------------------- |
| `list[int]`       | `nums: list[int] = [1, 2, 3]`          | List of integers                  |
| `tuple[int, str]` | `pair: tuple[int, str] = (1, "Alice")` | Tuple with fixed types            |
| `tuple[int, ...]` | `coords: tuple[int, ...] = (1, 2, 3)`  | Tuple of variable length          |
| `set[str]`        | `names: set[str] = {"Alice", "Bob"}`   | Set of strings                    |
| `dict[str, int]`  | `ages: dict[str, int] = {"Alice": 25}` | Dict with string keys, int values |

### ✅ From typing
```python
from typing import List, Dict, Tuple, Set, Optional, Union, Any, Generator
```

| Type Hint                    | Example                                | Meaning                           |
| ---------------------------- | -------------------------------------- | --------------------------------- |
| `List[int]`                  | `nums: List[int] = [1, 2, 3]`          | Same as `list[int]` (older style) |
| `Dict[str, int]`             | `ages: Dict[str, int] = {"Alice": 25}` | Same as `dict[str, int]`          |
| `Tuple[int, str]`            | `point: Tuple[int, str] = (1, "x")`    | Tuple with fixed types            |
| `Set[int]`                   | `s: Set[int] = {1, 2, 3}`              | Set of integers                   |
| `Optional[int]`              | `age: Optional[int] = None`            | `int` or `None`                   |
| `Union[int, str]`            | `value: Union[int, str] = "hi"`        | Either int or str                 |
| `Any`                        | `data: Any = "whatever"`               | Any type (no checking)            |
| `Generator[int, None, None]` | `gen: Generator[int, None, None]`      | Generator yielding ints           |

### ✅ Functions
```python
def greet(name: str, age: int) -> str:
    return f"Hello {name}, age {age}"
```
- Parameters: `name: str`, `age: int`

- Return type: `-> str`

### ✅ Special Cases

| Type Hint                   | Example                                      | Meaning                                                |
| --------------------------- | -------------------------------------------- | ------------------------------------------------------ |
| `Callable[[int, int], int]` | `add: Callable[[int, int], int]`             | A function taking 2 ints and returning int             |
| `Iterable[str]`             | `data: Iterable[str]`                        | Any iterable of strings (list, tuple, generator, etc.) |
| `Literal["yes", "no"]`      | `choice: Literal["yes", "no"]`               | Must be either `"yes"` or `"no"`                       |
| `TypedDict`                 | `class User(TypedDict): name: str; age: int` | Dict with specific key-value structure                 |

- Use built-in generics like `list[str]`, `dict[str, int]` (Python 3.9+).

- Use `typing` for more advanced cases (`Optional`, `Union`, `Callable`, etc.).

- Pair with `mypy` to catch mistakes early.
