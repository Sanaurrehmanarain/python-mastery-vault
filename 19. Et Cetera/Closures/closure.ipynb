# CLOSURE
In Python, a **closure** is a function that remembers the variables from its enclosing scope, even after the scope has finished executing. This allows the function to retain access to those variables even when it is called outside their original scope.

How Closures Work

A closure occurs when:

1. A nested function (inner function) is defined inside an enclosing function (outer function).
2. The inner function references variables from the enclosing function.
3. The outer function returns the inner function.

Example of a Closure

```python
def outer_function(message):
    def inner_function():
        print(message)  # Inner function "remembers" the message
    return inner_function  # Returning the inner function

closure_func = outer_function("Hello, Closure!")
closure_func() 
```

Even though outer_function has finished executing, closure_func still retains access to the message variable because of the closure.

Use Cases of Closures

- Data encapsulation: Closures can help store private variables.
- Function factories: Creating specialized functions dynamically.
- Decorators: Closures are used to modify the behavior of functions.

Example: Function Factory

```python
def multiplier(factor):
    def multiply(number):
        return number * factor  # Remembers 'factor' even after 'multiplier' exits
    return multiply

double = multiplier(2)
triple = multiplier(3)

print(double(5))  
print(triple(5)) 
```

Here, double remembers factor = 2, and triple remembers factor = 3, demonstrating closure behavior.

```python
def outerFunction(text):
    def innerFunction():
        print(text)
    return innerFunction

a = outerFunction("Hello")
del outerFunction
a()
```

A closure allows a function to remember and access variables from its enclosing scope, even after the outer function has been deleted.

In the given code, `a` retains access to the `text` variable, even though `outerFunction` is deleted. This is because `innerFunction` forms a closure, keeping a reference to `text` from `outerFunction`.

```python
def nth_power(exponent):
    def pow_of(base):
        return pow(base, exponent)
    return pow_of

square = nth_power(2)
print(square(4))
print(square(5))

print() # to print an empty line

cube = nth_power(3)
print(cube(4))
print(cube(5))
```
