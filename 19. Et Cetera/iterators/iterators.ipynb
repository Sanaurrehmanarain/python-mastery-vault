# ITERATION
refers to the process of going over elements in a collection (like a list, tuple, or dictionary).
# ITERATOR
Is an object which can be used to iterate over a collection.
It has two special methods

`__iter__()` and `__next__()`

Iterator is an object that enables iteration. It implements the `__iter__()` method (which returns the iterator object itself) and the `__next__()` method (which returns the next item in the sequence and raises `StopIteration` when there are no more items).

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
it = iter(a)
next(it)
```

```python
next(it) # it gives you the values in the list one by one
```

```python
next(it)
```

When your iterator is exhausted, it will give you exception `StopIteration`.

```python
for i in a:
    print(i)
```

`for loop` uses the same iterator mechanism to iterate over all these element of the collection.

### Custom Iterator Class

```python
class ListIterator:
    def __init__(self, list):
        self.__list = list
        self.__index = -1

    def __iter__(self):
        return self
    
    def __next__(self):
        self.__index += 1
        if self.__index == len(self.__list):
            raise StopIteration
        return self.__list[self.__index]

a = [1, 2, 3, 4, 5, 6]
mylist = ListIterator(a)
it = iter(mylist)

print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
class MyIterator:
    def __init__(self, start, end): #  Initializes the iterator with a start and end value.
        self.current = start
        self.end = end

    def __iter__(self): # Returns the iterator object itself.
        return self  # The iterator object itself

    def __next__(self): # Returns the next value and stops when the end is reached.
        if self.current >= self.end:
            raise StopIteration  # Stop when the limit is reached
        value = self.current
        self.current += 1
        return value

# Using the custom iterator
my_iter = MyIterator(1, 5)
for num in my_iter: # When used in a for loop, it iterates from 1 to 4 (excluding 5).
    print(num)
```

Why Use a Custom Iterator Class? (When For Loops Exist)

Yes, for loops work well with iterables, but custom iterator classes are useful when:

- You need to iterate over a custom data structure (e.g., linked lists, trees, or database records). A for loop works on sequences (lists, tuples, etc.), but you might want to define custom iteration logic.
- You want to control the iteration behavior, such as skipping elements, stopping under specific conditions, or iterating indefinitely.
- Memory efficiency: A custom iterator can generate values on demand instead of storing them all in memory at once (e.g., reading a huge file line by line instead of loading it all into memory).
- Lazy evaluation: Iterators allow you to work with sequences where computing all elements in advance is expensive or impossible (like infinite sequences).

For example, let's say you have a class that iterates over even numbers up to a limit:

```python
class EvenNumbers:
    def __init__(self, max_num):
        self.max_num = max_num
        self.current = 0

    def __iter__(self):
        return self  # An iterator must return itself

    def __next__(self):
        if self.current > self.max_num:
            raise StopIteration  # End iteration
        val = self.current
        self.current += 2
        return val

# Using the custom iterator in a for loop
for num in EvenNumbers(10):
    print(num)  
```
