# OPERATOR OVERLOADING
Python provides special methods (also called dunder methods) to define how operators behave with objects of a class.

Overloading the `+` Operator

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)  # Custom behavior for +

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(4, 5)
result = p1 + p2  # This calls p1.__add__(p2)

print(result)  
```

Here, we overloaded the `+` operator by defining the `__add__` method so that `Point(2,3) + Point(4,5)` results in `Point(6,8)`.

Example: Overloading `*` Operator

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return Number(self.value * other.value)

    def __str__(self):
        return str(self.value)

num1 = Number(3)
num2 = Number(4)
result = num1 * num2  # Calls num1.__mul__(num2)

print(result)  
```

Here, we overloaded the * operator so that multiplying two Number objects returns a new Number object.

Why Use Operator Overloading?

- ✅ Makes code more intuitive and readable.
- ✅ Allows user-defined objects to behave like built-in types.
- ✅ Helps in customizing class behavior to match real-world scenarios.

### Common Operator Overloading Methods

| Operator | Method |
|----------|--------|
| `+` | `__add__(self, other)` |
| `-` | `__sub__(self, other)` |
| `*` | `__mul__(self, other)` |
| `/` | `__truediv__(self, other)` |
| `//` | `__floordiv__(self, other)` |
| `%` | `__mod__(self, other)` |
| `==` | `__eq__(self, other)` |
| `!=` | `__ne__(self, other)` |
| `<` | `__lt__(self, other)` |
| `>` | `__gt__(self, other)` |
| `<=` | `__le__(self, other)` |
| `>=` | `__ge__(self, other)` |

```python
import math

class Circle:
    def __init__(self, radius):
        self.__radius = radius

    def setRadius(self, radius):
        self.__radius = radius

    def getRadius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def __add__(self, circle_object):
        return Circle(self.__radius + circle_object.__radius)
    
    def __lt__(self, circle_object):
        return (self.__radius < circle_object.__radius)
    
    def __gt__(self, circle_object):
        return (self.__radius > circle_object.__radius)

c1 = Circle(2)
c2 = Circle(3)
c3 = c1 + c2
print(c1.getRadius())
print(c2.getRadius())
print(c3.getRadius())

print(c1 < c2)
print(c1 > c2)
```
