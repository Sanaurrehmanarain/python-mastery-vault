# DECORATORS
A decorator in Python is a special function that modifies the behavior of another function without changing its code. Decorators allow you to wrap a function and add functionality before or after it executes.

### How Decorators Work

A decorator is just a function that takes another function as an argument and returns a new function that enhances or modifies the original one.

Basic Decorator Example

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator  # Applying the decorator
def say_hello():
    print("Hello!")

say_hello()
```

Here, @my_decorator modifies say_hello() without altering its original code.

### Using Decorators with Arguments

If the decorated function takes arguments, the wrapper function should accept them using `*args` and `**kwargs`.

```python
def repeat_decorator(func):
    def wrapper(*args, **kwargs):
        print("Function is about to run")
        result = func(*args, **kwargs)
        print("Function has finished")
        return result
    return wrapper

@repeat_decorator
def add(a, b):
    return a + b

print(add(3, 4))
```

Here, the decorator logs messages before and after running `add(a, b)`.

### Built-in Decorators in Python
Python has several built-in decorators, such as:

- `@staticmethod` → Defines a static method inside a class.
- `@classmethod` → Defines a method that takes cls as the first argument.
- `@property` → Converts a method into a read-only property.

### Key Use Cases of Decorators
- ✔ Logging
- ✔ Authentication
- ✔ Performance monitoring (timing a function)
- ✔ Modifying function behavior (e.g., caching, validation)

```python
def decorator_divide(func):
    def wrapper_func(a, b):
        print("divide", a, "and", b)
        if b == 0:
            print("division with zero is not allowed")
            return
        return a / b
    return wrapper_func

@decorator_divide
def divide(x, y):
    return x / y

print(divide(15, 5)) 
```

```python
from time import time

def timing(func):
    def wrapper_func(*args, **kwargs):
        start = time()
        print("Start time: ", start)
        result = func(*args, **kwargs)
        end = time()
        print("Ending time: ", end)
        print("Elapsed time: {}".format(end - start))
        return result

    return wrapper_func

@timing 
def my_func(num):
    sum = 0
    for i in range(num + 1):
        sum += i
    return sum

print(my_func(2000000))
    
```
