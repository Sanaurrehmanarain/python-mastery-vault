# Property Decorator
The `@property` decorator in Python is used to turn a method into a read-only attribute, and optionally give it getter, setter, and deleter behavior.

It allows you to add logic when reading, writing, or deleting an attribute, while still using attribute-style access (not method calls).

**🔹 Benefits**

- Encapsulation: Hide internal details while exposing a clean interface.

- Add validation when setting attributes.

- Compute values dynamically when accessed.

- Still access them like normal attributes (`obj.attr` instead of `obj.method()`).

```python
# Example
class Person:
    def __init__(self, name):
        self._name = name  # convention: private attribute

    @property
    def name(self):
        """Getter method (read-only by default)"""
        return self._name

    @name.setter
    def name(self, value):
        """Setter method (called when assigning)"""
        if not value.strip():
            raise ValueError("Name cannot be empty")
        self._name = value

    @name.deleter
    def name(self):
        """Deleter method (called when using del)"""
        print("Deleting name...")
        del self._name


# Usage
p = Person("Sana")
print(p.name)   # calls getter -> Sana

p.name = "Ali"  # calls setter
print(p.name)

del p.name      # calls deleter
```

- `@property` makes a method act like an attribute.

- You can add getter, setter, and deleter methods using decorators.

- It’s mainly used for controlled access to attributes.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

rectangle = Rectangle(3, 4)

print(rectangle.width)
print(rectangle.height)
```

```python
# First Version
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property
    def width(self):
        return f'{self._width: .1f}cm'

    @property
    def height(self):
        return f'{self._height: .1f}cm'

    @width.setter
    def width(self, new_width):
        if new_width > 0:
            self._width = new_width

        else:
            print("Width must be greater than zero")

    @height.setter
    def height(self, new_height):
        if new_height > 0:
            self._height = new_height

        else:
            print("Height must be greater than zero")

rectangle = Rectangle(3, 4)
rectangle.width = 5
rectangle.height = 6

print(rectangle.width)
print(rectangle.height)
```

```python
# Second Version
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property
    def width(self):
        return f'{self._width: .1f}cm'

    @property
    def height(self):
        return f'{self._height: .1f}cm'

    @width.setter
    def width(self, new_width):
        if new_width > 0:
            self._width = new_width

        else:
            print("Width must be greater than zero")

    @height.setter
    def height(self, new_height):
        if new_height > 0:
            self._height = new_height

        else:
            print("Height must be greater than zero")

    @width.deleter
    def width(self):
        del self._width
        print('Width has been deleted')

    @height.deleter
    def height(self):
        del self._height
        print('Height has been deleted')


rectangle = Rectangle(3, 4)
rectangle.width = 5
rectangle.height = 6

del rectangle.width
del rectangle.height
```

## Explanation of the above codes

### 📒 First Version (Getter + Setter)

```python
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height
```

- `__init__` initializes the object with `width` and `height`.

- Notice the underscore (`_width`, `_height`): that’s a naming convention for “private” attributes (not meant to be accessed directly).

#### Getter Methods

```python
    @property
    def width(self):
        return f'{self._width: .1f}cm'

    @property
    def height(self):
        return f'{self._height: .1f}cm'
```

- `@property` turns these methods into attributes.

- Instead of calling `rectangle.width()`, you just write `rectangle.width`.

- The formatting `f'{self._width: .1f}cm'` means:

    - `.1f` → one decimal place (e.g., `5.0`)

    - `"cm"` → add "cm" after the number

So if `_width = 5`, then `rectangle.width` returns `" 5.0cm"`.

#### Setter Methods

```python
    @width.setter
    def width(self, new_width):
        if new_width > 0:
            self._width = new_width
        else:
            print("Width must be greater than zero")
```

    - `@width.setter` defines what happens when you assign a new value:

```python
rectangle.width = 5   # Calls this setter
```

    - It validates the value. If `new_width <= 0`, it prints a warning instead of setting it.

The same logic applies to `height`.

#### Usage

```python
rectangle = Rectangle(3, 4)
rectangle.width = 5     # calls setter
rectangle.height = 6    # calls setter

print(rectangle.width)  # calls getter → " 5.0cm"
print(rectangle.height) # calls getter → " 6.0cm"
```

### 📒 Second Version (Getter + Setter + Deleter)

This code is the same as the first one but adds deleters:

```python
    @width.deleter
    def width(self):
        del self._width
        print('Width has been deleted')

    @height.deleter
    def height(self):
        del self._height
        print('Height has been deleted')
```

- `@width.deleter` defines what happens if you call:

```python
del rectangle.width
```

- This deletes the underlying attribute (`_width`) from the object and prints a message.

#### Usage

```python
rectangle = Rectangle(3, 4)
rectangle.width = 5
rectangle.height = 6

del rectangle.width   # calls deleter → removes _width
del rectangle.height  # calls deleter → removes _height
```

1. Getter (`@property`) → allows reading values like attributes, but you can add logic (like formatting or calculations).

2. Setter (`@<name>.setter`) → allows assigning values with validation.

3. Deleter (`@<name>.deleter`) → defines what happens when you `del` an attribute.
