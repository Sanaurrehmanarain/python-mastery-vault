# Lambda
A Lambda function in Python is an anonymous function defined using the lambda keyword and are defined without a name. It can take any number of arguments but has only one expression. Lambda functions are used for short, simple operations where defining a full function with def would be unnecessary.

lambda arguments: expression
- lambda → Keyword to define the function
- arguments → Inputs to the function (can be multiple, separated by commas).
- expression → The single computation or operation returning a value.

```python
def add(x, y):
    return x + y
print(add(1, 2))
```

You can write the same above function in single line using lambda

```python
print((lambda x, y: x + y) (1, 2))
```

```python
# Example:
add = lambda x, y: x + y
print(add(5, 3))
```

```python
def sqr(x):
    return x ** 2
print(sqr(3))
```

```python
sqr = lambda x : x ** 2
print(sqr(3))
```

### map()

```python
# Example 1
# using iteration
num = [1, 2, 3, 4, 5]
for i in num:
    print(i * i, end = ' ')
```

```python
# using map and lambda
x = map(lambda a : a * a, num)
print(list(x))
```

```python
# Example 2
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared) 
```

### filter()

```python
# filter
num = [1, 2, 3, 4, 5]
even_num = filter(lambda x : x % 2 == 0, num)
print(list(even_num))
```

Lambdas are useful for quick operations but should not replace regular functions for complex logic.
