### lambda

```python
# def double(x):
#     return x * 2

# def add(x, y):
#     return x + y

# def product(x, y, z):
#     return x * y * z
```

```python
# lambda
double = lambda x : x * 2
add = lambda x, y : x + y
product = lambda x, y, z : x * y * z
# These lambda func are equivalent to the above func
```

```python
print(double(8))
print(add(10, 15))
print(product(3, 4, 5))
```

### filter, reduce & map

```python
# map
my_list = [2, 5, 8, 10, 9, 3]

a = map(lambda x : x * 2, my_list) # double each element in the list
print(a)
```

to see the proper output you need to cast the output into the list

```python
print(list(a))
```

```python
my_list2 = [1, 4, 7, 8, 5, 1]

b = map(lambda x, y : x + y, my_list, my_list2) # add elements of both lists in the new list
print(list(b))
```

```python
# filter
c = filter(lambda x : x % 2 == 0, my_list) # filter out the even num
print(list(c))
```

```python
d = filter(lambda x : True if x > 5 else False, my_list) # return the values > than 5
print(list(d))
```

```python
# reduce
from functools import reduce # to use the reduce func we need to import this
e = reduce(lambda x, y : x + y, my_list)
print(e)
```
