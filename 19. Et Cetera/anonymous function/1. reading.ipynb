# 1. Lambda Function
A lambda function is an anonymous function (a function without a name) that can have multiple arguments but only one expression.

Syntax:
```python
lambda arguments: expression

Example:

````python
add = lambda x, y: x + y
print(add(5, 3))  
````

This is equivalent to:

````python
def add(x, y):
    return x + y
````

# 2. Map Function
The map() function applies a given function to all items in an iterable (list, tuple, etc.) and returns a new iterable.

Syntax:

```python
map(function, iterable)

Example:

````python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared) 
````

This applies lambda x: x ** 2 to each element in numbers.

# 3. Filter Function
The filter() function filters elements in an iterable based on a condition (a function that returns True or False).

Syntax:
```python
filter(function, iterable)

Example:

````python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  
````

This filters out odd numbers and keeps only even numbers.

# 4. Reduce Function
The reduce() function applies a function cumulatively to all elements in an iterable to reduce it to a single value. It is available in the functools module.

Syntax:
```python
from functools import reduce
reduce(function, iterable)

Example:

````python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  
````

Here, reduce computes 1 * 2 * 3 * 4 = 24.

### Key Differences:

| Function  | Purpose |
|-----------|---------|
| `lambda`  | Creates small anonymous functions. |
| `map`     | Applies a function to all items in an iterable. |
| `filter`  | Selects items that match a condition. |
| `reduce`  | Combines elements into a single value using a function. |
