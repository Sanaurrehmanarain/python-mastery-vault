Python's built-in types (like strings, lists, dictionaries, sets, etc.) come with various methods to perform operations. Below is a list of some famous and commonly used methods grouped by type:
# Python Built-in Type Methods
---
## String Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `str.upper()`               | Converts a string to uppercase.                                                |
| `str.lower()`               | Converts a string to lowercase.                                                |
| `str.capitalize()`          | Capitalizes the first character.                                               |
| `str.title()`               | Converts the string to title case.                                             |
| `str.strip()`               | Removes leading and trailing whitespace.                                       |
| `str.split()`               | Splits the string into a list based on a delimiter.                            |
| `str.join(iterable)`        | Joins elements of an iterable with the string as a separator.                  |
| `str.find(sub)`             | Returns the index of the first occurrence of a substring, or -1 if not found.  |
| `str.replace(old, new)`     | Replaces occurrences of a substring with another.                              |
| `str.isalpha()`             | Checks if all characters are alphabetic.                                       |
| `str.isdigit()`             | Checks if all characters are numeric.                                          |
| `str.startswith(prefix)`    | Checks if the string starts with a prefix.                                     |
| `str.endswith(suffix)`      | Checks if the string ends with a suffix.                                       |

---

## List Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `list.append(item)`         | Adds an item to the end of the list.                                            |
| `list.extend(iterable)`     | Extends the list by appending elements from an iterable.                       |
| `list.insert(index, item)`  | Inserts an item at a specified position.                                        |
| `list.remove(item)`         | Removes the first occurrence of an item.                                       |
| `list.pop([index])`         | Removes and returns the item at a given position (or last item by default).     |
| `list.index(item)`          | Returns the index of the first occurrence of an item.                         |
| `list.count(item)`          | Counts the number of occurrences of an item.                                   |
| `list.sort()`               | Sorts the list in place.                                                       |
| `list.reverse()`            | Reverses the list in place.                                                    |
| `list.copy()`               | Returns a shallow copy of the list.                                            |

---

## Dictionary Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `dict.keys()`               | Returns a view of all keys.                                                    |
| `dict.values()`             | Returns a view of all values.                                                  |
| `dict.items()`              | Returns a view of key-value pairs.                                             |
| `dict.get(key[, default])`  | Returns the value for a key or a default if the key is missing.                |
| `dict.update(other_dict)`   | Updates the dictionary with key-value pairs from another dictionary.           |
| `dict.pop(key[, default])`  | Removes and returns the value for a key.                                       |
| `dict.clear()`              | Removes all items from the dictionary.                                         |
| `dict.setdefault(key[, default])` | Returns the value for a key, or inserts it with a default if missing.   |

---

## Set Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `set.add(item)`             | Adds an item to the set.                                                       |
| `set.remove(item)`          | Removes an item from the set; raises an error if not present.                  |
| `set.discard(item)`         | Removes an item without raising an error if it's not present.                  |
| `set.pop()`                 | Removes and returns an arbitrary set element.                                  |
| `set.union(other_set)`      | Returns the union of sets.                                                     |
| `set.intersection(other_set)` | Returns the intersection of sets.                                            |
| `set.difference(other_set)` | Returns the difference of sets.                                                |
| `set.symmetric_difference(other_set)` | Returns the symmetric difference of sets.                           |
| `set.issubset(other_set)`   | Checks if the set is a subset.                                                 |
| `set.issuperset(other_set)` | Checks if the set is a superset.                                               |

---

## Tuple Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `tuple.count(item)`         | Counts occurrences of an item.                                                 |
| `tuple.index(item)`         | Returns the index of the first occurrence of an item.                          |

---

## Common Type Methods
| **Method**                  | **Description**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| `len(obj)`                  | Returns the length of an object (string, list, etc.).                          |
| `type(obj)`                 | Returns the type of the object.                                                |
| `isinstance(obj, class)`    | Checks if an object is an instance of a class.                                 |
| `id(obj)`                   | Returns the memory address of an object.                                       |


## Some more examples of different functions

`divmod(x, y)` outputs the quotient and the remainder in a tuple

````python
divmod(38, 9)
````

`round()` round the number to the closest decimal

````python
round(15.85467)
````

````python
round(15.85467, 3)
````

`isintance()` returns True, if the first argument is an instance of that class, Multiple classes can also be checked at once.
```python
print(isinstance(1, int))
print(isinstance(1.0, int))
print(isinstance(1.0,(int, float)))
```

````python
isinstance(3, int) # indeed 3 is an instance of integer
````

````python
isinstance(3.4, int)
````

`pow(x, y, z)` x raise to the power y and remainder by z

````python
pow(2, 4, 7) # (2 ** 4 ) % 7
````
