## Exercise 3 - Replacing characters

### 🚨 New Concept: Method chaining

In programming, we often want to apply multiple methods to the same data in sequence. Similar to Exercise 2, we can accomplish this by assigning each intermediate step back to the variable like this...

```python
my_var = "     text!!!    "
my_var = my_var.strip()
my_var = my_var.rstrip("!")
my_var
```

However, Python provides us with a more efficient way to do this called method chaining. Method chaining allows us to apply a method to the result of another method like this...

```python
my_var = "     text!!!    "
my_var = my_var.strip().rstrip("!")
my_var
```

Method chaining works by applying the method to the data from left to right. This means that in this example `.strip()` is applied to `my_var` first, and then `.rstrip()` is applied to the result of `my_var.strip()` second. If we applied the methods in reverse order like this...

```python
my_var = "     text!!!    "
my_var = my_var.rstrip("!").strip()
my_var
```

We get a different result since here `.rstrip()` was applied first.

When chaining lots of methods together, you can use the `\` character at the end of a line to help format your code like this...

```python
my_var = "     text!!!    "
my_var = my_var.rstrip("!") \
               .strip()
my_var
```

## Exercise 5 - Boolean return values

### 🚨 New Concept: Booleans

Say you wanted to know if you can cast a string to an integer. You want to avoid getting an error, so you would like to know first if your string contains only numeric data. Conveniently, the string type provides a method `.isdigit()` that allows us to check whether or not our string contains exclusively numeric values.

```python
x = "1234"

x.isdigit()
```

As you may have expected, the `.isdigit()` results in `True`. However, this `True` is special. While it looks like a string in the output box, it is actually another Python data type: a boolean.

```python
type(x.isdigit())
```
