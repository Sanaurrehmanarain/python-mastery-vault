# While loop
control flow structure
- Allows for repetitive execution of code **while** a control condition is true

```python
i = 1   # count variable is a `iterator`

while i <= 5:   # running something in loop is called `iteration`
    print("Hello!")
    i += 1

print(i)
```

```python
i = 1   

while i <= 5:  
    print("Hello!", i) # we can also print count/sr.num
    i += 1

print("Loop Ended")
```

## Break & Continue
**Break**: Used to termintate the loop when encountered.

**Continue**: Terminates execution in the current iteration & continues execution of the loop with the next iteration.

```python
i = 1

while i <= 5:
    print(i)
    if i == 3:
        break
    i += 1
```

```python
i = 1

while i <= 5:
    if i == 3:
        i += 1  # Increment before continue to avoid an infinite loop
        continue
    print(i)
    i += 1
```

```python
# Printing odd numbers.
i = 1

while i <= 10:
    if i %2 == 0:
        i += 1 
        continue
    print(i)
    i += 1
```

```python
# Printing even numbers.
i = 1

while i <= 10:
    if i %2 != 0:
        i += 1 
        continue
    print(i)
    i += 1
```

```python
# Method 1
while True:
    n = int(input("What's n?"))
    if n < 0: # This loop keeps asking for input until the user enters a non-negative (n >= 0) number.
        continue # f n < 0, it uses continue to restart the loop.
    else:
        break # If n >= 0, it breaks the loop.

# more explanation is mentioned below:
```

The while True creates an infinite loop that keeps running until a break statement is encountered.

Explanation:
- while True: starts a loop that will run indefinitely.
- Inside the loop, the program asks the user for input (n).
- If n < 0, continue restarts the loop, prompting the user again.
- If n >= 0, break exits the loop, allowing the program to continue execution.
- Key Role of while True:
- It ensures the user keeps entering input until a valid (n >= 0) number is provided. Without it, the program would only take one input and exit, even if the input was negative.

```python
# Method 2
while True:
    n = int(input("What's n?"))
    if n > 0:
        break  # No need for an else clause and continue
```

```python
should_iterate = True

while should_iterate == True:
    try:
        price = float(input("Enter a price (-1 to quit): "))
        if price == -1:
            should_iterate = False
        else:
            print(f"Discounted price: ${round(price * 0.7, 2)}")
    except ValueError:
        print("Invalid input. Please enter a numeric value or '-1' to quit.")

print("Done discounting!")
```

| **Aspect**                 | **For Loops**                                                                 | **While Loops**                                                      |
|----------------------------|------------------------------------------------------------------------------|----------------------------------------------------------------------|
| **Iterations Known**       | Number of iterations must be known before the loop starts.                   | Number of iterations can be unknown and may continue indefinitely.  |
| **Use Case**               | Ideal for iterating over a sequence or range of values.                     | Useful when the loop's termination depends on a condition.          |
| **Type of Loop**           | Definite loop (executes a set number of times).                              | Indefinite loop (executes as long as the condition remains true).    |
| **Common Use**             | Often used for iterating through lists, arrays, or ranges.                  | Frequently used for user input or tasks where conditions evolve.    |
| **Syntax Simplicity**      | Straightforward syntax for iterating a fixed number of times.               | Requires careful condition setup to avoid infinite loops.           |
| **Risk of Infinite Loop**  | Minimal, as the number of iterations is predefined.                         | Higher, if the condition is improperly defined or never changes.    |
| **Flexibility**            | Less flexible for dynamically changing conditions during execution.          | Highly flexible for handling dynamic or evolving conditions.        |
| **Examples**               | `for i in range(5): print(i)`                                                | `while user_input != "exit": user_input = input("Enter command: ")` |

```python
ounces_remaining = 144

while ounces_remaining > 0:
    user_input = float(input("How many ounces of water you want to drink? (-1 to quit): "))
    if user_input == -1:
        print(f"Your ended your hike with {ounces_remaining} ounces of water remaining.")
        break
    elif ounces_remaining > user_input:
        ounces_remaining -= user_input
        print(f"Drinking {user_input} ounces leave you with {ounces_remaining} ounces of water.")
    else:
        print(f"You drank {ounces_remaining} ounces and have no more water remaining.")
        ounces_remaining = 0
```
