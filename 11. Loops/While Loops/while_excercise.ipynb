##### Exercise 1 - Basic while loops
As you wander in the outskirts of a nearby city, you discover an abandoned apple orchard near the market. This presents a perfect opportunity to start a fruit retail business without raising suspicion about your sudden appearance in town. To help you manage your inventory when you will be selling apples at the market, you decide to create an inventory management program.

The program must ask the user to input the initial number of apples they brought to the market. Then, within a while loop, the program should prompt the user to enter the number of apples sold in a transaction and report the number of apples remaining in inventory.

```python
# Step 1: Request the initial number of apples in inventory
apples_remaining = int(input("Enter the number of apples you brought to the market: "))
apples_sold = 0

# Step 2: While there are apples remaining, request the number of apples sold in a transaction
while apples_remaining > 0:
    
    # Step 3: Update and display the number of apples sold and the number of apples remaining
    transaction = int(input("Enter the number of apples sold in this transaction: "))
    apples_sold += transaction
    apples_remaining -= transaction
    print(f"Apples sold: {apples_sold} | Apples remaining {apples_remaining}")

print("All sold out!")
```

The program starts by prompting the user to input the initial apple stock and converts it to an integer that is then stored in the variable apples_remaining. Before entering the while loop, apples_sold is initialized to 0. The loop runs while apples_remaining > 0, each iteration prompting for apples sold in a transaction and updating both apples_sold and apples_remaining accordingly. When there are no remaining apples, the program exits the loop and prints "All sold out!".

##### Exercise 2 - Multiple while conditions
##### 🚨 New Concept: Multiple conditions
In Python, the control condition of a while loop can incorporate multiple boolean expressions, which can be combined using logical operators such as and and or. This capability allows for flexible control conditions that depend on multiple criteria.

```python
battery_level = 20  # Initial battery level in percentage
temperature = 25    # Initial temperature in degrees Celsius

while battery_level < 100 and temperature < 40:
    battery_level += 5  # Charge the battery
    temperature += 0.5  # Increase temperature slightly during charging
    print(f"Battery level: {battery_level}%, Temperature: {temperature}°C")

print("Charging stopped.")
```

In this example, the while loop iterates as long as both the battery level is not fully charged and the temperature is below 40 degrees. That is, the program exits as soon as either condition becomes False.

After selling all your apples, you use the proceeds to buy wheat and barley. You need at least 20 units of each for dinner tonight—you're very hungry! Given that no single merchant carries such large quantities, you'll need to visit multiple vendors, making it challenging to mentally track your purchases. To assist with this, you decide to create a program that allows you to input the amount of wheat and barley acquired during each transaction. The program will utilize a while loop to repeatedly prompt for the amount of wheat and barley acquired after each transaction. This loop will continue until you have accumulated at least 20 units of both grains, at which point it will exit, signaling that you've reached your goal.

```python
# Step 1: Initialize the wheat and barley tracker variables
wheat = 0
barley = 0

# Step 2: While the quantity of wheat or barley is less than 20 for either grain, keep prompting the user to enter additional transactions
while wheat < 20 or barley < 20: # ensure loop whill continue as long as either the wheat or the barley variable are below 20.
    wheat += int(input("Enter units of wheat gained in this transaction: "))
    barley += int(input("Enter units of barley gained in this transaction: "))
    print(f"You currently have {wheat} units of wheat and {barley} units of barley.\n")

print(f"You have enough wheat and barley!")
```

The program starts by initializing two variables, wheat and barley, both set to 0. These variables will keep track of the total amount of each grain acquired.

A while loop is then implemented with the condition wheat < 20 or barley < 20. This means the loop will continue as long as either the wheat quantity is less than 20 or the barley quantity is less than 20.

Inside the loop:

- The user is prompted to enter the amount of wheat gained in the current transaction. This input is converted to an integer and added to the wheat total.
- Similarly, the user is asked for the amount of barley gained, which is added to the barley total.
- After each input, the program prints out the current totals of wheat and barley.

This process repeats until both wheat and barley quantities are at least 20, at which point the program exits the loop.

Finally, the program confirms that sufficient quantities of both grains have been acquired by printing a message.

##### Exercise 3 - Combining control flow mechanisms
##### Part A
The prowess of your inventory management systems have made you very successful at the local market. This, in turn, has attracted praise from fellow merchants who have just promoted you to the position of Supreme Leader of the Merchants' Guild!

Your first task as leader is to recruit three apprentices from a large pool of candidates. To be deemed worthy, a candidate must successfully roll a cheese wheel down a hill and catch it at the bottom without it breaking — twice. To help you perform your evaluation duty, you create a recruitment program.

This program must record the attempts of each candidate and notify the user when three suitable candidates have been identified.

```python
# Step 1: Initialize the number of available vacancies and set the while loop header.
member_needed = 3
while member_needed > 0:
    
    # Step 2: For each recruit, initialize the counter of trial passed to zero, and iterate through the two trials.
    trials_passed = 0
    for i in range (1,3):

        # Step 3: Prompt the user regarding the trial success of the candidate and update the trials_passed counter accordingly.
        trial = input(f"Did the recruit pass trial {i}? (yes/no): ")
        if trial == "yes":
            trials_passed += 1

    # Step 4: Test if the recruit passed both trials; if yes, decrement the number of members needed, and move to next iteration.
    if trials_passed == 2:
        member_needed -= 1
        print(f"New member recruited! {member_needed} mumber spots remaining.")
    else:

        # Step 5: If the recruit did not pass, notify the user and move to the next iteration.
        print(f"Recruit did not pass all trials. {member_needed} membership spots remaining.")

# Step 6: Once the required number of members has been recruited, end the loop and announce the completion of the recruitment process.
print("Guild recruitment complete.")
```

In this solution, we use a while loop to continuously search for candidates until three of them have been successfully recruited. The control condition of the while loop, members_needed > 0, ensures that candidate testing continues as long as their are unfilled apprentice positions.

In the body of the while loop, we first initialize the number of trials the candidate has passed to zero.

Then, we use a for loop to iterate over each candidate's attempts. In this loop, the user is prompted to input the result of the trial. The trial_passed variable is updated according to the user input.

After the for loop completes, we check if the candidate has passed both trials. If the candidate did pass, we decrement the number of members needed. If the candidate failed, we move to testing the next candidate without decrementing the number of members needed.

##### Part B
Your current experience with the crop of guild candidates has been disappointing. Only one candidate passed your rigorous cheese-rolling test. Kids these days...

Recognizing the need to adapt, you've decided to adjust your recruitment strategy by allowing candidates five opportunities to obtain two successful attempts.

Adapt your code from Part A accomodate this more lenient policy.

Hint: you may want to consider using a nested while loop instead of a nested for loop.

```python
# Step 1: Initialize the number of vacancies and set the while loop header.
members_needed = 2

while members_needed > 0:
    # Step 2: For each recruit, initialize counters for passed and attempted trials.
    trials_passed = 0
    trials_attempted = 0

    # Step 3: Allow the recruit to attempt trials until they pass two trials or attempt five trials.
    while trials_passed < 2 and trials_attempted < 5:
        trial = input(f"Did the recruit pass trial {trials_attempted + 1}? (yes/no): ").strip().lower()
        if trial == "yes":
            trials_passed += 1
        elif trial != "no":  # Validate input
            print("Invalid input. Please enter 'yes' or 'no'.")
            continue  # Skip incrementing trials_attempted for invalid input
        trials_attempted += 1

    # Step 4: Test if the recruit passed the required trials.
    if trials_passed == 2:
        members_needed -= 1
        print(f"New member recruited! {members_needed} member spot{'s' if members_needed != 1 else ''} remaining.")
    else:
        # Step 5: If the recruit did not pass, notify the user.
        print(f"Recruit did not pass enough trials. {members_needed} member spot{'s' if members_needed != 1 else ''} remaining.")

# Step 6: Announce the completion of the recruitment process.
print("Guild recruitment complete.")
```

With respect to Part A we made the following modifications:

- We created a new counter for the number of trials attempted. We use this counter in the control condition of the nested loop.
- We replaced the inner for loop with a while loop. The control condition tests both the number of successful and attempted trials.

##### Exercise 4 - Managing infinite while loops
##### 🚨 New Concept: Infinite while loops and break keyword
An infinite loop occurs when a while loop's control condition perpetually evaluates to True, causing the loop to run indefinitely. This is often an unintended consequence of improperly structured loop conditions or failure to update critical variables within the loop body.

When a program is stuck in an infinite loop in Google Colab, you can stop the execution by:

- Clicking the cell's "stop" button
- Holding the Ctrl key and pressing M followed by I
- Clicking the "Runtime" menu item and selecting the "Restart session" option

For example, the code that follows will run indefinitely. To stop the execution, follow one of the three methods above.

```python
while True:
    print("This will run forever.")
```

Clearly, the loop's control condition in the previous cell is always True, resulting in an infinite loop, which is problematic.

However, in some situations, pairing an infinite loop with a break statement is a convenient way to execute a task.

For example, the code below will stop running when the user enters the word "exit":

```python
while True:
    user_input = input("Enter 'exit' to quit: ")
    if user_input == 'exit':
        break
    print("You entered:", user_input)

print("Program finished.")
```

This while loop could theoretically run forever. But if the user inputs the exit value "exit", the program breaks out of the loop.

##### Part B
After selling a few jokes, you realize there are limits to your creativity. Therefore, you need to add a mechanism that allows you to end the program when you feel you've run out of ideas.

Modify your code from Part A to allow the user to input an exit value. If this exit value is entered, the loop must terminate, allowing the program to exit gracefully.

```python
jokes_sold = 0

while True:
    transaction = int(input("Enter the number of jokes the customer bought (-1 to stop selling jokes): "))
    if transaction == -1:
        break
    jokes_sold += transaction
    print(f"Total jokes sold: {jokes_sold}")
```

This code improves on Part A by adding a way for the user to exit the infinite while loop (entering the value -1). The user is still prompted to enter the number of jokes sold, with no fixed limit on the number of times they can do so.

##### Part A
As you are selling jokes, one of your potential customers confesses that they do not have the funds to purchase your famous jokes. So, he proposes to barter: five of your finest jokes in exchange for five rare battle items: a pair of boots, an helmet, an armour, a shield, and a sword. You accept the trade since you know you'll be able to sell these items at a good price later.

To maintain an up to date inventory of your battle items as you sell them, create a program that takes the inventory list below as given and:

1) Prompts the user to enter the name of the item the customer wishes to purchase 
2) If the item the customer wishes to purchase is in the inventory, remove it. Otherwise, alert the user that the item is not in the inventory.

```python
inventory = ["boots", "helmet", "armour", "shield", "sword"]
print(f"Initial inventory: {inventory}\n")

# Step 1: Initiate a while loop that iterates as long as there are items in the inventory
while inventory:
    # Step 2: Prompt the user for the item sold
    item = input(f"Enter an item: ")

    # Step 3: Determine if item is in inventory
    if item in inventory:
        # Step 4: If so, remove it and output the remaining inventory
        inventory.remove(item)
        print(f"Sold {item}.")
    else:
        # Step 5: Otherwise, alert the user the item is not in the inventory
        print(f"{item.capitalize()} not in inventory.")

    # Step 6: Output remaining inventory items
    print(f"Remaining battle items in inventory: {inventory}\n")

print("All items sold!")
```

In this program, we use a while loop to prompt the user to enter items as long as the inventory list has items. Note that the loop condition we used, while inventory, is equivalent to while len(inventory) > 0.

Inside the loop, we prompt the user to input an item. Since the user can enter any item, we must test if the item is in the inventory. If it is, we use the .remove() list method to remove the item from the list. If it is not, we alert the user that the item is not in the inventory. We then output the remaining items in inventory.

##### Part B
Now that you have become a well respected and successful merchant, you've begun wondering about your retirement prospects. The life of a merchant is tough, and you know you won't be able to do it forever. To ensure your ability to retire at a reasonable age, you target making 1,000 coins in daily profit at the market. To help you track your intra-day progress toward that goal, you want to create a profit tracking program. While the current daily profit is less than 1,000 coins, the program should:

1. Prompt the merchant (user) for the number of goods sold in the current transaction.
2. Prompt the merchant (user) for the average price of the goods sold.
3. Update the current profit.
4. Print the remaining profit needed or alert the user upon reaching the 1000-coin goal.

Solution

We propose two possible solutions:

The first solution uses a control condition that tests whether the current_profit is less than the target_profit. If the profit target has not been reached, the loop executes another iteration:

```python
target_profit = 1000
current_profit = 0

# Your code here
while current_profit < target_profit:
    goods_sold = int(input("Enter number of goods sold: "))
    price_per_good = int(input("Enter the average price per good: "))
    current_profit += goods_sold * price_per_good
    if current_profit < target_profit:
        print(f"Total profit so today: {current_profit} | Profit remaining: {target_profit - current_profit}")
    else:
        print("Target profit reached!")
```

Inside the body of the loop, we prompt the user to input the number of goods sold and their average price, then increment the profit accordingly. We also check whether the updated profit is still less than the target profit in order to print out the correct message.

Alternatively, we could use the infinite loop with a break pattern:

```python
target_profit = 1000
current_profit = 0

while True:
    goods_sold = int(input("Enter number of goods sold: "))
    price_per_good = int(input("Enter the average price per good: "))
    current_profit += goods_sold * price_per_good
    if current_profit < target_profit:
        print(f"Total profit today: {current_profit} | Profit remaining: {target_profit - current_profit}")
    else:
        print("Target profit reached!")
        break
```

With this approach, we need to include the break keyword in the else statement. This ensures the while loop will end once the target profit is reached.
