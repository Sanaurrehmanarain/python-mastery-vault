# The enumerate() function

The enumerate() function is a useful tool for tracking the index of the current item during a for loop. When we use enumerate() on a list, Python returns the index of the current item along with the item itself at the start of each iteration. Therefore, using enumerate() requires two variables: one for the index and one for the item.

Using the enumerate() function provides a convenient way to keep track of the current index of the list we are iterating over.

```python
tasks = ["Wash the dishes", "Do the laundry", "Write the report"]
for index, task in enumerate(tasks):
    print(f"Task {index}: {task}")
```

Using the enumerate() function provides a convenient way to keep track of the current index of the list we are iterating over.

# The _ placeholder

When iterating through a collection of collections, it is often convenient to unpack the nested collections in the header of the loop:

```python
cities = [
    ("San Francisco", "California", "USA"),
    ("New York", "New York", "USA"),
    ("Chicago", "Illinois", "USA"),
    ("Los Angeles", "California", "USA"),
    ("Miami", "Florida", "USA"),
    ("Houston", "Texas", "USA"),
    ("Phoenix", "Arizona", "USA"),
    ("Philadelphia", "Pennsylvania", "USA"),
    ("Seattle", "Washington", "USA")
]

for city, state, country in cities:
    print(f"{city} is in {state}.")
```

However, sometimes part of the unpacked information isn't needed in the rest of the loop. This is the case with the country variable in the example above. Instead of assigning data that isn't needed to a variable, we can use the _ character as a placeholder.

Using the _ character as a placeholder makes it clear that the corresponding information isn't necessary for the loop's execution.

```python
for city, state, _ in cities:
    print(f"{city} is in {state}.")
```

# The range() function
##### Excercise 4

The range() function is a valuable tool for iterating over a sequence of numbers. For instance, it can be used in a for loop to repeat a block of code a given number of times:

```python
for num in range(3):
    print(num)
```

Note that, in the example above, the iteration variable is the value 0 during the first iteration and is incremented by 1 each iteration until it reaches the exclusive upper limit set by the range() function.

Alternatively, the range() function can accept two arguments: the first sets the initial value for the iteration variable, and the second defines the exclusive upper limit.

```python
for num in range(2, 5):
    print(num)
```

Finally, the range() function can also be called with three arguments.

In this scenario, the first two arguments define the bounds for the iteration variable, while the third argument specifies how much the iteration variable should be incremented after each loop.

In summary, the range() function is an effective tool for easily creating sequences of numbers for iteration.

```python
for num in range(1, 6, 2):
    print(num)
```

#  Loop control mechanisms
##### Excercise 6

Python offers two tools to manage the execution flow of a loop loops: the break and continue keywords.

The **break** Keyword

The break keyword is used to exit a loop prematurely. When Python encounters a break statement inside a loop, it halts the loop's execution and resumes execution at the next line of code following the loop. The break keyword is commonly used to exit a loop when a specific condition is met.

For example, the following code uses the break keyword to search for a particular number in a list and stop once it is found:

```python
numbers = [8, 2, 10, 4, 6, 7, 9, 3, 1, 5]
target = 7

for number in numbers:
    if number == target:
        print(f"Found the target: {number}")
        break
    else:
        print(f"Current number {number} - didn't find target yet")
```

# The **continue** Keyword

The continue keyword is used to skip the remainder of the current iteration of a loop and move directly to the next iteration. When Python encounters a continue statement inside a loop, it jumps back to the loop's header and starts the next iteration, bypassing the rest of the code in the current iteration.

The continue keyword is useful for:

- Skipping certain iterations based on a condition
- Avoiding the execution of specific parts of the loop's body when certain criteria are met

For example, the following code uses the continue keyword to print only the even numbers in a list:

In summary, the break and continue keywords offer powerful tools for controlling the execution flow of loops, allowing you to make your code more flexible.

```python
numbers = [8, 2, 10, 4, 6, 7, 9, 3, 1, 5]

for number in numbers:
    if number % 2 != 0:
        continue
    print(number)
```

##### Part B
Since your boss is keen on productivity, he has set a strict 32-minute limit for the party. To signal the end of the party, you should stop playing music once the total duration reaches this limit. However, if the song currently playing would cause the total time to exceed 32 minutes, you are allowed to let it finish — your boss isn’t a monster.

Building on the code from Part A, modify it to print each song along with the cumulative total duration of music played so far. Once the total duration exceeds 32 minutes, stop printing songs. Use the break keyword to achieve this.

```python
party_playlist = [
    ("Bohemian Rhapsody", 6),
    ("Stairway to Heaven", 8),
    ("Hotel California", 6.5),
    ("California Dreamin'", 2.5),
    ("Smells Like Teen Spirit", 5),
    ("Going to California", 3.5),
    ("Yellow Submarine", 3),
    ("California Gurls", 3.5),
    ("Footloose", 4),
    ("Take On Me", 3.5),
    ("Livin' on a Prayer", 4),
    ("Don't Stop Believin'", 4.5),
    ("Eye of the Tiger", 4),
    ("Another One Bites the Dust", 3.5),
    ("I Want to Break Free", 4)
]

# Your code here
total_duration = 0

for song, duration in party_playlist:
    if total_duration > 32:
        break

    if "California" in song:
        continue

    total_duration += duration
    print(f"{song} playing - {total_duration} minutes of music played")
```

To ensure that we stop playing songs when the 32-minute mark is reached (while still allowing the final song to finish), we created a variable total_duration to track the cumulative duration of songs played. ##### Part A

To create a report for the data analytics team, you need a list of songs with titles longer than 15 characters. Use a for loop to create this list from the playlist provided.

```python
playlist = [
    "Bohemian Rhapsody",
    "Stairway to Heaven",
    "Hotel California",
    "Imagine",
    "Smells Like Teen Spirit",
    "Sweet Child O' Mine",
    "Like a Rolling Stone",
    "Comfortably Numb",
    "Space Oddity",
    "Lucy in the Sky with Diamonds"
]

# Your code here
long_title_songs = []

for song in playlist:
    if len(song) > 15:
        long_title_songs.append(song)

print(long_title_songs) 
```

Using a for loop, we can iterate through each song in the playlist and check if its title has more than 15 characters. If it does, we append it to the list.

While this approach works, Python offers a more concise syntax to achieve the same result: list comprehensions.

##### List comprehensions
List comprehensions offer a concise way to transform and filter lists in Python. To understand the syntax of list comprehensions, it’s useful to compare it with how a for loop can accomplish the same task. In this example, we’ll convert a list of temperatures from Celsius to Fahrenheit.

Using a For Loop

We can convert Celsius to Fahrenheit using a for loop:

```python
celsius = [0, 10, 20, 30, 40]
fahrenheit = []

for temp in celsius:
    fahrenheit.append((temp * 9/5) + 32)

print(fahrenheit)
```

This approach works, but is somewhat cumbersome.

Introducing List Comprehensions

List comprehensions enable you to perform operations on each element of a list in a single line of code. The basic syntax is: [expression for item in collection]

- **expression**: The operation to apply to each item in the collection.
- **item**: The loop variable representing an item in the collection.
- **collection**: The collection you're iterating through.

Using a list comprehension, the previous example can be rewritten more succinctly:

```python
celsius = [-10, 0, 10, 20, 30, 40]
fahrenheit = [(temp * 9/5) + 32 for temp in celsius]
print(fahrenheit)
```

Adding Conditions

We can also filter out elements from the original list using an if statement. With an if statement, only the elements of the original list that satisfy the condition are included in the resulting list: [expression for item in collection if condition]

For example, we can create a list of Fahrenheit temperatures for Celsius temperatures above the freezing point as follows:


```python
above_freezing = [(temp * 9/5) + 32 for temp in celsius if temp > 0]
print(above_freezing)
```

Now, only the temperatures that are greater than 32 degrees in the farienheit list have been added to the above_freezing list.

Benefits of List Comprehensions

Conciseness: They enable you to create lists in a single, compact line of code.
Readability: They enhance the readability of your code, especially for straightforward list transformations and filters.
List comprehensions are a powerful feature in Python that can make your code both more concise and easier to understand.

##### Part B
Use a list comprehension to generate a new list containing only the songs with titles longer than 15 characters.

```python
playlist = [
    "Bohemian Rhapsody",
    "Stairway to Heaven",
    "Hotel California",
    "Imagine",
    "Smells Like Teen Spirit",
    "Sweet Child O' Mine",
    "Like a Rolling Stone",
    "Comfortably Numb",
    "Space Oddity",
    "Lucy in the Sky with Diamonds"
]

# Your code here
long_songs = [song for song in playlist if len(song) > 15]

print(long_songs)
```

To output the songs with titles longer than 15 characters using a list comprehension, we start by typing []. Inside the brackets, we specify the expression as song, since we don't need to modify the items from the original list. Next, we define the loop variable, which is also song, and the iterable, which is playlist. Finally, we add the condition for including an item in our filtered list: len(song) > 15.
