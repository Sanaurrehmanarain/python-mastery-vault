```python
veggies = ["Potato", "Brinjal", "Ladyfinger", "Tomato", "Cucumber", "Onion"]

for _ in veggies:
    print(_)
```

```python
countries = ["China", "Portugal", "Spain", "Germany", "UK", "U.S.A", "Pakistan", "India", "Sweden", "Norway"]

for i in range(len(countries)):
    print(i, countries[i])
```

**How It Works**
1. `len(countries)` → finds how many items are in the list.

   * Here: `len(countries)` = **10**

2. `range(len(countries))` → generates numbers from **0** to **9**.

3. For each `i` in that range:

   * `countries[i]` accesses the country at index i.

   * Prints `i` (index) and `countries[i]` (country name).

**Visual**

| `i` | `countries[i]` |
| --- | -------------- |
| 0   | China          |
| 1   | Portugal       |
| 2   | Spain          |
| 3   | Germany        |
| 4   | UK             |
| 5   | U.S.A          |
| 6   | Pakistan       |
| 7   | India          |
| 8   | Sweden         |
| 9   | Norway         |

```python
for i, j in enumerate(countries):
    print(i)
    print(j)
```

**How It Works**
1. `enumerate(countries)` returns **pairs** → `(index, value)`.

2. For each `(i, j)`:

   - `i` → the index of the country.

   - `j` → the country name.

3. Prints the index (`i`) **on one line** and the country (`j`) **on the next**.

**Visual**
| `i` | `j` (country) |
| --- | ------------- |
| 0   | China         |
| 1   | Portugal      |
| 2   | Spain         |
| 3   | Germany       |
| 4   | UK            |
| 5   | U.S.A         |
| 6   | Pakistan      |
| 7   | India         |
| 8   | Sweden        |
| 9   | Norway        |

**Key Difference:**

- `range(len(...))` → You manually get indexes, then access the value.

- `enumerate(...)` → Python automatically gives you both index and value.

## Range
Range functions returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

```python
for _ in range(3): # starts from 0, ends at 2, range(stop)
    print(_)
```

```python
for _ in range(1, 5): # starts from 1, end at 4, range(start, stop)
    print(_)
```

```python
for _ in range(2, 10, 2): # step size 2, range(start, stop, step)
    print(_)
```

```python
# printing even numbers
for i in range(2, 10, 2):
    print(i)
```

```python
# printing odd numbers
for i in range (1, 10, 2):
    print(i)
```

##### WAP to find the factorial of first n natural numbers. 

```python
# 1 * 2 * 3 * 4 * 5
n = 5

fact = 1
for i in range(1, n + 1):
    fact *= i

print("Factorial = ", fact)

# If we used range(1, n), it would generate:
# 1, 2, 3, 4  (excluding `5`, which is incorrect for factorial)
# Thus, n + 1 ensures the loop runs from 1 to n, covering all necessary numbers for factorial calculation.
```

##### WAP to find the sum of first n natural numbers.

```python
# 1 + 2 + 3 + 4 + 5
n = 5

sum = 0
for i in range(1, n + 1):
    sum += i

print("Total sum = ", sum)
```

The grocery owner wants to discount all inventory items by 30%. We can do this easily by using for loops.

```python
def main():
    names = ["Mario", "Daisy", "Luigi", "Yoshi"]   
    
    # method 2, better version
    for name in names:
        print(write_letter(name, "Princess Peach"))

def write_letter(receiver, sender):
    return f"""
    +~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~+
        Dear {receiver},

        You are cordially invited to a ball at
        Peach's Castle this evening, 7:00 PM.

        Sincerely,
        {sender}
    +~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~+
    """
main()
```

```python
prices = [
    51.79, 39.37, 17.30, 38.42, 12.41, 19.38, 52.23, 43.92, 11.88, 7.49,
    46.02, 79.85, 50.51, 97.94, 38.57, 76.62, 53.61, 45.44, 75.90, 21.01, 
    6.66, 71.08, 86.49, 70.11, 98.59, 9.52, 72.78, 91.21, 32.01, 28.35
]
discounted_prices = []   # make a empty list to store discounted prices

for price in prices: # take each item from the list prices one by one and put it in the Jar(iteration / loop_variable price). iterate through each item until the list is finished.
    discounted_prices.append(round(price * .7, 2))  # multiply them with .7 and append it to the list named discounted_prices.

print(discounted_prices)
```

Python reaches the loop's header for the first time.

It creates the iteration variable price and assigns it the first item of the prices list.

Then Python executes the code in the body of the loop.

In this case, the body of the loop contains a single line of code, but there could be more.

Since the iteration variable price takes the value 51.79 during the first iteration, the value 36.25

is appended to the discounted prices list following the execution of the code in the body of the loop.

Python returns to the header.

At that point, the iteration variable is reassigned.

The second item in the prices list.

Then Python executes the code in the body of the loop again.

Since the iteration variable takes the value 39.37 during the second iteration, the value 27.56 is

appended to the discounted price list.

Python then returns to the header reassigns the iteration variable and executes the code in the body

of the loop once again.

This process continues until the iteration variable has been assigned every price in the prices list.

When Python runs out of item, the loop execution is complete.

```python
waypoints = [
    "Granite Falls Crossing",
    "Starlight Meadow",
    "Cedar Creek Junction",
    "Evergreen Grove",
    "Pine Ridge Summit"
]

tasks = [
    "Drink water",
    "Eat a snack",
    "Inventory supplies",
    "Take pictures"
]
```

```python
for waypoint in waypoints:
    print(f"Arrived at {waypoint}, completing tasks:")
    print(f"\t{tasks[0]}")
    print(f"\t{tasks[1]}")
    print(f"\t{tasks[2]}")
    print(f"\t{tasks[3]}")
    print(f"Leaving {waypoint}.\n")
```

### we can make a better version by creating a nested loop. 

```python
for waypoint in waypoints:
    print(f"Arrived at {waypoint}, completing tasks:")
    for task in tasks:
        print(f"\t{task}")
    print(f"Leaving {waypoint}.\n")
```
