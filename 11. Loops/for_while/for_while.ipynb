# BREAK & CONTINUE

##### **Break** in for loop

```python
for x in a:
	if x == 3:
		break
	print(x) # output: 0  1  2
```

##### **Break** in while loop

```python
i = 0

while i < 5:
	if i == 3:
		break
	print(i) # output: 0  1  2 
	i += 1
```

##### **Continue** in for loop

```python
for x in a:
	if x == 2:
		continue
	print(x) # output: 0  1  3  4  5
```

##### **Continue** in while loop

```python
i = -1

while i < 5:
	i += 1
	if i == 2:
		continue
	print(i) # output: 0  1  3  4  5
```

```python
def main():
    number = get_number()
    meow(number)
    
def get_number():
    while True:
        n = int(input("What's n?: "))
        if n > 0:
            return n
        
def meow(n):
    for _ in range(n):
        print("meow")
        
main()
```

# Nested Loops Example Code

```python
# Representing the spreadsheet as a nested list
students_scores = [
    ["Alice", 85, 88, 95],
    ["Bob", 70, 90, 87],
    ["Charlie", 92, 86, 90],
    ["Diana", 78, 82, 85],
    ["Ethan", 89, 95, 84]
]

# Using a nested loop to calculate total scores
for student in students_scores:
    name = student[0]
    total = 0
    for score in student[1:]:  # Skip the name, sum only scores
        total += score
    print(f"{name}: {total}")
```

# Parallelization with Joblib example code

```python
# Without Parallelization
import time

my_list = [1,2,3,4]
my_squares = []

for i in my_list:
    time.sleep(1)
    squared = i * i
    my_squares.append(squared)

print(my_squares)
```

```python
# With Parallelization
from joblib import Parallel, delayed

my_list = [1,2,3,4]
my_squares = []

def slow_square(i):
    time.sleep(1)
    squared = i * i
    return squared

my_squares = Parallel(n_jobs=-1)(delayed(slow_square)(i) for i in my_list)

print(my_squares)
```
