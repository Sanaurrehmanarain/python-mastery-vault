##### make the dictionary using for loop.
1. Make an empty dictionary and Create variables
2. Create outer layer of keys
3. Create inner layer of keys

```python
trail_log = {}
mile_markers = ["Mile 3", "Mile 6", "Mile 9"]
questions = [
    "How are you feeling?",
    "Did you see any cool wildlife?",
    "How are the trail conditions?"
]

for mile in mile_markers:
    trail_log[mile] = {}
    for  question in questions:
        trail_log[mile][question] = ""
    
trail_log
```

```python
trail_log ["Mile 3"]["How are you feeling?"] = "Energetic!"
trail_log ["Mile 3"]["Did you see any cool wildlife?"] = "Two elk and a mosse."
trail_log ["Mile 3"]["How are the trail conditions?"] = "Rocky and steep"

trail_log

# writing each answer seperately is cumbersome.
```

```python
# To update multiple key/values items at the same time. We use .update method
trail_log["Mile 6"].update({"How are you feeling?": "A little tired and hungry.",
                            "Did you see any cool wildlife?": "A Bear",
                            "How are the trail conditions?": "Rooty and overgrown."})

trail_log
```

```python
trail_log["Mile 9"].update({questions[0]: "Tired, but accomplished.",
                            questions[1]: "2 Bison",
                            questions[2]: "Well maintained."})

trail_log
# Instead of typing manually the questions, we can give the 'questions' list with the indexing. It will prevent any spelling mistake.
```
