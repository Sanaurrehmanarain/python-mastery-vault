# Dictionary Comprehensions
## What is a Dict Comprehension?

- Similar to list/set comprehensions, but builds a dictionary.

- Uses curly braces {} with a key: value expression.

- Syntax:

```python
{key_expr: value_expr for item in iterable if condition}
```

✅ Example 1: Basic Dict Comprehension

````python
numbers: list[int] = [1, 2, 3, 4, 5]
squared_dict: dict[int, int] = {n: n**2 for n in numbers}

print(squared_dict)
````

✅ Example 2: With a Condition

````python
numbers: list[int] = [1, 2, 3, 4, 5, 6]
even_square_dict: dict[int, int] = {n: n**2 for n in numbers if n % 2 == 0}

print(even_square_dict)
````

✅ Example 3: Transforming Data

````python
words: list[str] = ["apple", "banana", "cherry"]
length_dict: dict[str, int] = {word: len(word) for word in words}

print(length_dict)
````

✅ Example 4: Swapping Keys and Values

````python
data: dict[str, int] = {"a": 1, "b": 2, "c": 3}
swapped: dict[int, str] = {v: k for k, v in data.items()}

print(swapped)
````

🔹 When to Use Dict Comprehension

- When building a dictionary from an iterable.

- When transforming or filtering data into key-value pairs.

- More concise and Pythonic than writing a full `for` loop with `dict.update()`.

List comprehension → builds lists.

Set comprehension → builds sets (unique values).

Generator comprehension → builds generators (lazy evaluation).

Dict comprehension → builds dictionaries (`{key: value}`).

### 🔹 Comprehensions in Python — Comparison Table
| Type                        | Syntax Example                | Result Type | Key Features                                                  |
| --------------------------- | ----------------------------- | ----------- | ------------------------------------------------------------- |
| **List Comprehension**      | `[n**2 for n in range(5)]`    | `list`      | Keeps order, allows duplicates, eagerly computes all values   |
| **Set Comprehension**       | `{n**2 for n in range(5)}`    | `set`       | Removes duplicates automatically, unordered                   |
| **Dict Comprehension**      | `{n: n**2 for n in range(5)}` | `dict`      | Builds key-value pairs, allows transformation and filtering   |
| **Generator Comprehension** | `(n**2 for n in range(5))`    | `generator` | Lazy evaluation (produces values on demand), memory efficient |

✅ Quick Examples

```python
# List comprehension
nums = [n**2 for n in range(5)]
print(nums)  # [0, 1, 4, 9, 16]

# Set comprehension
nums_set = {n**2 for n in range(5)}
print(nums_set)  # {0, 1, 4, 9, 16}

# Dict comprehension
nums_dict = {n: n**2 for n in range(5)}
print(nums_dict)  # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# Generator comprehension
nums_gen = (n**2 for n in range(5))
print(nums_gen)        # <generator object <genexpr> at 0x...>
print(list(nums_gen))  # [0, 1, 4, 9, 16]
```
