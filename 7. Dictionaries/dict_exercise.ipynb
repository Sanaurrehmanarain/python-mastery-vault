```python
works = ["Starry Night", "Mona Lisa", "The Persistence of Memory", "Girl with a Pearl Earring", "The Scream", "The Night Watch", "The Birth of Venus", "Guernica", "The Last Supper", "The Kiss"]
artists = ["Vincent van Gogh", "Leonardo da Vinci", "Salvador Dalí", "Johannes Vermeer", "Edvard Munch", "Rembrandt", "Sandro Botticelli", "Pablo Picasso", "Leonardo da Vinci", "Gustav Klimt"]
genres = ["Post-Impressionism", "Renaissance", "Surrealism", "Baroque", "Expressionism", "Baroque", "Renaissance", "Cubism", "Renaissance", "Symbolism"]


# Method 1
# Your code here
art_gallery = {} # initializes an empty dictionary to store the structured data.
for i in range (len(works)): 
# range(len(works)): Generates a sequence of indices from 0 to len(works) - 1. 
# This is because len(works) gives the total number of art pieces, and range() creates a range object from 0 to that number (excluding the last value).
# i: Represents the current index in the iteration.
    art_gallery[works[i]] = {"artist": artists[i], "genre": genres[i]}
    # works[i]: Gets the art piece name at index i.
    # artists[i]: Gets the artist's name at index i.
    # genres[i]: Gets the genre at index i.
    # art_gallery[works[i]] = {"artist": artists[i], "genre": genres[i]}:
    # Creates a new key-value pair in the art_gallery dictionary.
    # The key is the art piece name (works[i]).
    # The value is a dictionary containing the artist and genre, constructed as:
    # {"artist": artists[i], "genre": genres[i]}

# The loop repeats for every index i in the range, building up the dictionary one entry at a time.

art_gallery
```

```python
# Method 2
art_gallery = {}
for work, artist, genre in zip (works, artists, genres): # more concise and readable way
    art_gallery[work] = {"artist": artist, "genre": genre}

art_gallery
```

##### Exercise 2 - Keys, values, and items
🚨 **New Concept**: The .keys(), .values(), and .items() methods

The .keys(), .values(), and .items() methods are very useful for iterating through items in a dictionary. They work as follows:

- .keys(): Returns the keys in the dictionary.
- .values(): Returns the values in the dictionary.
- .items(): Returns the key-value pairs in the dictionary, with each pair grouped together as a tuple.

For example, calling the .keys() method on the dictionary below would produce the following output:

```python
employees = {
    'Tony Stark': 'Mechanical Engineer',
    'Steve Rogers': 'History Teacher',
    'Natasha Romanoff': 'Private Investigator',
    'Bruce Banner': 'Biochemist',
    'Thor Odinson': 'Physicist',
}

employees.keys()
```

The .keys() method (as well as the .values() and .items() methods) does not return its output as a proper list. Instead, it returns a structure that for loops can use for iteration. For the purpose of this course, this is a subtlety we will largely ignore. We will treat the return values of these methods as if they were lists. Note that you can always use type casting to explicitly convert the return values into proper collections:

```python
list(employees.keys())
```

##### Exercise 4 - Removing items from a dictionary
🚨 **New Concept**: The .pop() method

To remove an item from a dictionary, Python provides the .pop() method. This method takes the key to remove as an argument, removes the corresponding key-value pair from the dictionary, and returns the value associated with that key. For example:

```python
house_plants = {
    "Spider Plant": {
        "light": "Indirect sunlight",
        "water": "Moderate, allow soil to dry between waterings",
        "care_level": "Easy"
    },
    "Snake Plant": {
        "light": "Low to bright indirect light",
        "water": "Sparingly, every 2-6 weeks",
        "care_level": "Easy"
    },
}

removed_plant = house_plants.pop("Snake Plant")
print("Snake Plant")
print(f"Light: {removed_plant.get('light')}")
print(f"Water: {removed_plant.get('water')}")
print(f"Care level: {removed_plant.get('care_level')}")

house_plants
```

From the output, we see that the .pop() method removed the Snake Plant from the dictionary and returned its data, which we then stored in the removed_plant variable.

##### Part B
The gallery wants to summarize the number of art pieces by each artist. Use a for loop to count the number of works by each artist and store that information in a new dictionary. Then, use another for loop to print the summary.

```python
# Count the number of works for each artist
artist_summary = {}
for details in art_gallery.values():
    artist = details["artist"]
    if artist in artist_summary:
        artist_summary[artist] += 1
    else:
        artist_summary[artist] = 1

# Print the summary
for artist, count in artist_summary.items():
    print(f"{artist}: {count} works")
```
