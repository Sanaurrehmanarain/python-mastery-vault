# DICTIONARY
- Dictionaries are used to store data values in key:value pairs
- They are unordered, mutable (changeable) & don't allow duplicate keys

```python
info = {
    "name": "Sana",
    "college": "Apna College",
    "subjects": ["Python", "SQL", "Tableau", "BI","R"],
    "topics": ("data exploration", "data cleaning", "data extracting"),
    "age": 37,
    "is_adult": True,
    "profession": "Data Analyst",
    "married": "Yes",
    "children": "Yes",
    "qualification": "Masters",
    "nationality": "Pakistani",
    "city": "Braga"
}

print(type(info))
info
```

## Dictionary Methods

```python
# To access any value in the dict.
info["city"] # Raises a KeyError if the key doesn’t exist.
# Use info["key"] when you're sure the key exists.
```

```python
# Returns the value according to key.
info.get("city") # Returns None (or a default value) instead of raising an error if the key is missing.
# Use info.get("key") when the key might be missing, preventing errors.
```

```python
# To get all the keys
info.keys()
```

```python
# To get all the values
info.values()
```

```python
# To assign a new value / change the value
info["name"] = "Sana Ur Rehman"
```

```python
# To add a new key : value
info["surname"] = "Arain"
```

```python
# Return all (key, val) pairs as tuples
info.items()
```

```python
# Inserts the specified items to the dictionary.
info.update({"weight": 80})
info
```

```python
# Adding Multiple Key-Value Pairs to a Dictionary
# Method 1
# Using .update() (Recommended)
my_dict = {"a": 1, "b": 2}
my_dict.update({"c": 3, "d": 4})  # Adds multiple key-value pairs
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```

```python
# Method 2
# Using Dictionary Unpacking (**)
my_dict = {"a": 1, "b": 2}
my_dict = {**my_dict, **{"c": 3, "d": 4}}
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```

```python
# Removing Multiple Key-Value Pairs from a Dictionary
# Method 1
# Using a Loop with del
my_dict = {"a": 1, "b": 2, "c": 3, "d": 4}
for key in ["b", "d"]:
    del my_dict[key]  # Removes keys 'b' and 'd'
print(my_dict)  # Output: {'a': 1, 'c': 3}
```

```python
# Method 2
# Using Dictionary Comprehension
my_dict = {"a": 1, "b": 2, "c": 3, "d": 4}
my_dict = {k: v for k, v in my_dict.items() if k not in ["b", "d"]}
print(my_dict)  # Output: {'a': 1, 'c': 3}
```

WAP to enter marks of 3 subjects from the user and store them in a dictionary. Start with an empty dictionary & add one by one. Use subject name as key & marks as value.

```python
# Method 1
marks = {}

x = int(input("enter physics score: "))
marks.update({"physics": x})

x = int(input("enter maths score: "))
marks.update({"maths": x})

x = int(input("enter chemistry score: "))
marks.update({"chemistry": x})

print(marks)
```

```python
# Method 2
# Initialize an empty dictionary
marks = {} # Create an empty dictionary

# List of subjects
subjects = ["Physics", "Maths", "Chemistry"] # Store subject names in a list

# Loop to take input for each subject
for subject in subjects: # Use a loop to iterate through the subject list and take user input.
    marks[subject] = int(input(f"Enter marks for {subject}: ")) # Store the marks in the dictionary using subject names as keys.

# Print the dictionary
print(marks)
```
