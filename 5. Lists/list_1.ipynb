# LISTS ARE MUTABLE

```python
students = ["Ron", "Hermione", "Ginny", "Neville", "Draco", "Bellatrix", "Aurora"]
print(students[0 : 3]) # list slicing
print(students[-7 : -4])
```

```python
for student in students:    # to call each element of the list, we didn't use _ here becuase we are using the variable name "student" again.
    print(student)
```

```python
for i in range(len(students)):
    print(i + 1, students[i]) # This will give the index num
```

Strings are immutable but Lists are mutable.

```python
# Example strings
str = "hello"
print(str[0]) # you can access to this index. its allowed.
str[0] = "y" # This is not allowed. You can not add elements to strings.
```

```python
list = ["karan", 95.4, 17, "Delhi"]
print(list[0])   # you can access the element using index
list[0] = "arjun" # you can also add elements to the list as they are mutable.
print(list)
```

## Some basic List methods

```python
# add item to a list
list.append("India")
list
```

```python
# Method 1 add multiple elements using .extend()
my_list = [1, 2, 3]
my_list.extend([4, 5, 6])  # Adds multiple elements
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# Method 2 using '+' concatenation
my_list = [1, 2, 3]
my_list = my_list + [4, 5, 6]  # Creates a new list
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# Method 3 using a loop
my_list = [1, 2, 3]
for item in [4, 5, 6]:
    my_list.append(item)
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# sort elements
list1 = [10, 24, 565, 23, 1, 0, -1]
list1.sort()
list1
```

```python
# sort in reverse
list1.sort(reverse = True)
list1
```

```python
# reverse the elements in the list
list1.reverse()
list1
```

```python
# Insert element at index
list1.insert(2, "Hello")
list1
```

```python
# Remove the first occurance of element.
list1.remove("Hello")
list1
```

```python
# Removes the element at index
list1.pop(3)
list1
```

```python
# Method 1
# Removing Multiple Items from a List. Using list comprehension
my_list = [1, 2, 3, 4, 5, 6]
my_list = [x for x in my_list if x not in [3, 5]]  # Removes 3 and 5
print(my_list)  # Output: [1, 2, 4, 6]
```

```python
# Method 2
# Using a Loop with .remove()
my_list = [1, 2, 3, 4, 5, 6]
for item in [3, 5]:  
    my_list.remove(item)  # Removes elements
print(my_list)  # Output: [1, 2, 4, 6]
# ⚠️ Raises an error if an item is not in the list.
```

 ✔ Use .extend() for lists and .update() for dictionaries when adding multiple items efficiently.
 
✔ Use list comprehension or dictionary comprehension for removing multiple items.

```python
waypoints = ["Eagle's Crest Trailhead", "Crystal Brook Bend", "Echo Canyon",
"Misty Falls", "Mount Silvercrest Summit"]

# enter new item one by one at the end of the list.
waypoints.append("Starwatch Lookout")
waypoints.append("Gem Lake")
waypoints.append("Eagle's Crest Trailhead")

# to add a item at a specific position.
waypoints.insert(6, "Piper's Cave")

# to remove an item.
waypoints.remove("Crystal Brook Bend")

# to delete an item at the given index
waypoints.pop(4) # method 1

del waypoints[4] # method 2

print(waypoints)

```

WAP to check if a ist contains a palindrome of elements. (Hint: use copy() mehtod)

```python
lst_1 = [1, 2, 3, 2, 1]

lst_2 = lst_1.copy()
lst_2.reverse()

# Compare original and reversed lists
if lst_1 == lst_2:
    print("The list is a palindrome.")
else:
    print("The list is not a palindrome.")
```

```python
# There are two ways to copy a list
m = [1, 2, 3, 4, 5, 6, 7, 8]
# Method 1:
n = n[:]
print(m)
```
