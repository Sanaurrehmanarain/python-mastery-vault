A multidimensional list is a list containing other lists as elements. Each inner list can have multiple values.

```python
seasonal_foods = [
  ["artichokes", "asparagus", "peas"], # Spring foods
  ["corn", "cucumbers", "tomatoes"], # Summer foods
  ["butternut squash", "sweet potatoes", "pumpkins"], # Fall foods
  ["kale", "leeks", "turnips"] # Winter foods
]
```

We can access the elements in a multidimensional list by specifying multiple indexes like this...

```python
print(seasonal_foods[2][1])
```

The [2] corresponds to the sub-list at the 2nd index ["butternut squash", "sweet potatoes", "pumpkins"]. The [1] corresponds to the 1st index in the sub-list "sweet potatoes".

Elements can be added, inserted, and removed just like regular lists...

```python
seasonal_foods[1].append("zucchini")
print(seasonal_foods)
```

```python
seasonal_foods.remove(["kale", "leeks", "turnips"])
print(seasonal_foods)

# or you can use pop method to delete the list at index 2
# seasonal_foods.pop(3)
```

The zip() function
As we saw in Exercise 3, multidimensional lists can be used to store related data in a structured way. However, it can sometimes be cumbersome to create and use multidimensional lists, especially in scenarios where we want to group data that appears in the same index across lists. For example, the 0th index in each of the following lists stores information related to the employee Alice.

```python
employees = ["Alice", "Bob", "Charlie", "Dana"]
employee_ids = [1234, 1111, 5678, 9090]
employee_roles = ["Project Manager", "Intern", "Developer", "CEO"]
```

To accomplish this, Python provides a built-in function called zip() which takes the lists whose indexes you would like to match as arguments.

```python
zip(employees, employee_ids, employee_roles)
```

As you can see, the zip() function returns a similar output to the reversed() function we saw in Exercise 2 Part B. We can cast this return value to a list to reveal our desired output...

```python
employee_data = list(zip(employees, employee_ids, employee_roles))
print(employee_data)
```

Great! Now we have all the data related to a given employee stored at its own index in the employee_data list. We can access this data using the normal index notation...

```python
print(employee_data[0])
```

However, notice that this data is surround with (). This is because ('Alice', 1011, 'Project Manager') is of a new data type!

```python
type(employee_data[0])
```

The zip() function matches list data by index and creates tuples. When we cast the return value to a list, we create a list of tuples.

Tuples may look similar to lists, but they are two distinct types of collections.
