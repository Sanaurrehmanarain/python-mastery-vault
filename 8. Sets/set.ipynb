# SETS
- Set is the collection of the unordered items. Set itself is mutable.
- Each element in the set must be unique & immutable.

1. "Set is the collection of unordered items."
- A set in Python does not maintain the order of elements.
- Example:

```python
my_set = {3, 1, 4, 2}
print(my_set)  # Output order may vary, e.g., {1, 2, 3, 4}
```

2. "Set itself is mutable."
- You can add or remove elements from a set.
- Example:

```python
my_set = {1, 2, 3}
my_set.add(4)  # Adding a new element
my_set.remove(2)  # Removing an element
print(my_set)  # Output: {1, 3, 4}
```

3. "Each element in the set must be unique & immutable."
- Unique: A set cannot have duplicate elements.
- Immutable: Elements must be of types that cannot change (e.g., numbers, strings, tuples).
- Example

```python
my_set = {1, 2, 2, "hello", (3, 4)}
print(my_set)  # Output: {1, 2, "hello", (3, 4)} (no duplicate 2)
```

- Lists and dictionaries cannot be elements in a set because they can change, violating immutability.
- Invalid Example (mutable elements like lists cannot be in a set):

```python
my_set = {1, [2, 3]}  # ❌ TypeError: unhashable type: 'list'
```

```python
# list can hold multiple duplicates.
fruits = ["orange", "banana", "apple", "orange", "kiwi", "apple", "strawberry", "mango", "banana"]
print(fruits)
```

```python
# set can not hold duplicates.
print(set(fruits))
```

sets are mutable / modifiable and sets are unordered

to see the difference between two sets.

```python
name_1 = {"Sana", "Sami", "Roomi", "Umar", "Yahya", "Hamza"}
name_2 = {"Maha", "Yahya", "Taha", "Hamza"}

new_name = name_2.difference(name_1)

print(new_name)
```

to check the same items in both sets. 

```python
new_name_1 = name_2.intersection(name_1)
print(new_name_1)
```

## Set Methods

```python
# adds an element
c = set()
c.add(1)
```

```python
# add multiple elements
c.update([2, 3, 4, 5])
```

```python
# add using a loop
d = set()

for item in ["a", "b", "c"]:
    d.add(item)

print(d)
```

```python
# removes the element
c.remove(3)
```

```python
# removes multiple elements
c.difference_update([3, 4])  # Removes 3 and 4
print(c)  # Output: {1, 2, 5}
```

```python
# removes using loop
c = {1, 2, 3, 4, 5}
for item in [3, 4]:  
    c.discard(item)  # Removes if exists, no error if missing
print(c)  # Output: {1, 2, 5}
```

```python
# empties the set
c = {1, 2, 3, 4, 5}

c.clear()

print(c)
len(c)
```

```python
# removes a random value
c = {1, 2, 3, 4, 5}

c.pop()
c
```

```python
# Combines both set values & returns new
k = {1, 2, 3, 4, 5, 6}
l = {5, 6, 7, 8, 9, 10}

k.union(l)
```

```python
# Combines common values & returns new
k.intersection(l)
```
