# Set Comprehensions
What is a Set Comprehension?

- Similar to list comprehensions, but creates a set instead of a list.

- Uses curly braces `{}` instead of `[]`.

- A set is an unordered collection of unique elements (no duplicates).

✅ Example 1: Basic Set Comprehension

```python
numbers: list[int] = [1, 2, 2, 3, 4, 4, 5]
squared_set: set[int] = {n**2 for n in numbers}

print(squared_set)
```

✅ Example 2: With a Condition

```python
data: range = range(10)
even_squares: set[int] = {n**2 for n in data if n % 2 == 0}

print(even_squares)
```

✅ Example 3: Strings with Set Comprehension

```python
word: str = "banana"
unique_chars: set[str] = {char for char in word}

print(unique_chars)
```

🔹 When to Use Set Comprehension

- When you want to remove duplicates automatically.

- When the order of elements doesn’t matter.

- Useful for tasks like extracting unique values, deduplicating data, or filtering with conditions.

```python
values: list[int] = [5, 5, 5, 11, 11, 5, 20, 30, 30, 5, 11, 2, 2, 100]
filtered: set[int] = {v for v in values}
print(filtered)
```

```python
filtered = set(values)
print(filtered)
```

both give the same result, and `set(values)` is shorter.

🔹 When to Use set(values)

- Best for just removing duplicates.

- Simple, fast, and more readable when you don’t need extra logic.

- Example:

```python
values = [5, 5, 11, 20, 30]
unique_values = set(values)
print(unique_values)
```

🔹 When to Use `Set Comprehension` `{v for v in values}`

- Gives you more control:

    1. You can filter items.

    2. You can transform items before putting them in the set.

✅ Example 1: Filtering

```python
# Only keep numbers greater than 10
filtered = {v for v in values if v > 10}
print(filtered) 
```

✅ Example 2: Transformation

```python
# Store squares instead of original numbers
squares = {v**2 for v in values}
print(squares)  
```

✅ Example 3: Combined

```python
# Square only even numbers
even_squares = {v**2 for v in values if v % 2 == 0}
print(even_squares)
```
