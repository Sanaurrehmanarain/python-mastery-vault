Define a collection called new_book that will contain the following information: "The Hunger Games," Suzanne Collins, 2008.

Note that book information doesn't change over time, but the order in which the information is stored within the collection is important. Later we will rely on the fact that the title is the 0th index, the author is the 1st index, and the publication year is the 2nd index.

```python
new_book = ("The Hunger Games", "Suzanne Collins", 2008)
```

When working with collections of items, it can be useful to assign each item in the collection to a variable name. Of course, we can accomplish this by assigning each index to a variable on separate lines of code...

```python
person = ["Alice", 30, "New York"]

name = person[0]
age = person[1]
city = person[2]

print(f"{name} is {age} years old and lives in {city}.")
```

But, it can be tedious and inefficient to write out variable assignments for each element in a collection. Luckily, Python provides an easier way of doing this called "unpacking". We can unpack a collection on a single line of code like this...

```python
person = ["Bob", 46, "London"]

name, age, city = person

print(f"{name} is {age} years old and lives in {city}.")
```

Unpack the new_book collection into the variables title, author, and year.

```python
title, author, year = new_book
```

The librarian maintains an inventory of the genres of the books that are currently checked out. Each time a book is checked out, its genre is added to this collection. When a book is checked back in, the book's genre is removed from the collection. Often, the librarian wants to count how many books of a certain genre are checked out.

These are the genres of the seven books that are currently checked out...

Science Fiction
Mystery
Non-Fiction
Fantasy
Science Fiction
Non-Fiction
Mystery
Define a collection called checked_out_genres to store this information.

```python
checked_out_genres = [
    "Science Fiction", 
    "Mystery", 
    "Non-Fiction", 
    "Fantasy", 
    "Science Fiction", 
    "Non-Fiction", 
    "Mystery"
]
print(checked_out_genres)
```

Two science-fiction books and a mystery book were just returned. Three biographies, two poetry books, and a fantasy book were just checked out.

Update checked_out_genres accordingly.

```python
checked_out_genres.remove("Science Fiction")
checked_out_genres.remove("Science Fiction")
checked_out_genres.remove("Mystery")

checked_out_genres.append("Biography")
checked_out_genres.append("Biography")
checked_out_genres.append("Biography")
checked_out_genres.append("Poetry")
checked_out_genres.append("Poetry")
checked_out_genres.append("Fantasy")

print(checked_out_genres)
```

The librarian is wondering how many fantasy and poetry books are currently checked out. Complete the following code to answer this question.

```python
fantasy_checked_out = checked_out_genres.count("Fantasy")  
poetry_checked_out = checked_out_genres.count("Poetry")

print(f"There are {fantasy_checked_out} fantasy and {poetry_checked_out} poetry books checked out.")
```

The librarian wants to compare the genres of books checked out to those not checked out to get a better sense of what people like to read.

First, pull out the unique values from checked_out_genres in Part B and store them back in the checked_out_genres variable.

```python
checked_out_genres = set(checked_out_genres)
print(checked_out_genres)
```

Create another collection called available_genres which stores the following unique genres of the books that are not currently checked out:

Thriller
Poetry
Drama
Biography
Adventure
Historical Fiction
Fantasy
Memoir
Philosophy

```python
available_generes = {
    "Thriller",
    "Poetry",
    "Drama",
    "Biography",
    "Adventure",
    "Historical Fiction",
    "Fantasy",
    "Memoir",
    "Philosophy"
}

print(available_generes)
```

Now that the librarian has a collection storing the unique genres of the books checked out and the books not checked out, they would like to know the following...

How many total genres does the library have (checked out and available genres)?

Which genres are both checked out and available?

Which genres are available but not checked out?

How many genres are checked out or available, but not both?

Use set operations and the variables checked_out_genres and available_genres to answer these questions.

```python
q1 = len(checked_out_genres.union(available_generes))
print(f"There are {q1} total genres.")

q2 = checked_out_genres.intersection(available_generes)
print(f"The {q2} genres are both checked out and available.")

q3 = available_generes.difference(checked_out_genres) 
print(f"The {q3} genres are available but not checked out.")

q4 = len(available_generes.symmetric_difference(checked_out_genres))
print(f"There are {q4} genres checked out or available, but not both.")
```
