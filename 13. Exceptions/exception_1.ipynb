An ``exception`` is an event, which occurs during the execution of a program, that disrupts the normal flow of the program.

Under ``try``, you write the code that might cause an error.

Under `except`, you handle the error gracefully.

````python
try:
    x = 10 / 0  # Code that may cause an error
except ZeroDivisionError:
    print("Cannot divide by zero!")  # Handling the error
````

```python
try:
    # Code that may raise an error
except:
    # Code that runs if an error occurs

General Structure
```python
try:
    # Risky code (e.g., file operations, user input, API calls)
except ExceptionType:
    # Handle the specific error (log it, show a message, retry, etc.)

````python
# example
try:
    x = 10 / 0  # This will cause a ZeroDivisionError
except:
    print("An error occurred!")  # Runs when an error happens
````

### Best Practice – Catch All with Logging
```python
try:
    risky_function()
except Exception as e:
    print(f"An error occurred: {e}")  # Logs the specific error

````python
a = float(input("number 1: "))
b = float(input("number 2: "))

try: # under try, write whatever you want to test
    result = a / b

except Exception as e: # under except, write the code to handle the error if an exception is thrown.
    print(f"An error occurred: {e}")

result
````

````python
# Multiple Exceptions Handling
try:
    num = int(input("Enter a number: "))  # Might cause ValueError
    result = 10 / num  # Might cause ZeroDivisionError
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero!")
````

## Raising an Exception

````python
class CoffeeCup:
    def __init__(self, temperature):
        self.__temperature = temperature

    def drink_coffee(self):
        if self.__temperature > 85:
            raise Exception("Coffee Too Hot!")  # Raising an Exception
        elif self.__temperature < 65:
            raise Exception("Coffee Too Cold!")  # Raising an Exception
        else:
            print("Coffee OK to Drink")

# Example Usage
try:
    cup = CoffeeCup(90)  # Too hot
    cup.drink_coffee()
except Exception as e:
    print(f"Error: {e}")
````

To create your own `Custom Exception Class`

````python
class CoffeeTemperatureError(Exception):
    """Custom exception for incorrect coffee temperature."""
    pass

class CoffeeCup:
    def __init__(self, temperature):
        self.__temperature = temperature

    def drink_coffee(self):
        if self.__temperature > 85:
            raise CoffeeTemperatureError("🔥 Coffee Too Hot! Let it cool down.")
        elif self.__temperature < 65:
            raise CoffeeTemperatureError("❄️ Coffee Too Cold! Reheat it.")
        else:
            print("☕ Coffee is perfect to drink!")

# Example Usage with Exception Handling
try:
    cup = CoffeeCup(90)  # Too hot
    cup.drink_coffee()
except CoffeeTemperatureError as e:
    print(f"Error: {e}")
````

````python
class CoffeeTooHotException(Exception):
    def __init__(self, msg):
        super().__init__(msg)

class CoffeeTooColdException(Exception):
    def __init__(self, msg):
        super().__init__(msg)

class CoffeeCup:
    def __init__(self, temperature):
        self.__temperature = temperature

    def drink_coffee(self):
        if self.__temperature > 85:
            raise CoffeeTooHotException("Coffee Temperature: " + str(self.__temperature))
        elif self.__temperature < 65:
            raise CoffeeTooColdException("Coffee Temperature: " + str(self.__temperature))
        else:
            print("Coffee OK To Drink")

# Handling exceptions
try:
    cup = CoffeeCup(100)
    cup.drink_coffee()
except CoffeeTooHotException as e:
    print("Too Hot:", e)
except CoffeeTooColdException as e:
    print("Too Cold:", e)
````
