### 🚨 New Concept: ' ' vs " "

Why does Python provide two methods for creating strings?

If your string contains an apostrophe ('), you must use double quotes (" ") to create your string. If your string contains double quotes (" "), you must use single quotes to create your string.

## Exercise 1 - When to create a string with '' or " "

You are helping out your favorite restaurant by writing a program that generates tonight's dinner menu. The first item on the menu is Vegetarian Shepard's Pie.

### Part A

Your first task is to create a string with the following sentence: We are having Vegetarian Shepard's Pie for dinner tonight.

**Solution**

Because the word *Shepard's* contains an apostrophe, we must use double quotes to create our string. This tells Python that the apostrophe is part of the string, not the end of it.

```python
"We are having Vegetarian Shepard's Pie for dinner tonight."
```

Otherwise we get the following...

```python
'We are having Vegetarian Shepard's Pie for dinner tonight.'
```

### Part B

Next, the head chef would like you to represent the list of ingredients for the Vegetarian Shepard's Pie in your code. Create a string with the following ingredients for the Vegetarian Shepard's Pie: onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn.

**Solution**

Because the word *"beef"* has double quotes, we must use single quotes to create our string. By using single quotes, we are telling Python that quotes around "beef" are considered part of the string.

```python
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn'
```

Otherwise we get the following...

```python
"onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn"
```

## Exercise 2 - Indexing and slicing

### Part A

The chef would like your program to retrieve certain ingredients from the ingredients string you created in Exercise 1. Let's first use indexing to retrieve the `g` in `ground "beef"`.

**Solution**

The first comma is located at index 5. Remember to start counting indexes from 0 and not 1.

```python
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn'[7]
```

### Part B

Next, use slicing to retrieve the characters `ground "beef"` from the ingredients string.

**Solution**

To retrieve `ground "beef"`, the index we wish to start at is 7 (which corresponds to the `g` character) and the index we wish to retrieve up to but not including is 20 (which is one index beyond the second `"` character).

```python
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn'[7:20]
```

### Part C

### 🚨 New Concept: Slicing shorthand notation

Python also provides a shorthand way of specifying indexes when we wish to include every index from the beginning up until a point in the string. For example...

```python
'my string'[:4]
```

Using the opposite syntax, we can also create a slice from a point in the string until the end.  This is especially useful if you do not know how long a string will be, so you don't know the last index to specify. For example...

```python
'my string'[4:]
```

### Question

Use this shorthand notation to output two slices: `onion` and `spices, tomato paste, veggie broth, peas, carrots, corn`.

**Solution**

```python
# onion
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn'[:5]
```

```python
# spices, tomato paste, veggie broth, peas, carrots, corn
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn'[22:]
```

## Exercise 3 - Multiple concatenations and using * for string repetition

### 🚨 New Concept: Multiple concatenations

Using the concatenation operator `+`, you can join together as many strings as you would like.

### Part A

The chef has just notified you that the Shepard’s Pie recipe calls for a few more ingredients! Now, we need to add "green beans", "mashed potatoes", and "cheese" to the ingredients list. In a single line of code, concatenate `'green beans'`, `'mashed potatoes'`, and `'cheese'` to the ingredients string.

**Solution**

Multiple concatenation statements can be used in sequence to join several strings together. For better formatting, you can add a comma and space before each ingredient.

```python
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn' + ', green beans' + ', mashed potatoes' + ', cheese'
```

### Part B

### 🚨 New Concept: Multiple concatenations



Similar to the `+` operator which performs different operations when applied to , the `*` operator has a different purpose when applied to string data. When applied to string data, the `*` operator concatenates multiple copies of the string it is applied to. For example...

```python
'a' * 3
```

### Question

The chef also notes that the recipe calls for two cloves of garlic, so we need to add it to the ingredients list twice. Using the `*` operator to duplicate `"garlic clove"`, concatenate the result to the list of ingredients you created in Part A.

**Solution**

You can add the comma and space before the words "garlic clove" for better formatting.

```python
'onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn' + ', green beans' + ', mashed potatoes' + ', cheese' + ', garlic clove' * 2
```

### Part C

Finally, measure the length of your new ingredients string in Part B.

**Solution**

```python
len('onion, ground "beef", spices, tomato paste, veggie broth, peas, carrots, corn' + ', green beans' + ', mashed potatoes' + ', cheese' + ', garlic clove' * 2)
```

The `len()` function can be applied to strings with concatenation operations.
