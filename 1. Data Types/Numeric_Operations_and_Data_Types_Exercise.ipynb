The minor difference between the expected value (12.65) and the value reported by Python (12.64999999999999) is due to how Python handles floating-point numbers. Python approximates floating-point numbers, which can lead to slight variations from the expected values when performing calculations. It's important to note that these discrepancies don't always happen. In practice, such small differences are typically negligible and can often be ignored.

## Exercise 2 - Order of operations

### 🚨 New Concept: Order of operations

In order to ensure the proper order of operations, we can group terms in an expression with parenthesis. This works the same as if you were doing a math problem.

For example...

```python
2 + 3 * 4
```

has a different result than...

```python
(2 + 3) * 4
```

As a reminder, the mathematical order of operations is...
1. Parenthesis
2. Exponents
3. Multiply
4. Divide
5. Add
6. Subtract

Because Python follows this order, we get 14 in the first example. 3 is multiplied by 4 first to get 12. Then, the 2 is added to get 14. In the second example, we evaluate the addition expression within the parenthesis first to get 5, and then multiply 5 by 4 to get 20.

### Part A
The bakery has two ovens, each can bake 25 cookies at a time. The baker needs to bake 200 cookies and each batch takes 15 minutes. How long will it take to bake all cookies?

To solve this problem, you must use parenthesis to group numbers together.

```python
(200 / (25 * 2)) * 15
```

If you do not use parenthesis you get the following incorrect result.


```python
200 / 25 * 2 * 15
```

```python
type((200 / (20 * 3)) * 15)
```


## Exercise 3 - Floor division and modulus




### Part A

The baker has 250 grams of chocolate chips and wants to evenly distribute them into batches of cookies, with each batch using 40 grams. How many full batches can they make, and how many grams of chocolate chips will remain?

Use the floor division operator to get the whole number of batches.

```python
# Full batches
250 // 40
```

Use the modulus operator to get the number of grams of chocolate chips left over.

```python
# Remaining grams
250 % 40
```

### Part B

The baker is also making pastries. They have 500 grams of dough and each pastry requires 37 grams of dough. How many pastries can the baker make, and how much dough will be left over?

Use the floor division operator to get the whole number of pastries.

```python
# Pastries
500 // 37 # // this gives the quotient
```

Use the modulus operator to get the number of grams of dough left over.

```python
# Leftover dough
500 % 37 # % this gives the remainder
```

You can use `divmod()` to get the `quotient //` and `remainder %` at the same time.

```python
divmod(500, 37)
```


## Exercise 4 - Exponentiation

The type of yeast that the baker uses doubles in mass every hour. If they start with 3 grams of yeast, how much yeast will they have after 6 hours?


```python
# Yeast mass after 6 hours
3 * 2**6
```

Use the exponent operator to raise 2 to the power of 6. This captures the doubling every hour for six hours. Then multiply that by 3 to account for the three grams of yeast the baker starts with.

For example, if we wish to take the cubed root of 27 we would write...
