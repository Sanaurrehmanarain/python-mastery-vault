# Practice with Python f-Strings

This notebook will help you explore additional formatting options using f-strings in Python.
Each section introduces a new feature you can try!

```python
temp = 67

f'The temperature is {temp} degree Fahrenheit'
```

We can directly convert the temp from F to C within the f string. So, no need to convert in the variable.

```python
f'The temperature is {(temp - 32) / 1.8} degree Celsuis'
```

You can round the decimals.

```python
f'The temperature is {(temp - 32) / 1.8:.2f} degree Celsuis'
```

### Big numbers with f-strings

```python
temp_sun = 27000000 # sun core temp is 27 million degree farhenheit
f'The temperature is {(temp_sun - 32) / 1.8:.2f} degree Celsuis'
```

f-string allows us to use seperators to handle the big numbers.

```python
f'The temperature is {(temp_sun - 32) / 1.8:,.2f} degree Celsuis'
```

```python
temp_sun = 27000000
temp_earth = 58

print(f'The temperature is {(temp_sun - 32) / 1.8:,.2f} degree Celsuis')
print(f'The temperature is {(temp_earth - 32) / 1.8:,.2f} degree Celsuis')
```

```python
print(f'The temperature is {(temp_sun - 32) / 1.8:*^15,.2f} degree Celsuis')
print(f'The temperature is {(temp_earth - 32) / 1.8:*^15,.2f} degree Celsuis')
```

## 1. Binary, Octal, and Hexadecimal
You can display numbers in different bases using `b`, `o`, or `x`:

```python
num = 42
print(f"Binary: {num:b}")
print(f"Octal: {num:o}")
print(f"Hexadecimal: {num:x}")
```

## 2. Padding Numbers with Zeros
Use `0` before width to pad numbers with zeros:

```python
score = 7
print(f"Score: {score:03}")
```

## 3. Sign Formatting
Use `+` to always show sign:

```python
pos_balance = 42
neg_balance = -21
print(f"Normal: {pos_balance} | {neg_balance}")
print(f"Signed: {pos_balance:+} | {neg_balance:-}")
```

## 4. Number Grouping with Underscores
Underscores can be used instead of commas (e.g., for code readability):

```python
population = 83200000
print(f"Comma: {population:,}")
print(f"Underscore: {population:_}")
```

## 5. Percentage with Precision
You can specify how many decimals to show in percentages:

```python
ratio = 0.72345
print(f"Rounded to 1 decimal: {ratio:.1%}")
print(f"Rounded to 2 decimals: {ratio:.2%}")
```
