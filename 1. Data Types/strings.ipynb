## ESCAPE SEQUENCES

When writing out a longer string, it is often desirable to split the string up into multiple lines or insert a tab for formatting. Python provides us with 
special character sequences to accomplish tasks like this. These sequences always begin with the escape character backslash \. To tell Python that we could like a new line to be inserted in our string when outputted, we can use the new line character sequence \n like this

Escape sequences start with a backslash (\) and have special meanings. Some common ones:

- \n → Newline
- \t → Tab space
- \\ → Backslash
- \' → Single quote
- \" → Double quote

Example of escape sequences:

```python
print("Hello\nWorld")   # Prints on two lines
print("Python\tRocks")  # Adds a tab space
print("This is a backslash: \\")  # Prints a backslash
```

Additionally, if you want to include ' O ' characters within a string and do not want to worry about which way you need to declare the string, you can escape single and double quotes with the backslash like this...

```python
print("She said \"My name is Mary O\'Connor\".")
```

```python
print('Hello, my name is "Sana".')
```

```python
print("c:\name\drive") # This will not give the required output, as python think its `\n` new line. 

print(r"c:\name\drive") # Correct approach, r => raw string
```

In Python, end=" " is an argument used in the print() function to control what is printed at the end of the output instead of the default newline (\n).

```python
# Example Without end (Default Behavior):
print("Hello")
print("World")
```

```python
# Example With end=" " (Custom Behavior):
print("Hello", end=" ")
print("World")
```

Here, end=" " replaces the default newline with a space, keeping both prints on the same line.

```python
print("I love Python.", "My name is Sana.", end = "\n") # end="\n" is the default. You write it or not, the output will be the same.
```

```python
print("I love Python\n", "My name is Sana")
```

```python
print("I love Python", "My name is Sana", sep = " & ")
```

### Multi Line String
A multi-line string is a string that spans across multiple lines. You can create one using triple quotes — either ''' or """. Here's a simple example:

```python
multi_line_string = """This is a multi-line string.
It can span across several lines.
Each line is part of the same string."""
print(multi_line_string)
```

```python
print('''This following options are available:
                -a      : does nothing
                -b      : also does nothing.
''')
```

```python
a = """This is line 1.
This is line 2.
This is the last line and this is line 3.
"""

print(a)
```
