Types of Operators:

An operator is a symbol that performs a certain operation between operands.
- Arithmetic Operators (+, -, *, /, %, **)
- Relational / Comparison Operators (==, !=, >, <, >=, <=)
- Assignment Operators (=, +=, -=, *=, /=, %=, **=)
- Logical Operators (not, and, or)

Write a program to input 2 numbers & print their sum.

```python
num1 = float(input("Enter 1st num: "))
num2 = float(input("Enter 2nd num: "))

sum = num1 + num2

print(f"The sum is {sum}.")
```

Write a program to input side of a square & print its area.

```python
side = float(input("Enter square side: "))

print("area = ", side * side)
```

Write a program to input 2 floating point numbers & print their average.

```python
a = float(input("Enter the 1st num: "))
b = float(input("Enter the 2nd num: "))

avg = a + b / 2

print("The average is: ", avg)
```

Write a program to input 2 int numbers, a and b. Print True if a is greater than or equal to b. If not print False.

```python
a = int(input("Enter the 1st num: "))
b = int(input("Enter the 2nd num: "))

print(a >= b)
```

Assigning the same value to multiple variables

```python
a = b = c = 10  # All three variables will have the value 10
```

Assigning Different Values to Multiple Variables

```python
x, y, z = 1, 2, 3  # x = 1, y = 2, z = 3
```

Swapping Two Variables Without a Temporary Variable

```python
a, b = b, a  # Swaps the values of a and b
```

Using a List or Tuple for Multiple Assignments

```python
values = [5, 10, 15]
x, y, z = values  # x = 5, y = 10, z = 15
```
