# Regular Expression

A **regular expression (regex)** in Python is a sequence of characters that defines a search pattern, usually for working with text. You use it to check if a string matches a pattern, extract parts of text, or replace text that matches a certain rule.

Python provides the `re` module to work with regular expressions.

## Common Uses of Regex

1. Searching – Check if a string contains a pattern.

2. Extracting – Pull out specific parts of a string (like emails, dates, phone numbers).

3. Replacing – Substitute text that matches a pattern.

4. Splitting – Break a string by a pattern.

```python
import re

# Simple pattern
pattern = r"\d+"   # one or more digits

text = "My order number is 12345."

# Search for the first match
match = re.search(pattern, text)
if match:
    print("Found:", match.group())   

# Find all matches
numbers = re.findall(pattern, text)
print(numbers)  
```

### Key Regex Symbols

- `.` → any character except newline

- `\d` → digit (0–9)

- `\w` → word character (letters, digits, underscore)

- `\s` → whitespace

- `+` → one or more

- `*` → zero or more

- `?` → zero or one

- `^` → start of string

- `$` → end of string

- `[]` → match any character inside (e.g., `[abc]`)

- `|` → OR

```python
import re
```

## 2. Basic Patterns
Digits `\d` and Words `\w`

```python
import re

text = "Order ID: 4567, Customer: Sana_89"

# Find all digits
digits = re.findall(r"\d+", text)
print("Digits:", digits)

# Find all words
words = re.findall(r"\w+", text)
print("Words:", words)
```

## 3. Special Symbols

```python
text = "cat bat rat mat"

# Find words ending with 'at'
pattern1 = r"\w+at"
print(re.findall(pattern1, text))

# Find only 'cat' or 'rat'
pattern2 = r"cat|rat"
print(re.findall(pattern2, text))
```

## 4. Extracting Email Addresses

```python
text = "Contact us at support@example.com or sales@company.org"

emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print("Emails:", emails)
```

## 5. Extracting Phone Numbers

```python
text = "Call me at +46-123-456789 or 08-123456"

phones = re.findall(r"\+?\d{1,3}-?\d{2,3}-?\d{5,}", text)
print("Phone numbers:", phones)
```

## 6. Splitting Text

```python
text = "apple, orange; banana | mango"
fruits = re.split(r"[;|,]\s*", text)
print("Fruits:", fruits)
```

## 7. Substitution (Replacing Text)

```python
text = "My password is 12345"
safe_text = re.sub(r"\d+", "*****", text)
print(safe_text)
```

## 8. Useful Cheatsheet

| Pattern  | Meaning                          | Example Match       |
| -------- | -------------------------------- | ------------------- |
| `\d`     | Digit (0-9)                      | `3`                 |
| `\w`     | Word char (a-z, A-Z, 0-9, \_)    | `Sana_89`           |
| `\s`     | Whitespace (space, tab, newline) | `" "`               |
| `.`      | Any char except newline          | `a`, `9`, `#`       |
| `^abc`   | Starts with `abc`                | `abc123`            |
| `abc$`   | Ends with `abc`                  | `123abc`            |
| `a*`     | 0 or more `a`                    | `aaa`, \`\`         |
| `a+`     | 1 or more `a`                    | `aaa`               |
| `a?`     | 0 or 1 `a`                       | `a`, \`\`           |
| `[abc]`  | One of `a, b, c`                 | `a`                 |
| `[^abc]` | Not `a, b, c`                    | `d`, `z`            |
| `a{2,4}` | 2 to 4 `a`                       | `aa`, `aaa`, `aaaa` |
