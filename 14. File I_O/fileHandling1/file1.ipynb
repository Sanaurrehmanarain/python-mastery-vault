# FILE I/O
Python can be used to perform operations on a file. (read & write data)

Types of all files:

- Binary files: .mp4, .mov, .png, .jpeg etc.

### Reading a file

````python
# Method 1
f = open("demo.txt", "r")
data = f.read() # Reads entire file
print(data)
print(type(data))
f.close()
````

````python
# Method 2
# Using with open(...) automatically closes the file after execution.
with open("demo.txt", "r") as f:
    data = f.read()
    print(data)
````

````python
f = open("demo.txt", "r")
data = f.read(5) # Reads only first five characters including spaces
print(data)
f.close()
````

````python
# Reads one line at a time
f = open("demo.txt", "r")
line1 = f.readline()
print(line1)
````

````python
f.close() # To close the file
````

### Writing to a file

````python
f = open("demo.txt", "w")
data = f.write("I also want to learn R. ") # This will overwrite the file. Means it will delete the old data and write this into the file.
f.close()
````

````python
# Better version
with open("test.txt", "w") as f:
    f.write("Hello, world!")
````

### Appending to a file

````python
f = open("demo.txt", "a")
data = f.write("\nThis is a demo file. \nI am learning python. \nIt is a fantastic first programming language to learn.") # This will append the data at the end.
f.close()
````

````python
# Better version
with open("test.txt", "a") as f:
    f.write("\nThis is an additional line.")
````

# File I/O in Python  

## File Modes  

In Python, the `open()` function is used to work with files. The argument mode specifies how the file should be opened.  

### Common File Modes  

| Mode  | Description  |
|--------|----------------------------------|
| **`r`**  | Open text file for reading. The stream is positioned at the beginning of the file. |
| **`r+`** | Open for reading and writing. The stream is positioned at the beginning of the file. |
| **`w`**  | Truncate file to zero length or create a new text file for writing. The stream is positioned at the beginning of the file. |
| **`w+`** | Open for reading and writing. The file is created if it does not exist; otherwise, it is truncated. The stream is positioned at the beginning of the file. |
| **`a`**  | Open for writing. The file is created if it does not exist. The stream is positioned at the end of the file. Subsequent writes will always append to the end of the file. |
| **`a+`** | Open for reading and writing. The file is created if it does not exist. The stream is positioned at the end of the file. Subsequent writes will always append to the end of the file. |

Best Practices

- Always use with open(...) to ensure proper resource management and it will automatically closes the file after execution.
- Use f.read() to read the entire file, or f.readlines() to get a list of lines.
- When writing, be aware that "w" mode overwrites existing content, while "a" mode appends to it.

Q: WAF to find in which line of the file does the word "learning" occur first.

Print -1 if word not found.

````python
# Method 1
def check_for_line():
    word = "learning"
    data = True
    line_no = 1
    with open("practice.txt", "r") as f:
        while data:
            data = f.readline()
            if word in data:
                print(line_no)
                return
            line_no += 1
    return -1

check_for_line()
````

````python
# Method 2
def check_for_line():
    word = "learning"
    line_no = 1
    with open("practice.txt", "r") as f:
        for line in f:
            if word in line:
                return line_no  # Return the first occurrence
            line_no += 1
    return -1  # If the word is not found, return -1

print(check_for_line())
````

Q: From a file containing numbers separated by comma, print the count of even numbers.

````python
with open("numbers.txt", "r") as f:
    data = f.read()
    print(data)
    print(type(data))
````

````python
nums = data.split(",")
print(nums)
````

````python
count = 0
for value in nums:
    if int(value) % 2 == 0:
        count += 1

print(count)
````
