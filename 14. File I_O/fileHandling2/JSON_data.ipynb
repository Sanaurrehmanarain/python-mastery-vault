`JSON` (JavaScript Object Notation) is a lightweight data format used for storing and exchanging data. It is human-readable and easy for machines to parse.
In other words, JSON is a text format. JSON is a syntax for storing and exchanging data.

Example JSON:

```python
{
    "name": "Sana",
    "age": 25,
    "skills": ["Python", "Java"],
    "is_student": False
}
```

Key Features:
- Uses key-value pairs ("name": "Sana")
- Supports data types: strings, numbers, booleans, arrays, objects
- Commonly used in APIs and data interchange between systems

In Python, you can work with JSON using the json module.

Example:

```python
import json

data = '{"name": "Sana", "age": 25}'
parsed_data = json.loads(data)  # Convert JSON string to dictionary
print(parsed_data["name"])  
```

```python
a = {
    "name": "Maha",
    "Profession": "Doctor",
    "Age": "33",
    "Education": "MBBS"
}

print(json.dumps(a)) # convert some python values to JSON values
```

```python
print(json.dumps(a, indent = 4, sort_keys = True))
```

```python
# to save json file
with open("demo.json", "w") as fh:
    fh.write(json.dumps(a,indent = 2, sort_keys = True))
```

```python
# to read json files
with open("demo.json", 'r') as fh:
    print(fh.read())
```

```python
with open("demo.json", 'r') as fh:
    print(type(fh.read()))# fh.read func will return the string values
```

```python
# to convert the string into json value
with open("demo.json", 'r') as fh:
    json_str = fh.read()
    json_value = json.loads(json_str) # convert string to json value
    print(type(json_value))
    print(json_value["name"])
```
