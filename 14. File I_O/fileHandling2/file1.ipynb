## File Handling
Use open() function to open the file. The open() function takes two parameters; filename and mode

```python
fh = open("demo.txt", "w") # make/open and write/overwrite the file demo.txt
for i in range (10):
    fh.write("This is line no %d\n" % (i + 1)) 
fh.close() # close the file
```

```python
fh = open("demo.txt", "a") # append the file
for i in range (11, 21):
    fh.write("This is line no %d\n" % i) 
fh.close() 
```

```python
# This is a better method to open/create the file as it will close it automatically
with open ("demo.txt", "a") as fh:
    fh.write("Hello, how are you all?")
```

```python
with open("demo.txt", "r") as fh: # read the file
    print(fh.read())
```

```python
with open("demo.txt", "r") as fh:
    lines = fh.readlines()  # Read all lines
    
with open("demo.txt", "w") as fh:
    fh.writelines(lines[:5])  # Keep only the first 5 lines
```

```python
with open("demo.txt", "r") as fh: 
    print(fh.read(4)) # give the first four char
```

```python
with open("demo.txt", "r") as fh:
    print(fh.readline()) # read the first line
```

```python
with open("demo.txt", "r") as fh:
    print(fh.readline()) # read the first three lines
    print(fh.readline()) 
    print(fh.readline()) 
```

```python
with open("demo.txt", "r") as fh:
    print(fh.readlines()) # returns the list 
```

```python
with open("demo.txt", "r") as fh:
    print(fh.readlines()[4]) 
```

```python
with open("demo.txt", "r") as fh:
    for line in fh:
        print(len(line)) # give you the char of each line
```

```python
with open("demo.txt", "r") as fh:
    for line in fh:
        print(len(line.split(" "))) # give you the count of num of words
```
