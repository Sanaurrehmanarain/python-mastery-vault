##### 🚨 New Concept: The else clause
So far, we've learned that placing potentially error-prone code inside a try block allows us to handle errors and make our programs more resilient. For example, the following code puts the "risky" code in a try block, so if b = 0, we can catch and handle the error effectively:

```python
a = 10
b = 2

try:
    result = a / b
    print(f"The result is {result}")
except ZeroDivisionError:
    print(f"Error: Cannot divide by zero.")
```

However, not all the code within the try block is truly "dangerous." For instance, if the line result = a / b executes without error, the print statement that follows it won't cause any issues either. Therefore, including the print statement inside the try block isn't necessary. In situations like this, Python provides the else clause, which lets you define a block of code that runs only if no exceptions are raised in the try block.

For example, in the code below, we place the print statement inside an else clause. This way, only the "dangerous" code is inside the try block, and the print statement will run only if result = a / b executes without causing an error.

```python
a = 10
b = 2

try:
    result = a / b
except ZeroDivisionError:
    print(f"Error: Cannot divide by zero.")
else:
    print(f"The result is {result}")
```

This is a better programming practice because it clearly indicates which code should run only if there are no errors in the try block.

##### 🚨 New Concept: The raise keyword
The raise keyword in Python is used to manually trigger an exception in your code. It allows you to create and throw exceptions when needed.

For example, if we want to ensure that inches and feet quantities are positive when calculating height in inches, we can raise a ValueError:

```python
feet = 5
inches = -10

if feet <= 0 or inches <= 0:
    raise ValueError()

print(f"{feet} feet {inches} inches in inches is {feet * 12 + inches} inches.")
```

We can then handle these manually triggered exceptions using a try/except block:

```python
feet = 5
inches = -10

try:
    if feet <= 0 or inches <= 0:
        raise ValueError()
    print(f"{feet} feet {inches} inches in inches is {feet * 12 + inches} inches.")
except ValueError:
    print("Feet and inches should be positive.")
```

##### Part A
Currently, you have a translator that converts English words to Xylonian. However, since you're collaborating closely with the Xylonian scientists, you'd like to create a translator that also accepts Xylonian words and translates them into English.

Follow the steps below to write a program that accepts a single Xylonian word as input and translates it to English using the language_translator dictionary.

```python
# Step 1: Prompt the user to enter a word in Xylonian.
word = input("Enter a word in Xylonian to translate to English: ").strip().lower()

# Step 2: Initialize a variable to track if the word is found in the dictionary.
word_found = False

# Step 3: Loop through the language_translator dictionary to find the inputted word
for key, value in language_translator.items():
  # Step 4: If the current value matches the input word, print the translation and set 'found' to True.
  if value == word:
    print(f"The translation of '{word}' is '{key}'")
    word_found = True
    break # Step 5: Exit the loop since the word has been found.

# Step 6: After the loop, check if the word was not found in the dictionary.
try:
  
  # Step 7: If the word was not found, raise a ValueError to handle the case.
  if not word_found:
    raise ValueError()
except ValueError:
  # Step 8: Handle the ValueError.
  print(f"'{word}' is not in the translation dictionary.")
```

##### 🚨 New Concept: The as e notation
In Python, the as e notation is used in error handling to capture the exception for further processing. It is especially useful when raising custom exceptions.

For example, instead of printing a message in the except block like this:

```python
feet = 5
inches = -10

try:
    if feet <= 0 or inches <= 0:
        raise ValueError()
    print(f"{feet} feet {inches} inches in inches is {feet * 12 + inches} inches.")
except ValueError:
    print("Feet and inches should be positive.")
```

Python allows us to pass a string containing an error message to the error that we raise. Then,we use the as e notation to capture the exception text in the except statement and reference it inside the except block.

```python
feet = 5
inches = -10

try:
    if feet <= 0 or inches <= 0:
        raise ValueError("Feet and inches should be positive.")
    print(f"{feet} feet {inches} inches in inches is {feet * 12 + inches} inches.")
except ValueError as e:
    print(e)
```

While you can technically use any valid variable name instead of e, e is the standard convention among Python programmers.

##### Part B
Similar to your English-to-Xylon translator, you want your Xylon-to-English translator to handle phrases.

Follow the steps below to translate a phrase from Xylonain to English.

```python
# Step 1: Get the input phrase from the user in Xylonian and prepare it for translation
phrase = input("Enter a phrase in Xylonian to translate to English: ").strip().lower()

# Step 2: Split the input phrase into individual words
words = phrase.split()

# Step 3: Initialize a list to store the translated words
translated_phrase = []

# Step 4: Iterate through each word in the input phrase
for word in words:
  word_found = False
  
  # Step 5: Search for the word in the translation dictionary
  for key, value in language_translator.items():
    if value == word:
      translated_phrase.append(key)
      word_found = True
      break

  # Step 6: Handle cases where the word is not found in the dictionary, use as e notation
  try:
    if not word_found:
      raise ValueError(f"{word}' is not in the translation dictionary.")
  except ValueError as e:
    print(e)
    break

# Step 7: Check if all words were successfully translated and display the final translated phrase
if len(words) == len(translated_phrase):
  translated_phrase = " ".join(translated_phrase)
  print(f"The translated phrase is: {translated_phrase}")
```

##### Part C
As a final upgrade to your measurement converter program, you want to be able to handle multiple comma-separated measurement inputs and convert them.

Follow the steps below to write code that accepts comma-separated measurements as input.

```python
# Step 1: Get the input measurements from the user and prepare the input string for processing
input_string = input("Enter measurements (e.g., '5 meters, 10 kilograms'): ").strip().lower()

# Step 2: Split the input string into individual measurement entries
entries = input_string.split(", ")

# Step 3: Process each measurement entry using a loop
for entry in entries:
  
  # Step 4: Attempt to split each entry into an amount and a unit
  try:
    quantity, unit = entry.split()
  except ValueError as e:
    print(e)
    break
    
  # Step 5: Attempt to convert the amount to a floating-point number
  try:
    quantity_float = float(quantity)
  except ValueError as e:
    print(e)
    break

  # Step 6: Check if the unit is recognized in the measurement translator
  try:
    if unit not in measurement_translator:
      raise ValueError(f"'{unit}' is not a recognized unit.")
  except ValueError as e:
      print(e)
      break

  # Step 7: Perform the conversion using the translation dictionary and output the result
  converted = quantity_float * measurement_translator[unit]["conversion_factor"]
  print(f"{quantity_float} {unit} is {converted} {measurement_translator[unit]['xylonian_unit']}")
```
