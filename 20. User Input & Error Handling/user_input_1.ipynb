##### input()
prompts user to input data for program to process.
##### Remarks
1. Program execution is paused until user input is submitted
2. Text prompts should be clear and precise
3. The function only returns string data
4. Input should be immediately cast to the expected type
5. Provide confirmation that user input was received

```python
name = input("What is your name?")
print(f"Hello {name}!")
```

```python
age = int(input("What is your age?"))
print(f"You have been an adult for {age - 18} years.")
```

```python
name = input("What is your name? ")

match name:
    case "Harry" | "Ron" | "Hermione":
        print("Gryffindor")
    case "Draco" | "Neville" | "Voldemort":
        print("Slytherin")
    case _:
        print("You are not in Hogwarts!")
```
