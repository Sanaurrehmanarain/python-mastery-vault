### 🚨 New Concept: Truthiness and `bool()`

The concept of "truthiness" in programming refers to whether a given value evaluates to `True` or `False`. Boolean values have an obvious truthiness, but other Python values can be truthy or falsy as well.

Truthy means that a value evolutes to `True` and falsy means that a value evaluates to `False`.

Python provides the built-in function `bool()` which allows us to check whether a value is considered `True` or `False`.

For example, `0` is considered falsy...

```python
bool(0)
```

While any other numeric value is considered truthy...

```python
print(bool(1))
print(bool(18))
print(bool(123.4))
print(bool(-1))
```

Additionally, the empty string `''` is considered falsy...

```python
bool('')
```

While every other string value is considered truthy...

```python
print(bool('hello world'))
print(bool('a!'))
print(bool('123'))
print(bool(' '))
print(bool('False'))
```

## Exercise 1 - Expressions in comparison statements

You have just gotten a job as a data analyst at the National Weather Service. Congrats! On your first day, your boss asks you to find answers to several questions using their newest data set. The data set includes average temperature data for New York City.

```python
avg_temp_mon = 45 # Monday's average temperature
avg_temp_tues = 51 # Tuesdays's average temperature
avg_temp_wed = 53 # Wednesday's average temperature
avg_temp_thurs = 54 # Thursday's average temperature
avg_temp_fri = 49 # Friday's average temperature
```

### Part A

Determine if the average temperature early in the week (Monday-Wednesday) is greater than the average temperature later in the week (Thursday and Friday).

**Solution**

```python
avg_early = (avg_temp_mon + avg_temp_tues + avg_temp_wed) / 3
avg_late = (avg_temp_thurs + avg_temp_fri) / 2

print(f"It is {avg_early > avg_late} that the average temperature on Monday, Tuesday, and Wednesday is greater than the average temperature on Thursday and Friday.")
```

Or you can write it all inside the `print()` statement...

```python
print(f"It is {(avg_temp_mon + avg_temp_tues + avg_temp_wed) / 3 > (avg_temp_thurs + avg_temp_fri) / 2} that the average temperature on Monday, Tuesday, and Wednesday is greater than the average temperature on Thursday and Friday.")
```

### Part B

Determine if the day-to-day absolute temperature change between Monday and Tuesday is at least as large as the absolute temperature change between Thursday and Friday.

**Solution**

```python
result = abs(avg_temp_mon - avg_temp_tues) >= abs(avg_temp_thurs - avg_temp_fri)

print(f"It is {result} that the day to day temperature change between Monday and Tuesday is at least as large as the temperature change between Thursday and Friday.")
```

## Exercise 3 - The `not` operator

### 🚨 New Concept: The `not` operator

The `not` operator is another logical operator provided by Python like `and` and `or`. When applied to a comparison or logical expression, it negates the result. This means that if an expression is originally evaluated as `True` and we apply the `not` operator, then the expression will be evaluated as `False` and vice versa.  For example...

```python
print(True)
print(not True)
```

```python
print(True and False)
print(not (True and False))
```

Note that the parenthesis make a difference to the result when using the `not` operator. For example...

```python
print(not True and False)
```

This is because here the `not` operator is being applied only to the `True`, turning it into `False`. And of course, `False and False` is `False`.

Like the mathematical operations, logical operations have a presentence order. The `not` keyword has the highest precedence, followed by `and`, followed by `or`. This means that an expression like this...

```python
not True and False or True
```

Evaluates to `True` because the `not` is applied to the first `True` turning it `False`. Meaning, `not True and False or True` becomes `False and False or True`.

Then because the `and` operator has precedence over the `or` operator, the `False and False` portion of the expression is evaluated to `False`. Meaning, `False and False or True` becomes `False or True`.

Finally, we evaluate `False or True` to `True`.

### Question

You also have access to the weather data for Austin, Texas. Your boss asks you to determine if the National Weather Service should issue a heat advisory for the area.

Given the following variables...


```python
temp = 92
will_rain = False
temp_will_drop = False
```

Use the `not` operator to write a logical expression to evaluate whether a heat advisory should be issued. A heat advisory should be issued if the temperature is at least 90 degrees and it is not expected to rain and it is not expected to cool off.

**Solution**

We can write this expression in two different ways. We can check if `will_rain` and `temp_will_drop` are `False` separately by applying the `not` operator to both variables individually.

```python
should_issue_advisory = temp >= 90 and not will_rain and not temp_will_drop

print(f"It is {should_issue_advisory} that the National Weather Service should issue a heat advisory.")
```

We could also link `will_rain` and `temp_will_drop` with the `or` operator and apply the `not` operator to the group.

```python
should_issue_advisory = temp >= 90 and not (will_rain or temp_will_drop)

print(f"It is {should_issue_advisory} that the National Weather Service should issue a heat advisory.")
```

This logical equivalence is known as De Morgan's Laws. You can read more about them here if you are interested: https://brilliant.org/wiki/de-morgans-laws/.

For our purposes, De Morgan's laws say two things...
1. `not A or not B` is the same as `not (A and B)`
2. `not A and not B` is the same as `not (A or B)`

So in this problem, we can take advantage of the fact that `not A and not B` is the same as `not (A or B)` and turn `not will_rain and not temp_will_drop` into `not (will_rain or temp_will_drop)`.

## Exercise 4 - Advanced comparison and logical expressions

Alert! Alert! The national radar just picked up a storm brewing across the Midwest. You need to determine if it will be safe to drive during the storm.

For driving conditions to be considered safe, it needs to not be hailing and one of the following conditions must also be true...
- at least 500 yards of visibility and a precipitation rate of no more than 0.2 inches per hour
- no less than 600 yards of visibility and a precipitation rate of no more than 0.3 inches per hour

Given the following variables...

```python
is_hailing = False
visibility = 550
precipitation_rate = 0.32
```

Write an expression to determine if it is safe to drive.

**Solution**

```python
safe_to_drive = not is_hailing and ((visibility >= 500 and precipitation_rate <= 0.2) or (visibility >= 600 and precipitation_rate < 0.3))

print(f"It is {safe_to_drive} that it is safe to drive during the storm.")
```

Because the `and` operator takes precedence over the `or` operator, we can leave out the parenthesis around each of the `and` expressions.

```python
safe_to_drive = not is_hailing and (visibility >= 500 and precipitation_rate <= 0.2 or visibility >= 600 and precipitation_rate < 0.3)

print(f"It is {safe_to_drive} that it is safe to drive during the storm.")
```

## Exercise 5 - Testing the truthiness of values

### 🚨 New Concept: Placeholder values

In programming, we often want to create a variable before we have data to store in it. In Python, a variable must be set to some data when it is created, so programmers use placeholder values. Placeholder values are used to represent data that is incomplete, missing, or yet to be provided.

String variables are often given the empty string `''` as a placeholder value, while numeric data is often given the placeholder values of `0` or `-1`.

Additionally, Python provides the keyword `None` which also represents the absence of a value. You can set a variable equal to `None` just as you would expect...

```python
my_var = None
print(my_var)
```

However, you should note that `None` is a value of its own type: `NoneType`. If you choose to use it as a placeholder value, be mindful that even though it looks like a string when outputted, it does not always act like a string.

```python
type(my_var)
```

Further, `None` is falsy...

```python
bool(None)
```
