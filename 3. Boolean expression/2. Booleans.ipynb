Write a boolean function, is_between(x, y, z), that returns True if x < y < z or if
z < y < x, and False otherwise.

```python
def is_between(x, y, z):
    if x < y < z or z < y < x:
        return True
    else:
        return False
```

```python
is_between(10, 15, 20)
```
